---

## ğŸ§  Key Observations

* Brute force generates `len(nums1) * len(nums2)` pairs â†’ âŒ too slow
* Arrays are **already sorted**
* Smallest sums come from **small indices**
* This is a classic **k-way merge using a min-heap**

---

## ğŸš€ Optimal Strategy (Min-Heap + Index Tracking)

### Core Idea

Treat each element in `nums1` as the **start of a sorted list**:

```
(nums1[i] + nums2[0]), (nums1[i] + nums2[1]), ...
```

Then merge these lists using a **min-heap**, similar to merging k sorted lists.

---

## âœï¸ Short Approach Sketch

1. Push pairs `(nums1[i] + nums2[0], i, 0)` into a min-heap
   (only first `min(k, len(nums1))` rows)
2. While heap not empty and result size < `k`:

   * Pop smallest sum
   * Add the pair to result
   * Push next pair from same row `(i, j+1)`
3. Stop after collecting `k` pairs

---

## â±ï¸ Complexity

* **Time**: `O(k log k)`
* **Space**: `O(k)`
* Optimal and interview-approved

## ğŸ§© Python Solution (Optimal)

```python
import heapq

class Solution:
    def kSmallestPairs(self, nums1, nums2, k):
        if not nums1 or not nums2 or k == 0:
            return []
        
        min_heap = []
        res = []
        
        # Initialize heap with pairs (nums1[i], nums2[0])
        for i in range(min(k, len(nums1))):
            heapq.heappush(min_heap, (nums1[i] + nums2[0], i, 0))
        
        while min_heap and len(res) < k:
            _, i, j = heapq.heappop(min_heap)
            res.append([nums1[i], nums2[j]])
            
            # Push next pair in the same row
            if j + 1 < len(nums2):
                heapq.heappush(
                    min_heap,
                    (nums1[i] + nums2[j + 1], i, j + 1)
                )
        
        return res
```

# My Solution

````python
import heapq

class Solution(object):
    def kSmallestPairs(self, nums1, nums2, k):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :type k: int
        :rtype: List[List[int]]
        """
        return_arr = []
        heap = []
        i = 0
        while i < min(k, len(nums2)):
            heapq.heappush(heap, (nums1[0]+nums2[i], 0, i))
            i += 1
        while len(return_arr) < k :
            _, i, j = heapq.heappop(heap)
            return_arr.append([nums1[i], nums2[j]])
            if i + 1 < len(nums1):
                heapq.heappush(heap, (nums1[i+1]+nums2[j], i+1, j))

        return return_arr
````
