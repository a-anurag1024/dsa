---

#### ðŸ” Problem Recap

Youâ€™re given an `m Ã— n` grid of **non-negative numbers**.

* Start at **top-left** `(0,0)`
* End at **bottom-right** `(m-1,n-1)`
* You can move **only right or down**

ðŸ‘‰ **Goal:** Find the **minimum sum path**.

---

### ðŸ’¡ Core Insight

This is a **classic 2D Dynamic Programming grid problem**.

To reach a cell `(i, j)`, you can only come from:

* `(i-1, j)` â†’ from **top**
* `(i, j-1)` â†’ from **left**

So the optimal path to `(i, j)` depends on the **minimum** of those two.

---

## ðŸ§  DP State Definition

Let:

```
dp[i][j] = minimum path sum to reach cell (i, j)
```

---

## ðŸ” DP Transition

```
dp[i][j] = grid[i][j] + min(dp[i-1][j], dp[i][j-1])
```

---

## ðŸ§± Base Cases

* `dp[0][0] = grid[0][0]`
* First row: only reachable from the left
* First column: only reachable from the top

---

## âœï¸ Short Approach Sketch

1. Initialize a DP table (or reuse the grid)
2. Fill the first row and first column
3. For each cell, add its value to the min of top & left
4. Return `dp[m-1][n-1]`

---

## ðŸ§® Python Solution (Space Optimized â€” In-Place)

```python
class Solution(object):
    def minPathSum(self, grid):
        m, n = len(grid), len(grid[0])
        
        for i in range(m):
            for j in range(n):
                if i == 0 and j == 0:
                    continue
                elif i == 0:
                    grid[i][j] += grid[i][j - 1]
                elif j == 0:
                    grid[i][j] += grid[i - 1][j]
                else:
                    grid[i][j] += min(grid[i - 1][j], grid[i][j - 1])
        
        return grid[m - 1][n - 1]
```

## â±ï¸ Complexity Analysis

* **Time Complexity:** `O(m Ã— n)`
* **Space Complexity:** `O(1)` (modifies input grid)

# My Solution

````python
# Not 100% optimized

class Solution(object):
    def minPathSum(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        m, n = len(grid), len(grid[0])
        dp = [[float('inf') for _ in range(n)] for _ in range(m)]
        k = min(m, n)
        for i in range(k):
            # down the column
            for j in range(i, m):
                min_path_sum = float('inf')
                if j > 0: min_path_sum = min(min_path_sum, dp[j-1][i]+grid[j][i])
                if i > 0: min_path_sum = min(min_path_sum, dp[j][i-1]+grid[j][i])
                if j==0 and i==0: min_path_sum = grid[j][i]
                dp[j][i] = min_path_sum
            # across the row
            for j in range(i, n):
                min_path_sum = float('inf')
                if j > 0: min_path_sum = min(min_path_sum, dp[i][j-1]+grid[i][j])
                if i > 0: min_path_sum = min(min_path_sum, dp[i-1][j]+grid[i][j])
                if i==0 and j==0: min_path_sum = grid[i][j]
                dp[i][j] = min_path_sum
        return dp[-1][-1]
````
