## âš™ï¸ Approach (Greedy Range Expansion)
### Idea:
At any index, you can jump within a range determined by your previous steps.  
We track:
- `current_end`: the farthest index reachable within the current jump.  
- `farthest`: the farthest index reachable with the **next** jump.  

When we reach `current_end`, we:
- increment the jump count,  
- and update `current_end = farthest`.

This ensures we make the **fewest possible jumps** to reach the last index.

---

### ğŸ” Steps:
1. Initialize:
   - `jumps = 0`
   - `current_end = 0`
   - `farthest = 0`
2. Traverse the array (up to `n - 2`):
   - Update `farthest = max(farthest, i + nums[i])`.
   - If `i == current_end`:  
     â†’ increment `jumps`,  
     â†’ set `current_end = farthest`.
3. Return `jumps`.

## â±ï¸ Complexity
- **Time:** O(n) â€” single pass through array.  
- **Space:** O(1) â€” constant extra space.

## âœ… Code Snippet (Python)
```python
def jump(nums):
    jumps = 0
    current_end = 0
    farthest = 0

    for i in range(len(nums) - 1):
        farthest = max(farthest, i + nums[i])
        if i == current_end:
            jumps += 1
            current_end = farthest

    return jumps
