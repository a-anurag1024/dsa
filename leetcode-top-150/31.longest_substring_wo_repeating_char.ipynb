## ðŸ§© Problem Summary

Given a string `s`, find the **length of the longest substring** without repeating characters.

A substring is a contiguous sequence of characters within a string.

---

## ðŸ§  Examples

### Example 1:
**Input:**  
`s = "abcabcbb"`  
**Output:** `3`  
**Explanation:**  
The longest substring without repeating characters is `"abc"` (length = 3).

---

### Example 2:
**Input:**  
`s = "bbbbb"`  
**Output:** `1`  
**Explanation:**  
Every substring has only `"b"`, so length = 1.

---

### Example 3:
**Input:**  
`s = "pwwkew"`  
**Output:** `3`  
**Explanation:**  
The longest substring is `"wke"`.  
Note: `"pwke"` is a subsequence, not a substring.

---

## âš™ï¸ Approach (Sliding Window + Hash Map)

### ðŸ”¹ Key Idea:
Use a **sliding window** to maintain a current substring without duplicates.  
A **hash map** (dictionary) tracks the last seen index of each character.

---

### ðŸ”¹ Step-by-Step Logic

1. Initialize:
   - `left = 0` â†’ window start index
   - `max_len = 0`
   - `seen = {}` â†’ stores last seen position of characters
2. Iterate through the string with `right` pointer:
   - If `s[right]` is already in `seen` and `seen[s[right]] >= left`,  
     move `left` to `seen[s[right]] + 1` (skip duplicate).
   - Update `seen[s[right]] = right`
   - Update `max_len = max(max_len, right - left + 1)`
3. Return `max_len` at the end.

## ðŸ§‘â€ðŸ’» Python Sketch

```python
def lengthOfLongestSubstring(s: str) -> int:
    seen = {}
    left = 0
    max_len = 0

    for right, ch in enumerate(s):
        if ch in seen and seen[ch] >= left:
            left = seen[ch] + 1  # skip duplicate
        seen[ch] = right
        max_len = max(max_len, right - left + 1)

    return max_len


# My solution

````python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        curr_chars = {}
        left = 0
        max_len = 0
        for right in range(len(s)):
            if curr_chars.get(s[right], False):
                cur_del = s[left]
                while cur_del != s[right]:
                    curr_chars[s[left]] = False
                    left += 1
                    cur_del = s[left]
                curr_chars[s[left]] = False
                left += 1
            curr_chars[s[right]] = True
            cur_len = right - left + 1
            if cur_len > max_len: max_len = cur_len
        return max_len
                    
````
