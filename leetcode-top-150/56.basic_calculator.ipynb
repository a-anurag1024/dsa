## ðŸ“˜ Problem Summary
You need to evaluate a valid mathematical expression `s` containing:
- Integers  
- `+` and `-`  
- Parentheses `(` and `)`  
- Spaces  
- Unary minus is allowed (`-1`, `-(2+3)`)

No `eval()` or built-in parser is allowed.

Your job: return the **integer result**.

---

## ðŸ§  Key Idea
Use a **stack-based expression evaluator**.

Since only `+`, `-`, and parentheses exist, we can evaluate the expression by:
- Maintaining a **current result**
- Maintaining the **current sign (+1 or -1)**
- Using a **stack** to store previous results and signs whenever `(` appears.

This avoids full token parsing and keeps everything O(n).

---

## ðŸ”‘ Approach (O(n) Time, O(n) Space)
### Key variables:
- `result` â†’ accumulated sum for the current parenthesis level
- `sign` â†’ current sign (either +1 or -1)
- `num` â†’ current number being parsed
- `stack` â†’ stores previous (result, sign) when entering `(`

### Steps:
1. Iterate through the characters in the string:
   - **Digit**: build the number (`num = num * 10 + digit`)
   - **'+'**: add `num * sign` to `result`, reset `num`, set `sign = +1`
   - **'-'**: add `num * sign` to `result`, reset `num`, set `sign = -1`
   - **'('**:
     - Push current `result` and `sign` to stack
     - Reset `result = 0`, `sign = +1`
   - **')'**:
     - Finalize current number: `result += num * sign`
     - Pop previous sign and previous result:
       - `result = prev_result + prev_sign * result`
     - Reset `num = 0`
2. At end of string, add last `num * sign`.

---

## ðŸ“ Example
Expression:  
`(1 + (4 + 5 + 2) - 3) + (6 + 8)`

Evaluation flow (abstract):
- Push result, sign on '('
- Evaluate inner expression
- Pop and combine
- Repeat


Final result: **23**

```python
class Solution(object):
    def calculate(self, s):
        stack = []
        result = 0
        sign = 1
        num = 0

        for c in s:
            if c.isdigit():
                num = num * 10 + int(c)
            elif c in "+-":
                result += num * sign
                num = 0
                sign = 1 if c == '+' else -1
            elif c == '(':
                stack.append(result)
                stack.append(sign)
                result = 0
                sign = 1
            elif c == ')':
                result += num * sign
                num = 0
                prev_sign = stack.pop()
                prev_result = stack.pop()
                result = prev_result + prev_sign * result

        return result + num * sign

````python
class Solution(object):
    def calculate(self, s):
        """
        :type s: str
        :rtype: int
        """
        st = []
        def is_number(l):
            try:
                float(l)
                return True
            except:
                return False

        def backtrack():
            l = ""
            val = 0
            curr_val = None
            while l != '(' and len(st)>0:
                l = st.pop()
                if l == '+':
                    val += int(curr_val)
                    curr_val = None
                elif l == '-':
                    val -= int(curr_val)
                    curr_val = None
                elif is_number(l):
                    curr_val = l
            if curr_val: val += int(curr_val)
            return str(val)

        curr_num = ""
        for c in s:
            if c != ')':
                if c.isdigit():
                    curr_num += c
                elif c in ['+', '-', '(']:
                    if curr_num != "":
                        st.append(curr_num)
                        curr_num = ""
                    st.append(c)
                elif c == ' ' and curr_num != "":
                    st.append(curr_num)
                    curr_num = ""
            else:
                if curr_num: st.append(curr_num)
                curr_num = ""
                val = backtrack()
                st.append(val)
        if curr_num: st.append(curr_num)
        return int(backtrack())
````
