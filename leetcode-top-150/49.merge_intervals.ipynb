## ðŸ“˜ Problem Summary
You're given a list of intervals `[[start, end]]`.  
Your task is to **merge all overlapping intervals** and return the final list of **non-overlapping** merged intervals.

Two intervals overlap if:
- `current_start <= previous_end`

---

## ðŸ§  Key Idea
If intervals are sorted by start time, merging becomes a simple linear scan.  
Always compare the current interval with the **last merged** one.

---

## âœ… Approach (O(n log n) for sorting + O(n) scan)
1. **Sort** intervals by their start value.
2. Initialize a result list with the first interval.
3. For each next interval:
   - If it overlaps with the last interval in result â†’ **merge** by updating the end:
     - `merged[-1][1] = max(merged[-1][1], current_end)`
   - Otherwise â†’ **append** as a new interval.
4. Return the merged list.

---

## ðŸ“ Example
Input: `[[1,3],[2,6],[8,10],[15,18]]`  
Output: `[[1,6],[8,10],[15,18]]`

---

## ðŸ§© Sample Python Implementation
```python
class Solution(object):
    def merge(self, intervals):
        intervals.sort(key=lambda x: x[0])
        merged = [intervals[0]]

        for start, end in intervals[1:]:
            last_start, last_end = merged[-1]

            if start <= last_end:  # Overlap â†’ merge
                merged[-1][1] = max(last_end, end)
            else:
                merged.append([start, end])

        return merged

````python
class Solution(object):
    def merge(self, intervals):
        """
        :type intervals: List[List[int]]
        :rtype: List[List[int]]
        """
        if len(intervals) == 0: return []
        intervals.sort(key=lambda x: x[0])
        curr = intervals[0]
        ret_intervals = []
        for i in range(1, len(intervals)):
            if curr[1] >= intervals[i][0]:
                curr[1] = max(curr[1], intervals[i][1])
            else:
                ret_intervals.append(curr)
                curr = intervals[i]
        ret_intervals.append(curr)
        return ret_intervals
````
