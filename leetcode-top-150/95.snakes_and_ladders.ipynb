# üêçüé≤ LeetCode 909: Snakes and Ladders ‚Äî Quick Review

## üìò Problem Summary
You are given an `n √ó n` board representing a **Snakes and Ladders** game.

- Cells are numbered from `1` to `n¬≤` in **Boustrophedon order**
  (left ‚Üí right, then right ‚Üí left, starting from bottom row).
- You start at square `1`.
- Each move:
  - Roll a die ‚Üí move `+1` to `+6` squares.
  - If the destination has a **snake or ladder**, jump **once** to its target.
- Goal:
  ‚û°Ô∏è Reach square `n¬≤` in the **minimum number of dice rolls**.

If unreachable ‚Üí return `-1`.

---

## üß† Key Insight
This is a **shortest-path problem on an unweighted graph**.

- Each square is a node.
- Each dice roll creates up to **6 edges**.
- Snakes/ladders are **forced edges**.

üëâ The correct strategy is **Breadth-First Search (BFS)**.

---

## üöÄ Approach (BFS)
### Steps:
1. Convert a square number into `(row, col)` using Boustrophedon logic.
2. Use BFS starting from square `1`.
3. For each square:
   - Try all dice outcomes `[+1 ‚Ä¶ +6]`.
   - If a snake/ladder exists ‚Üí jump once.
4. Use a `visited` set to avoid cycles.
5. Stop when reaching square `n¬≤`.

---

## üß© Boustrophedon Mapping
Given square `s` (1-indexed):

```text
row = n - 1 - (s - 1) // n
col = (s - 1) % n        if (n - 1 - row) is even
      n - 1 - (s - 1) % n otherwise
```

This converts square labels into board indices correctly.

üß© Time & Space Complexity

Time: O(n¬≤) ‚Äî each square visited once.

Space: O(n¬≤) ‚Äî queue + visited set.

### Sample BFS code
```python
from collections import deque

class Solution:
    def snakesAndLadders(self, board):
        n = len(board)
        
        def get_pos(s):
            r = n - 1 - (s - 1) // n
            c = (s - 1) % n if ((n - 1 - r) % 2 == 0) else n - 1 - (s - 1) % n
            return r, c
        
        visited = set([1])
        q = deque([(1, 0)])  # (square, moves)
        
        while q:
            curr, moves = q.popleft()
            if curr == n * n:
                return moves
            
            for nxt in range(curr + 1, min(curr + 6, n*n) + 1):
                r, c = get_pos(nxt)
                dest = board[r][c] if board[r][c] != -1 else nxt
                if dest not in visited:
                    visited.add(dest)
                    q.append((dest, moves + 1))
        
        return -1

````python
from collections import deque

class Solution(object):
    def snakesAndLadders(self, board):
        """
        :type board: List[List[int]]
        :rtype: int
        """
        n = len(board)
        visited = set()
        def square_to_i_j(sq_no):
            row = (sq_no-1) // n
            col = ((sq_no-1) % n)
            i = n - 1 - row
            j = col if row % 2 == 0 else n - 1 - col
            return i, j
        q = deque([(1, 0)])
        while q:
            pos, moves = q.popleft()
            if pos == n ** 2: return moves
            for nxt in range(1+pos, 1+min(6+pos, n**2)):
                i, j = square_to_i_j(nxt)
                next_pos = nxt if board[i][j] == -1 else board[i][j]
                if next_pos not in visited:
                    visited.add(next_pos)
                    q.append((next_pos, moves+1))
        return -1
````
