## ðŸ”¹ Quick Review: LeetCode 108 â€” Convert Sorted Array to Binary Search Tree

### **Problem Summary**

Given a sorted (ascending) integer array `nums`, convert it into a **height-balanced Binary Search Tree (BST)**.

* BST property:
  left subtree < root < right subtree
* Height-balanced means:

  * Depth difference of left & right subtrees â‰¤ 1

---

### **Key Observations**

* The **middle element** of a sorted array makes the best root
* Left half â†’ left subtree
* Right half â†’ right subtree
* Recursively applying this guarantees balance
* This is a **divide-and-conquer** problem

---

## ðŸ”¹ Short Approach Sketch (Recursive DFS)

1. Define a recursive function `build(l, r)`:

   * If `l > r`, return `None`
2. Choose middle index:

   ```python
   mid = (l + r) // 2
   ```
3. Create root with `nums[mid]`
4. Recursively build:

   * Left subtree from `[l, mid-1]`
   * Right subtree from `[mid+1, r]`
5. Return the root

---

### **Time & Space Complexity**

* **Time:** `O(n)` â€” every element is used once
* **Space:** `O(log n)` recursion depth (balanced tree)

---

## ðŸ”¹ Python Solution (Optimal)

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

class Solution(object):
    def sortedArrayToBST(self, nums):
        """
        :type nums: List[int]
        :rtype: Optional[TreeNode]
        """

        def build(l, r):
            if l > r:
                return None

            mid = (l + r) // 2
            root = TreeNode(nums[mid])

            root.left = build(l, mid - 1)
            root.right = build(mid + 1, r)

            return root

        return build(0, len(nums) - 1)
```

---

## ðŸ”¹ Why This Guarantees Balance

* Choosing the **middle element** ensures:

  * Left and right subtrees have nearly equal size
  * Height difference â‰¤ 1 at every node
