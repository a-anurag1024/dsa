### ü§ñ LeetCode 63 ‚Äî Unique Paths II (Quick Review)

---

#### üîç Problem Recap

You are given an `m √ó n` grid:

* `0` ‚Üí free cell
* `1` ‚Üí obstacle

A robot:

* Starts at **top-left** `(0,0)`
* Wants to reach **bottom-right** `(m-1,n-1)`
* Can move **only right or down**
* **Cannot step on obstacles**

üëâ **Goal:** Return the **number of unique paths** to reach the destination.

---

### üí° Core Insight

This is a **counting DP on a grid with blocked cells**.

For each cell:

* If it‚Äôs an obstacle ‚Üí **0 paths**
* Otherwise ‚Üí paths come from **top + left**

---

## üß† DP State Definition

Let:

```
dp[i][j] = number of unique paths to reach cell (i, j)
```

---

## üîÅ DP Transition

```
if grid[i][j] == 1:
    dp[i][j] = 0
else:
    dp[i][j] = dp[i-1][j] + dp[i][j-1]
```

---

## üß± Base Cases

* If `grid[0][0] == 1` ‚Üí return `0`
* `dp[0][0] = 1`
* First row / column must stop accumulating once an obstacle appears

---

## ‚úçÔ∏è Short Approach Sketch

1. Handle blocked start cell
2. Initialize DP table (or reuse grid)
3. For each cell:

   * If obstacle ‚Üí set paths to 0
   * Else sum paths from top & left
4. Return value at bottom-right

---

## üßÆ Python Solution (Space Optimized ‚Äî In-Place)

```python
class Solution(object):
    def uniquePathsWithObstacles(self, obstacleGrid):
        m, n = len(obstacleGrid), len(obstacleGrid[0])
        
        if obstacleGrid[0][0] == 1:
            return 0
        
        obstacleGrid[0][0] = 1
        
        for i in range(m):
            for j in range(n):
                if obstacleGrid[i][j] == 1 and not (i == 0 and j == 0):
                    obstacleGrid[i][j] = 0
                else:
                    if i > 0:
                        obstacleGrid[i][j] += obstacleGrid[i - 1][j]
                    if j > 0:
                        obstacleGrid[i][j] += obstacleGrid[i][j - 1]
        
        return obstacleGrid[m - 1][n - 1]
```

---

## ‚è±Ô∏è Complexity Analysis

* **Time Complexity:** `O(m √ó n)`
* **Space Complexity:** `O(1)` (grid reused)


# My Solution

````python
class Solution(object):
    def uniquePathsWithObstacles(self, obstacleGrid):
        """
        :type obstacleGrid: List[List[int]]
        :rtype: int
        """
        m, n = len(obstacleGrid), len(obstacleGrid[0])
        left_col = []
        num_paths = 1
        for i in range(m):
            if obstacleGrid[i][0] == 1: num_paths = 0
            left_col.append(num_paths)
        top_row = []
        num_paths = 1
        for i in range(n):
            if obstacleGrid[0][i] == 1: num_paths = 0
            top_row.append(num_paths)
        
        for i in range(1, m):
            current_row = [left_col[i]]
            for j in range(1, n):
                if obstacleGrid[i][j] == 1: current_row.append(0)
                else: current_row.append(current_row[-1]+top_row[j])
            top_row = current_row
        return top_row[-1]
````
