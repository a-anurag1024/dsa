This is a **classic DP on grid / triangle** problem.

Each cell depends **only on the two cells directly below it**.

Instead of building from the top, the cleanest solution is **bottom-up DP**.

## ğŸ§  DP Idea (Bottom-Up)

### State

Let:

* `dp[j]` = minimum path sum from the current rowâ€™s index `j` to the bottom

### Transition

From bottom to top:

```

```

### ğŸ§± Base Case

Initialize `dp` as the **last row** of the triangle.

---

### âœï¸ Short Approach Sketch

1. Copy the last row into a DP array
2. Move from the second-last row up to the top
3. For each element:

   * Add the minimum of its two children
4. The top element of DP is the answer

---

### ğŸ§® Python Solution (Space Optimized O(n))

```python
class Solution(object):
    def minimumTotal(self, triangle):
        dp = triangle[-1][:]   # copy last row
        
        for i in range(len(triangle) - 2, -1, -1):
            for j in range(len(triangle[i])):
                dp[j] = triangle[i][j] + min(dp[j], dp[j + 1])
        
        return dp[0]
```

# My Solution

````python
# Top Down Approach

class Solution(object):
    def minimumTotal(self, triangle):
        """
        :type triangle: List[List[int]]
        :rtype: int
        """
        last = triangle[0]
        for curr_row in triangle[1:]:
            curr = [float('inf')]*(len(curr_row))
            for i in range(len(curr_row)):
                min_sum = float('inf')
                if i > 0: min_sum = min(min_sum, curr_row[i]+last[i-1])
                if i < len(last): min_sum = min(min_sum, curr_row[i]+last[i])
                curr[i] = min_sum
            last = curr
        return min(last)
````
