# üîí LeetCode 20: Valid Parentheses ‚Äî Quick Review

## üìò Problem Summary
You are given a string `s` containing only brackets:  
`(` `)` `{` `}` `[` `]`  

Your task is to determine if the string is **valid** based on these rules:
1. Every opening bracket must be closed by the **same type**.
2. Brackets must close in the **correct order**.
3. Every closing bracket must match **an existing, unmatched** opening bracket.

---

## üß† Key Idea
Use a **stack** to track opening brackets.

- When you see an **opening bracket**, push it onto the stack.
- When you see a **closing bracket**, check:
  - Stack must not be empty.
  - It must match the top element (via a mapping).
- In the end, the stack must be empty for the string to be valid.

This gives a clean and efficient solution.

---

## üîë Approach (O(n) Time, O(n) Space)
1. Create a hashmap mapping closing ‚Üí opening:
')': '('
']': '['
'}': '{'
2. Initialize an empty stack.
3. Traverse the string:
- If it's an opening bracket ‚Üí push to stack.
- If it's closing:
  - Stack must be non-empty.
  - Top must match the corresponding opening bracket.
  - Pop if matched; return false otherwise.
4. After the loop, return `True` only if stack is empty.

---

## üìù Examples
| Input | Output | Explanation |
|-------|--------|-------------|
| `"()"` | `true` | Simple matching pair |
| `"()[]{}"` | `true` | Multiple correct pairs |
| `"(]"` | `false` | Wrong type |
| `"([])"` | `true` | Nested correctly |
| `"([)]"` | `false` | Wrong order |

---

## üß© Sample Python Implementation
```python
class Solution(object):
 def isValid(self, s):
     stack = []
     match = {')':'(', ']':'[', '}':'{'}

     for ch in s:
         if ch in match.values():          # opening bracket
             stack.append(ch)
         else:                              # closing bracket
             if not stack or stack[-1] != match[ch]:
                 return False
             stack.pop()

     return len(stack) == 0

````python
from collections import deque

class Solution(object):
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        stack = deque()
        combs = {
            '(': ')',
            '{': '}',
            '[': ']'
        }
        for c in s:
            if c in combs:
                stack.append(c)
            else:
                if len(stack) == 0: return False
                curr = stack.pop()
                if c != combs[curr]: return False
        if len(stack) == 0: return True
        return False
        
````
