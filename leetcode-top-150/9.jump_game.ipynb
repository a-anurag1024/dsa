## ðŸ“ Problem Summary
Given an array `nums`, where each element represents the **maximum jump length** from that position,  
determine if it is possible to **reach the last index** starting from index `0`.

## ðŸ’¡ Example
**Input:**

**Output:** `True`  
**Explanation:**  
Jump 1 step from index `0` â†’ `1`, then jump 3 steps from index `1` â†’ `4`.

**Input:**  
`nums = [3,2,1,0,4]`

**Explanation:**

## âš™ï¸ Approach (Greedy Reachability)
### Idea:
Keep track of the **farthest index** you can reach at each step.  
If at any point the current index exceeds your farthest reachable point,  
you canâ€™t proceed further â€” return `False`.

### Steps:
1. Initialize `max_reach = 0`.
2. Traverse the array:
   - If the current index `i` is **greater than** `max_reach`, return `False`.
   - Update `max_reach = max(max_reach, i + nums[i])`.
3. If the loop finishes, return `True` â€” last index is reachable.

def canJump(nums):
    max_reach = 0
    for i in range(len(nums)):
        if i > max_reach:
            return False
        max_reach = max(max_reach, i + nums[i])
    return True


# My solution

````python
class Solution(object):
    def canJump(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        n = len(nums)
        target = n - 1
        for start in range(target-1, -1, -1):
            if target <= start + nums[start]:
                target = start
        if target == 0: return True
        else: return False
````
