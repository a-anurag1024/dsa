# üß© LeetCode 48: Rotate Image ‚Äî Quick Review

## üìå Problem Summary
You're given an `n x n` matrix representing an image.  
Your task: **Rotate the matrix 90¬∞ clockwise, in-place**  
‚û°Ô∏è No extra 2D matrix can be used.

---

## ‚úÖ Key Insight
A 90¬∞ clockwise rotation can be achieved with **two simple in-place steps**:

1. **Transpose the matrix**  
   Swap elements across the main diagonal  
   `matrix[i][j] ‚Üî matrix[j][i]`

2. **Reverse each row**  
   This shifts columns into correct rotated positions.

These two operations produce the same effect as rotating the matrix by 90¬∞ clockwise.

---

## üß† Approach (Short & Practical)

### **Method: Transpose + Reverse Rows**
- Loop through upper triangle to transpose the matrix.
- Reverse every row of the transposed matrix.
- Both steps are in-place ‚Üí meets constraints.

**Time Complexity:** `O(n¬≤)`  
**Space Complexity:** `O(1)` (in-place)

## üîß Python Code (In-Place)
```python
class Solution:
    def rotate(self, matrix):
        n = len(matrix)

        # Step 1: Transpose
        for i in range(n):
            for j in range(i + 1, n):
                matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]

        # Step 2: Reverse each row
        for i in range(n):
            matrix[i].reverse()


# My Solution (Layer by Layer)

````python
class Solution(object):
    def rotate(self, matrix):
        n = len(matrix)
        
        # process layer by layer
        for layer in range(n // 2):
            first = layer
            last = n - layer - 1
            
            for i in range(first, last):
                offset = i - first

                # save top
                top = matrix[first][i]

                # left ‚Üí top
                matrix[first][i] = matrix[last - offset][first]

                # bottom ‚Üí left
                matrix[last - offset][first] = matrix[last][last - offset]

                # right ‚Üí bottom
                matrix[last][last - offset] = matrix[i][last]

                # top ‚Üí right
                matrix[i][last] = top
````
