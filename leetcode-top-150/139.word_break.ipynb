### üî§ LeetCode 139 ‚Äî Word Break (Quick Review)

---

#### üîç Problem Recap

You are given:

* A string `s`
* A dictionary `wordDict`

üëâ **Goal:** Determine if `s` can be segmented into **one or more dictionary words**,
where **dictionary words can be reused**.

---

### üí° Core Insight

This is a **prefix-based Dynamic Programming** problem.

At every index `i`, ask:

> *Is there a word in the dictionary that ends at `i` AND the string before it was also segmentable?*

This naturally leads to a **boolean DP** formulation.

---

### üß† DP State Definition

Let:

* `dp[i] = True` if `s[:i]` can be segmented using `wordDict`

---

### üîÅ DP Transition

For each `i` from `1 ‚Üí len(s)`:

```
dp[i] = True if ‚àÉ j < i such that:
         dp[j] == True and s[j:i] ‚àà wordDict
```

---

### üß± Base Case

```
dp[0] = True   # empty string is always segmentable
```

---

### ‚úçÔ∏è Short Approach Sketch

1. Convert `wordDict` to a **set** for O(1) lookups
2. Create a boolean DP array of size `n + 1`
3. For each position `i`, check all previous splits `j`
4. If any valid split exists ‚Üí mark `dp[i] = True`
5. Return `dp[n]`

---

### üßÆ Python Solution (Bottom-Up DP)

```python
class Solution(object):
    def wordBreak(self, s, wordDict):
        word_set = set(wordDict)
        n = len(s)
        
        dp = [False] * (n + 1)
        dp[0] = True
        
        for i in range(1, n + 1):
            for j in range(i):
                if dp[j] and s[j:i] in word_set:
                    dp[i] = True
                    break
        
        return dp[n]
```

---

### ‚è±Ô∏è Complexity Analysis

* **Time Complexity:** `O(n¬≤)`
* **Space Complexity:** `O(n)`

> In practice, performance is good due to early breaks and hash-set lookups.

---

# My Solution

````python
class Solution(object):
    def wordBreak(self, s, wordDict):
        """
        :type s: str
        :type wordDict: List[str]
        :rtype: bool
        """
        dp = [0]*len(s)
        word_set = set(wordDict)
        for i in range(len(s)):
            if s[:i+1] in word_set:
                dp[i] = 1
            else:
                for j in range(1, i+1):
                    if dp[j-1] == 1 and s[j:i+1] in word_set:
                        dp[i] = 1
                        break
        return True if dp[-1] else False
````
