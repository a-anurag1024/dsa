## âš™ï¸ Approach (Boyerâ€“Moore Voting Algorithm)
1. Initialize:
   - `count = 0`
   - `candidate = None`
2. Traverse through `nums`:
   - If `count == 0`, set `candidate = num`.
   - If `num == candidate`, increment `count`; else decrement it.
3. After traversal, `candidate` holds the majority element.

This works because the majority elementâ€™s total count **outweighs** all others combined. And because if the count becomes zero that means there are equal number of cancelling elements (must'nt be the case for the majority element)

def majorityElement(nums):
    count = 0
    candidate = None

    for num in nums:
        if count == 0:
            candidate = num
        count += (1 if num == candidate else -1)
    return candidate


# My Solution (Hashmap method. O(n) time/space)

````python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        hashmap = {nums[0]:1}
        max_element = nums[0]
        max_counter = 1
        for j in range(1, len(nums)):
            if nums[j] in hashmap:
                hashmap[nums[j]] += 1
                if hashmap[nums[j]] > max_counter:
                    max_counter = hashmap[nums[j]]
                    max_element = nums[j]
            else:
                hashmap[nums[j]] = 1
        return max_element
````
