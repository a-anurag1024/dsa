## üìå Problem Summary  
Given the root of a binary tree, determine whether it is a **valid Binary Search Tree (BST)**.

A valid BST must satisfy:
1. **Left subtree** values < node value  
2. **Right subtree** values > node value  
3. Both subtrees must *recursively* be valid BSTs  

Note: All comparisons are **strict**, not ‚â§ or ‚â•.

---

## üîç Example  

### Example  
**Input:** `[2,1,3]`  
**Output:** `true`  
Because:

follows the BST rules perfectly.

## üöÄ Key Idea  
Two standard solutions:

---

### ‚≠ê Approach 1 ‚Äî Min/Max Range Propagation (Recommended)  
During DFS, each node must lie within a valid numeric range:

left subtree: (min, node.val)

right subtree: (node.val, max)


At each node:
- If value violates range ‚Üí not a BST  
- Recursively validate children with updated bounds  

This captures sibling constraints the local-node check alone can‚Äôt detect.

---

### ‚≠ê Approach 2 ‚Äî Inorder Traversal (Sorted Order Check)  
BST inorder traversal gives a **strictly increasing sequence**.  
Track `prev` value and ensure each new value > prev.

---

## üìù Short Approach (Interview-Ready)

**Recursive min/max approach:**
1. Start with bounds `(-inf, +inf)`  
2. For each node:
   - If `node.val <= min` or `node.val >= max` ‚Üí invalid  
   - Recurse into left with `(min, node.val)`  
   - Recurse into right with `(node.val, max)`  
3. All nodes must satisfy the constraint  

Time Complexity: **O(n)**  
Space Complexity: **O(h)** (recursion stack)

---

## üßë‚Äçüíª Template Code ‚Äî Min/Max DFS

```python
class Solution:
    def isValidBST(self, root):

        def validate(node, low, high):
            if not node:
                return True
            if not (low < node.val < high):
                return False
            return (validate(node.left, low, node.val) and
                    validate(node.right, node.val, high))

        return validate(root, float('-inf'), float('inf'))
```

## Template code - InOrder Traversal Check
```python
class Solution:
    def isValidBST(self, root):
        self.prev = None

        def inorder(node):
            if not node:
                return True
            if not inorder(node.left):
                return False
            if self.prev is not None and node.val <= self.prev:
                return False
            self.prev = node.val
            return inorder(node.right)

        return inorder(root)


# My Solution

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def isValidBST(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: bool
        """
        self.is_bst = True
        def dfs(curr):
            if not curr:
                return float('inf'), -float('inf')
            if curr.left:
                lmin, lmax = dfs(curr.left)
            else:
                lmin, lmax = curr.val, -float('inf')
            if curr.val <= lmax: self.is_bst = False
            if curr.right:
                rmin, rmax = dfs(curr.right)
            else:
                rmin, rmax = float('inf'), curr.val
            if curr.val >= rmin: self.is_bst = False
            return lmin, rmax
        _, _ = dfs(root)
        return self.is_bst
````
