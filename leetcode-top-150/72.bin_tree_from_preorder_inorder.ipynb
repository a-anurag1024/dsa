## ðŸš€ Key Idea  
### **Preorder gives the root first.**  
`preorder[0]` is always the **root** of the current subtree.

### **Inorder tells how the tree splits.**  
If root = `x`, and `idx` is its position in `inorder`, then:

inorder: [ left-subtree | x | right-subtree ]

sizes: L elements 1 R elements


### Use recursive partitioning:  
1. Pick root from preorder  
2. Find position in inorder  
3. Recursively build left subtree (using first L nodes from preorder after root)  
4. Recursively build right subtree  

To make this efficient â†’ **use a hashmap** to lookup `inorder` indices in O(1).

---

## ðŸ“ Short Approach (Interview-Ready)

1. Build a map `{value â†’ inorder_index}`.  
2. Maintain a `pre_idx` pointer for preorder.  
3. Recursive function `build(l, r)` meaning:
   - build subtree from `inorder[l:r]` (inclusive range)  
4. Steps inside recursion:
   - If `l > r`: return None  
   - Root = preorder[pre_idx], increment `pre_idx`  
   - Split inorder using index map  
   - Build left subtree  
   - Build right subtree  
5. Return root.

Time Complexity: **O(n)**  
Space Complexity: **O(n)**  
(recursion + index map)

## ðŸ§‘â€ðŸ’» Template Code (Efficient DFS Construction)

```python
class Solution:
    def buildTree(self, preorder, inorder):
        # map inorder value â†’ index
        idx_map = {val: i for i in inorder}
        pre_idx = 0

        def build(left, right):
            nonlocal pre_idx
            if left > right:
                return None

            # Step 1: root from preorder
            root_val = preorder[pre_idx]
            pre_idx += 1
            root = TreeNode(root_val)

            # Step 2: find root in inorder
            mid = idx_map[root_val]

            # Step 3: build left and right
            root.left = build(left, mid - 1)
            root.right = build(mid + 1, right)

            return root

        return build(0, len(inorder) - 1)

````python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution(object):
    def buildTree(self, preorder, inorder):
        """
        :type preorder: List[int]
        :type inorder: List[int]
        :rtype: Optional[TreeNode]
        """
        in_order_map = {val:i for i, val in enumerate(inorder)}
        def helper(pre_ind, in_l, in_r):
            if in_l > in_r: return None, pre_ind
            if pre_ind >= len(inorder): return None, pre_ind
            curr_val = preorder[pre_ind]
            mid = in_order_map[curr_val]
            curr_node = TreeNode(val=curr_val)
            pre_ind += 1
            left_node, pre_ind = helper(pre_ind, in_l, mid-1)
            right_node, pre_ind = helper(pre_ind, mid+1, in_r)
            curr_node.left = left_node
            curr_node.right = right_node
            return curr_node, pre_ind
        return helper(0, 0, len(preorder))[0]
````
