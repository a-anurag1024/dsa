Given an integer `num` (1 â‰¤ num â‰¤ 3999), convert it into a **Roman numeral** representation.

| Symbol | Value |

## âš™ï¸ Roman Numeral Construction Rules

1. **Symbols are arranged from largest â†’ smallest**, left to right.
2. **Subtractive Notation** is used for numbers like 4, 9, 40, 90, 400, 900:
   - IV (4), IX (9)
   - XL (40), XC (90)
   - CD (400), CM (900)
3. Only `I`, `X`, `C`, and `M` can be repeated (up to 3 times).

## âš™ï¸ Approach (Greedy Conversion)

### ğŸ”¹ Key Idea:
Use **greedy subtraction** â€” always take the **largest possible Roman value** that fits into the current number, append its symbol, and subtract it.

---

### ğŸ”¹ Step-by-Step

1. Maintain a list of **(value, symbol)** pairs in descending order.
2. For each pair:
   - While `num >= value`, append `symbol` to result and subtract `value` from `num`.
3. Continue until `num` becomes 0.

```python
def intToRoman(num: int) -> str:
    val = [1000, 900, 500, 400,
           100, 90, 50, 40,
           10, 9, 5, 4, 1]
    syb = ["M", "CM", "D", "CD",
           "C", "XC", "L", "XL",
           "X", "IX", "V", "IV", "I"]
    
    roman = ""
    for i in range(len(val)):
        while num >= val[i]:
            roman += syb[i]
            num -= val[i]
    return roman


# My Solution

````python
class Solution(object):
    def intToRoman(self, num):
        """
        :type num: int
        :rtype: str
        """
        tens = {1:"I", 10:"X", 100:"C", 1000:"M"}
        fives = {5:"V", 50:"L", 500:"D"}
        cur_ten_deg = 3
        val = ""
        while cur_ten_deg >= 0:
            ten_base = 10**(cur_ten_deg)
            five_base = 5 * ten_base
            lead = int(num/ten_base)
            rem = num % ten_base
            if lead != 0:
                if lead <= 3:
                    val = val + tens[ten_base] * lead
                elif lead == 4:
                    val = val + tens[ten_base] + fives[five_base]
                elif lead <= 8:
                    val = val + fives[five_base] + tens[ten_base] * (lead-5)
                else:
                    val = val + tens[ten_base] + tens[10*ten_base]
            num = rem
            cur_ten_deg -= 1
                
        return val
````
