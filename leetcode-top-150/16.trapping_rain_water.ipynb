## üß© Problem Summary

You are given `n` non-negative integers representing an elevation map where the width of each bar is `1`.  
You need to find **how much water can be trapped** after it rains.

Each bar can trap water based on the **minimum of the tallest bars to its left and right** minus its own height.

## ‚öôÔ∏è Approach (Two-Pointer Greedy ‚Äî Optimal)

### üß† Key Idea:
At any position, the trapped water depends on:
$$
\text{water}[i] = \min(\text{max\_left}[i], \text{max\_right}[i]) - \text{height}[i]
$$
But we can do this in **one pass** with **two pointers** and **O(1) space**.

---

### üîπ Step-by-Step

1. Initialize two pointers:  
   `left = 0`, `right = n - 1`,  
   and two variables `left_max = 0`, `right_max = 0`.

2. Move the pointer with the **smaller boundary**:
   - If `height[left] < height[right]`:
     - If `height[left] >= left_max`, update `left_max`.
     - Else, water can be trapped: `ans += left_max - height[left]`.
     - Move `left += 1`.
   - Else:
     - If `height[right] >= right_max`, update `right_max`.
     - Else, water can be trapped: `ans += right_max - height[right]`.
     - Move `right -= 1`.

3. Continue until both pointers meet.

---

### üßÆ Complexity

| Type | Complexity |
|------|-------------|
| ‚è±Ô∏è Time | `O(n)` |
| üß† Space | `O(1)` |

---

## üßë‚Äçüíª Python Sketch

```python
def trap(height):
    left, right = 0, len(height) - 1
    left_max = right_max = 0
    water = 0

    while left < right:
        if height[left] < height[right]:
            if height[left] >= left_max:
                left_max = height[left]
            else:
                water += left_max - height[left]
            left += 1
        else:
            if height[right] >= right_max:
                right_max = height[right]
            else:
                water += right_max - height[right]
            right -= 1

    return water
