## ‚ûï LeetCode 66: Plus One ‚Äî Quick Review

### üß© Problem Summary

You‚Äôre given a non-negative integer represented as an **array of digits** (most significant ‚Üí least significant).
Your task is to **add 1** to the number and return the resulting digits array.

---

## üîç Key Observations

1. **Addition happens from the rightmost digit**
2. **Carry propagation** is the main challenge
3. Only digit `9` causes a carry
4. Edge case: **all digits are 9** ‚Üí array grows in size

   * `[9] ‚Üí [1,0]`
   * `[9,9] ‚Üí [1,0,0]`

---

## üß† Approach Sketch (Optimal)

### Idea: Simulate grade-school addition

* Traverse digits **from right to left**
* If digit < 9 ‚Üí increment and return
* If digit == 9 ‚Üí set to 0 and continue
* If all digits become 0 ‚Üí prepend `1`

---

## ‚úÖ Python Solution (Optimal & Interview-Preferred)

```python
class Solution(object):
    def plusOne(self, digits):
        n = len(digits)
        
        for i in range(n - 1, -1, -1):
            if digits[i] < 9:
                digits[i] += 1
                return digits
            digits[i] = 0
        
        # If all digits were 9
        return [1] + digits
```

---

## ‚è±Ô∏è Complexity Analysis

* **Time:** `O(n)`
* **Space:** `O(1)` (output array excluded)




# My Solution

````python
class Solution(object):
    def plusOne(self, digits):
        """
        :type digits: List[int]
        :rtype: List[int]
        """
        carry = 1
        i = len(digits) - 1
        while carry and i >= 0:
            curr = digits[i]
            csum = carry + curr
            digits[i] = csum % 10
            carry = csum // 10
            i -= 1
        if carry != 0:
            digits = [carry] + digits
        return digits
````
