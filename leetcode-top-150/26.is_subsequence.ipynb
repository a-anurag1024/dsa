## ðŸ§© Problem Summary

Given two strings `s` and `t`, determine if `s` is a **subsequence** of `t`.

A **subsequence** of a string is formed by deleting some (or none) of its characters **without changing the order** of the remaining characters.

---

## ðŸ§  Examples

### Example 1:
**Input:**  
`s = "abc"`, `t = "ahbgdc"`  
**Output:** `true`  
**Explanation:**  
Characters `'a'`, `'b'`, `'c'` appear in order in `t`.

---

### Example 2:
**Input:**  
`s = "axc"`, `t = "ahbgdc"`  
**Output:** `false`  
**Explanation:**  
Character `'x'` does not appear after `'a'` in the same sequence.

---

## âš™ï¸ Approach (Two-Pointer Method)

### ðŸ”¹ Key Idea:
Use two pointers â€” one for `s` and one for `t` â€” to track matching characters in order.

1. Start both pointers at the beginning of `s` and `t`.
2. Move the `t` pointer through each character.
3. Each time a matching character is found, move the `s` pointer forward.
4. If the `s` pointer reaches the end, all characters in `s` were found in order â†’ return `True`.
5. Otherwise, return `False`.

## ðŸ§‘â€ðŸ’» Python Sketch

```python
def isSubsequence(s: str, t: str) -> bool:
    i, j = 0, 0
    while i < len(s) and j < len(t):
        if s[i] == t[j]:
            i += 1
        j += 1
    return i == len(s)


# My solution

````python
class Solution(object):
    def isSubsequence(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        sp = tp = 0
        while tp < len(t) and sp < len(s):
            if s[sp] == t[tp]:
                sp += 1
            tp += 1
        if sp == len(s): return True
        else: return False
````
