## üìå Problem Summary  
Given the root of a **Binary Search Tree (BST)**, return the **minimum absolute difference** between values of any two **different** nodes.

Because it's a BST, performing **inorder traversal** yields values in **sorted order**, and the minimum difference must occur between **adjacent elements** in this sorted list.

---

## üîç Examples  

### Example 1  
Input: `[4,2,6,1,3]`  
Inorder: `1, 2, 3, 4, 6`  
Differences: `1, 1, 1, 2`  
Output: `1`

### Example 2  
Input: `[1,0,48,null,null,12,49]`  
Inorder: `0, 1, 12, 48, 49`  
Min diff = `1`  
Output: `1`

---

## üöÄ Key Idea  
### BST property:
**Inorder traversal ‚Üí sorted values**

### So:
To find minimum absolute difference between any two nodes, it's enough to compare **adjacent** values in inorder traversal.

### Strategy:
1. Keep a `prev` pointer to store previous node's value while doing inorder traversal  
2. Compute `min_diff = min(min_diff, current - prev)`  
3. Continue until traversal is complete  

Time Complexity: **O(n)**  
Space Complexity:  
- **O(h)** recursion (best method)  
- Or O(n) if you store all values first  

---

## üìù Short Approach (Interview-Ready)

- Perform **inorder traversal**  
- Track last seen value (`prev`)  
- Update global min difference  

This is optimal because:
- BST gives sorted order  
- Adjacent values give minimum possible difference  

---

## üßë‚Äçüíª Template Code (Optimal Inorder DFS)

```python
class Solution:
    def getMinimumDifference(self, root):
        self.prev = None
        self.min_diff = float('inf')

        def inorder(node):
            if not node:
                return
            
            inorder(node.left)

            if self.prev is not None:
                self.min_diff = min(self.min_diff, node.val - self.prev)
            
            self.prev = node.val

            inorder(node.right)

        inorder(root)
        return self.min_diff

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def getMinimumDifference(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """
        def dfs(curr):
            if not curr: return float('inf'), -float('inf'), float('inf')
            lmdiff, lmax, lmin = dfs(curr.left)
            rmdiff, rmax, rmin = dfs(curr.right)
            curr_min_diff = min(abs(curr.val-lmax), abs(rmin-curr.val))
            min_diff = min(lmdiff, rmdiff, curr_min_diff)
            return min_diff, max(curr.val, rmax), min(curr.val, lmin) 
            
        return dfs(root)[0]
````
