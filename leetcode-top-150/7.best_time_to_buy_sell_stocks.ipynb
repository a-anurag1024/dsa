# üß© LeetCode 121 ‚Äî Best Time to Buy and Sell Stock  
**Difficulty:** Easy  
**Topic:** Array, Dynamic Programming  

---

## üìù Problem Summary
Given an array `prices` where `prices[i]` represents the price of a stock on the `i-th` day,  
find the **maximum profit** achievable from **one buy and one sell transaction**.

You must buy before you sell ‚Äî if no profit can be made, return `0`.

---

### üí° Example
**Input:** `prices = [7,1,5,3,6,4]`  
**Output:** `5`  
**Explanation:**  
Buy on day 2 (`price = 1`), sell on day 5 (`price = 6`), profit = `6 - 1 = 5`.

---

## ‚öôÔ∏è Approach (Single Pass)

**Idea:**  
Track the **minimum price so far** while iterating, and compute potential profit at each step.  
Update the **maximum profit** when a higher profit is found.

---

### üîç Steps
1. Initialize `min_price = ‚àû` and `max_profit = 0`.  
2. Iterate through each price:
   - Update `min_price` if current price is lower.
   - Compute `profit = price - min_price`.
   - Update `max_profit = max(max_profit, profit)`.  
3. Return `max_profit`.

---

### üß† Key Insight
This is a **greedy one-pass** solution ‚Äî we only need to track the *best buy* (lowest price so far) and *best profit* achievable afterward.

---

### üß© Complexity
- **Time:** `O(n)`  
- **Space:** `O(1)`

---

### ‚úÖ Sample Code (Python)
```python
class Solution(object):
    def maxProfit(self, prices):
        min_price = float('inf')
        max_profit = 0
        for price in prices:
            if price < min_price:
                min_price = price
            else:
                max_profit = max(max_profit, price - min_price)
        return max_profit

````python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        max_profit = 0
        buy_day = 0
        for sell_day in range(1, len(prices)):
            if prices[buy_day] > prices[sell_day]:
                buy_day = sell_day
            else:
                curr_profit = prices[sell_day] - prices[buy_day]
                if curr_profit > max_profit: max_profit = curr_profit
        return max_profit
````
