Given the head of a linked list, remove the **nth node from the end** and return the updated head.

## ðŸš€ Key Idea  
Use a **two-pointer technique** (fast & slow).  
Move the fast pointer `n` steps ahead, then move both pointers together until fast reaches the end.  
Now slow is just before the node to delete.

---

## ðŸ“ Short Approach (Interview-Ready)
1. Create a `dummy` node pointing to `head` (helps remove the head itself).  
2. Move `fast` pointer **n steps forward**.  
3. Move both `fast` and `slow` until `fast` reaches the end.  
4. Now `slow.next` is the node to delete â†’ skip it.  
5. Return `dummy.next`.  

Time Complexity: **O(sz)**  
Space Complexity: **O(1)**

## ðŸ§‘â€ðŸ’» Template Code (Clean & Minimal)
```python
class Solution:
    def removeNthFromEnd(self, head, n):
        dummy = ListNode(0, head)
        fast = slow = dummy

        # Move fast n steps
        for _ in range(n):
            fast = fast.next

        # Move both until fast reaches end
        while fast.next:
            fast = fast.next
            slow = slow.next

        # Delete nth from end
        slow.next = slow.next.next

        return dummy.next

````python
# Better Implementation
# Definition for singly-linked list.

class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        

class Solution(object):
    def removeNthFromEnd(self, head, n):
        """
        :type head: Optional[ListNode]
        :type n: int
        :rtype: Optional[ListNode]
        """
        dummy = ListNode(next=head)
        k_minus_node = dummy
        curr = head
        i = 1
        while curr:
            if i > n :
                k_minus_node = k_minus_node.next
            curr = curr.next
            i += 1
        k_minus_node.next = k_minus_node.next.next if k_minus_node.next else None
        return dummy.next
````
