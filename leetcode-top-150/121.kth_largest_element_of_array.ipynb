## ðŸš€ Approaches (from interview perspective)

### 1ï¸âƒ£ Heap (Min-Heap of size `k`)

**Idea**

* Maintain a min-heap of size `k`
* Push elements into the heap
* If size exceeds `k`, pop the smallest
* The heapâ€™s top is the kth largest

**Why it works**

* You keep only the top `k` largest elements seen so far

**Complexity**

* Time: `O(n log k)`
* Space: `O(k)`

âœ… Very safe and commonly accepted in interviews

### 2ï¸âƒ£ Quickselect (Optimal)

**Idea**

* Based on the partition logic of QuickSort
* Instead of sorting both sides, recurse only where the answer lies
* Target index = `len(nums) - k`

**Complexity**

* Average: `O(n)`
* Worst-case: `O(nÂ²)` (rare with random pivot)

âœ… Preferred if interviewer asks for **no sorting** and **optimal**

2. Pick a pivot and partition array

### âœ… Heap Solution (Clean & Safe)

```python
import heapq

class Solution:
    def findKthLargest(self, nums, k):
        heap = []
        
        for num in nums:
            heapq.heappush(heap, num)
            if len(heap) > k:
                heapq.heappop(heap)
        
        return heap[0]
```

```python
import random

class Solution:
    def findKthLargest(self, nums, k):
        k = len(nums) - k  # convert to kth smallest
        
        def quickselect(l, r):
            pivot = nums[random.randint(l, r)]
            i, j = l, r
            
            while i <= j:
                while nums[i] < pivot:
                    i += 1
                while nums[j] > pivot:
                    j -= 1
                if i <= j:
                    nums[i], nums[j] = nums[j], nums[i]
                    i += 1
                    j -= 1
            
            if k <= j:
                return quickselect(l, j)
            if k >= i:
                return quickselect(i, r)
            return nums[k]
        
        return quickselect(0, len(nums) - 1)
```

````python
import heapq

class Solution(object):
    def findKthLargest(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        heapq.heapify(nums)
        for i in range(len(nums)-k+1):
            k = heapq.heappop(nums)
        return k
````
