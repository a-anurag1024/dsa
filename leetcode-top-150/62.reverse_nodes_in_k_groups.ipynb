You are given the head of a linked list and an integer `k`.

## ğŸš€ Key Idea
Reverse every **contiguous block of k nodes**.  
If fewer than k nodes remain â†’ **do not reverse**.

---

## ğŸ“ Short Approach (Interview-Ready)
1. **Check if k nodes exist** ahead of current pointer.  
   - If not enough nodes â†’ break.
2. **Reverse the next k nodes** using standard pointer reversal.
3. Maintain:
   - `groupPrev` â†’ node before the group  
   - `groupStart` â†’ first node of the group (will become tail after reversal)  
   - `nextGroupHead` â†’ node after the reversed block  
4. **Reconnect pointers** after each reversal.
5. Move to the next group and repeat.

This ensures **O(n)** time and **O(1)** extra space.
