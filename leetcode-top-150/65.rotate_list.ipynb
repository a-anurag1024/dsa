## üìå Problem Summary  
Given the head of a linked list and an integer `k`, rotate the list to the **right** by `k` places.

A right rotation moves the last `k` nodes to the front.

---

## üîç Examples  

### Example 1  
**Input:** `[1,2,3,4,5], k = 2`  
**Output:** `[4,5,1,2,3]`

### Example 2  
**Input:** `[0,1,2], k = 4`  
**Output:** `[2,0,1]`  
*(because rotating by 4 is the same as rotating by 1 when length is 3)*

---

## üöÄ Key Idea  
A rotation is equivalent to:  
1. Finding the list length `L`.  
2. Connecting the tail to the head ‚Üí forming a **circular list**.  
3. Breaking the circle at the correct spot to form the rotated list.

---

## üìù Short Approach (Interview-Ready)

1. If the list is empty or `k == 0`, return head.  
2. Traverse once to compute `length = L` and track the tail.  
3. Convert list into a **cycle**: `tail.next = head`.  
4. Compute effective rotations: `k = k % L`.  
   - If `k == 0`, break the cycle at the tail and return head.  
5. Find the new tail at position `L - k`.  
6. The next node becomes the **new head**.  
7. Break the cycle by setting `new_tail.next = None`.  

Time Complexity: **O(n)**  
Space Complexity: **O(1)**

---

## üßë‚Äçüíª Template Code (Clean & Minimal)
```python
class Solution:
    def rotateRight(self, head, k):
        if not head or not head.next or k == 0:
            return head

        # Step 1: find length and tail
        tail, length = head, 1
        while tail.next:
            tail = tail.next
            length += 1

        # Step 2: form a cycle
        tail.next = head

        # Step 3: find new tail (L - k%L steps from start)
        k = k % length
        steps_to_new_tail = length - k

        new_tail = head
        for _ in range(steps_to_new_tail - 1):
            new_tail = new_tail.next

        # Step 4: break the cycle
        new_head = new_tail.next
        new_tail.next = None

        return new_head

````python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def rotateRight(self, head, k):
        """
        :type head: Optional[ListNode]
        :type k: int
        :rtype: Optional[ListNode]
        """
        # Get the length of list
        n = 0
        curr = head
        last = head
        while curr:
            n += 1
            last = curr
            curr = curr.next
        if n == 0: return head
        else: kp = k % n
        if kp == 0: return head
        curr = head
        for i in range(n):
            if i == n - kp - 1:
                new_head = curr.next
                curr.next = None
                break
            curr = curr.next
        last.next = head
        return new_head
````
