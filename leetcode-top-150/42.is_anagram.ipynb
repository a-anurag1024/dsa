## ðŸ“Œ Problem Summary  
Given two strings `s` and `t`, determine if `t` is an **anagram** of `s`.  
An anagram means:
- Same characters  
- Same frequencies  
- Any order is allowed  

---

## ðŸ§  Key Idea  
Two strings are anagrams **iff**:
- They have the **same length**
- They have the **same character counts**

---

## âœ… Short Approach

### **Method 1: Count Frequency (Best Method)**
1. If lengths differ â†’ return false.
2. Count characters in `s`.
3. Decrement counts using characters in `t`.
4. If any count goes negative â†’ not an anagram.

**Time:** `O(n)`  
**Space:** `O(1)` (only 26 lowercase letters)

---

### **Method 2: Sorting (Simplest but slower)**
Sort both strings and compare results.

**Time:** `O(n log n)`  
**Space:** `O(1)` or `O(n)` depending on language

---

## ðŸ”§ Python Code (Optimal Counting Method)

```python
class Solution:
    def isAnagram(self, s, t):
        if len(s) != len(t):
            return False
        
        freq = [0] * 26
        
        for ch in s:
            freq[ord(ch) - ord('a')] += 1
        
        for ch in t:
            idx = ord(ch) - ord('a')
            freq[idx] -= 1
            if freq[idx] < 0:
                return False
        
        return True

````python
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        hm = {}
        if len(s) != len(t): return False
        for i in range(len(s)):
            if s[i] not in hm: hm[s[i]] = 1
            else: hm[s[i]] += 1
            if t[i] not in hm: hm[t[i]] = -1
            else: hm[t[i]] -= 1
        for c in hm:
            if hm[c] != 0 : return False
        return True
````
