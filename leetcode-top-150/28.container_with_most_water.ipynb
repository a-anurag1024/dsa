# ü™£ LeetCode 11 ‚Äî Container With Most Water (Medium)

---

## üß© Problem Summary

Given an array `height` where each element represents the height of a vertical line at position `i`,  
find two lines that form a container that holds the **maximum amount of water** between them.

You cannot tilt the container ‚Äî the sides are vertical.

---

## üß† Examples

### Example 1:
**Input:**  
`height = [1,8,6,2,5,4,8,3,7]`  
**Output:** `49`  
**Explanation:**  
The lines at indices `1` and `8` form the container with the maximum area.  
`min(8,7) * (8 - 1) = 7 * 7 = 49`

---

### Example 2:
**Input:**  
`height = [1,1]`  
**Output:** `1`  
**Explanation:**  
Only one possible container ‚Üí `min(1,1) * (1 - 0) = 1`

---

## ‚öôÔ∏è Approach (Two-Pointer Greedy)

### üîπ Key Idea:
Use two pointers to scan from **both ends inward**, calculating the area at each step.  
Move the pointer that has the **smaller height**, since a taller line won‚Äôt increase the area if the smaller one remains.

---

### üîπ Step-by-Step Logic

1. Initialize two pointers:  
   `left = 0`, `right = len(height) - 1`
2. Compute the area at each step:  
   `area = (right - left) * min(height[left], height[right])`
3. Track the **maximum area** found so far.
4. Move the pointer with the **smaller height** inward ‚Äî since it limits the container height.
5. Continue until `left >= right`.

---

## üßë‚Äçüíª Python Sketch

```python
def maxArea(height):
    left, right = 0, len(height) - 1
    max_water = 0

    while left < right:
        # Calculate current area
        width = right - left
        h = min(height[left], height[right])
        max_water = max(max_water, width * h)

        # Move the smaller height pointer inward
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1

    return max_water


# My solution

````python
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        i = 0
        j = len(height) - 1
        max_area = (j-i)*min(height[i], height[j])
        while i <= j:
            area = (j-i)*min(height[i], height[j])
            if area > max_area: max_area = area
            if height[i] > height[j]: j-= 1
            else: i+=1
        return max_area
````
