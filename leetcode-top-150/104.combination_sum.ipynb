### **Problem Summary**

Given an array of **distinct integers** `candidates` and an integer `target`, return **all unique combinations** where the chosen numbers sum to `target`.

* Each number can be used **unlimited times**
* Order of numbers **does not matter**
* Combinations must be **unique** (frequency matters, not order)

---

### **Key Observations**

* This is a **combination** problem, not permutation.
* Reuse of elements is allowed â†’ recursion can stay on the **same index**.
* Sorting helps with **early pruning**.
* Classic **DFS / backtracking with pruning**.

---

## ðŸ”¹ Short Approach Sketch (Backtracking)

1. Sort `candidates` to enable pruning.
2. Use DFS with:

   * `start`: index to avoid duplicate combinations
   * `path`: current combination
   * `remaining`: target left to reach
3. If `remaining == 0`, store the current combination.
4. If `remaining < 0`, stop exploring (prune).
5. For each candidate starting from `start`:

   * Choose it
   * Recurse with **same index** (since reuse is allowed)
   * Backtrack

---

### **Time & Space Complexity**

* **Time:** Exponential (bounded by valid combinations, â‰¤ 150)
* **Space:** `O(target / min(candidates))` recursion depth

---

## ðŸ”¹ Python Solution (Backtracking â€“ Optimal)

```python
class Solution(object):
    def combinationSum(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        candidates.sort()
        res = []

        def backtrack(start, path, remaining):
            # Found valid combination
            if remaining == 0:
                res.append(path[:])
                return

            # Prune invalid paths
            if remaining < 0:
                return

            for i in range(start, len(candidates)):
                # Early pruning since sorted
                if candidates[i] > remaining:
                    break
                path.append(candidates[i])
                # Reuse allowed â†’ pass i again
                backtrack(i, path, remaining - candidates[i])
                path.pop()

        backtrack(0, [], target)
        return res
```

---

## ðŸ”¹ Why `start` Index Is Important

* Prevents duplicate combinations like `[2,3]` and `[3,2]`
* Ensures combinations are built in **non-decreasing order**



# My Solution

````python
class Solution(object):
    def combinationSum(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        combinations = []
        curr = []
        candidates.sort()
        def helper(curr_sum, pos):
            if curr_sum == target:
                combinations.append(curr[:])
                return
            if curr_sum > target: return 
            for i in range(pos, len(candidates)):
                can = candidates[i]
                curr.append(can)
                curr_sum += can
                helper(curr_sum, i)
                curr_sum -= can
                curr.pop()
            return
        helper(0, 0)
        return combinations
````
