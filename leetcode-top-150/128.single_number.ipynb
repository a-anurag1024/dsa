## ðŸ§  Key Insight (XOR Magic)

The solution relies on the properties of **XOR (`^`)**:

| Property           | Explanation                    |
| ------------------ | ------------------------------ |
| `a ^ a = 0`        | A number cancels itself        |
| `a ^ 0 = a`        | XOR with zero keeps the number |
| XOR is commutative | Order doesnâ€™t matter           |

So if you XOR all numbers:

* All pairs cancel out
* Only the single number remains

3. Return `result`

## ðŸ§© Python Solution (Optimal)

```python
class Solution(object):
    def singleNumber(self, nums):
        res = 0
        for num in nums:
            res ^= num
        return res
```

# My Solution

````python
class Solution(object):
    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        p = 0
        for n in nums:
            p = p ^ n
        return p
````
