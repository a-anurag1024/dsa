### **Problem Summary**

You are given a **sorted array with distinct values** that has been **rotated at an unknown index**.
Given a `target`, return its index if present, otherwise return `-1`.

**Constraint:** `O(log n)` time.

---

### **Key Observations**

* The array is composed of **two sorted halves**
* At least **one half is always sorted**
* Binary search can still be applied by identifying the sorted half
* No duplicates â†’ comparisons are unambiguous

---

## ðŸ”¹ Short Approach Sketch (Modified Binary Search)

1. Initialize `left = 0`, `right = n - 1`
2. While `left <= right`:

   * Compute `mid`
   * If `nums[mid] == target`, return `mid`
3. Determine which half is sorted:

   * If `nums[left] <= nums[mid]` â†’ **left half sorted**

     * Check if `target` lies in this range
   * Else â†’ **right half sorted**

     * Check if `target` lies in this range
4. Narrow the search space accordingly
5. If not found, return `-1`

---

### **Time & Space Complexity**

* **Time:** `O(log n)`
* **Space:** `O(1)`

---

## ðŸ”¹ Python Solution (Optimal)

```python
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        left, right = 0, len(nums) - 1

        while left <= right:
            mid = (left + right) // 2

            if nums[mid] == target:
                return mid

            # Left half is sorted
            if nums[left] <= nums[mid]:
                if nums[left] <= target < nums[mid]:
                    right = mid - 1
                else:
                    left = mid + 1
            # Right half is sorted
            else:
                if nums[mid] < target <= nums[right]:
                    left = mid + 1
                else:
                    right = mid - 1

        return -1
```


### ðŸ’¡ Tip

If asked:

> *What breaks if duplicates are allowed?*

Answer:

> *We lose the guarantee of identifying a sorted half, which can degrade the algorithm to O(n).*


# My Solution

````python
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        st, en = 0, len(nums)-1
        if nums[0] == target: return 0
        if nums[-1] == target: return len(nums) - 1
        is_left_band = target > nums[0]
        while st <= en:
            mid = (st + en) // 2
            if nums[mid] == target: return mid
            move_what = None
            if nums[mid] < nums[st]:
                if is_left_band: move_what = "end"
            elif nums[mid] > nums[en]:
                if not is_left_band: move_what = "start"
            if not move_what:
                if nums[mid] > target: move_what = "end"
                else: move_what = "start"
            if move_what == "start":
                st = mid + 1
            else: en = mid - 1
        return -1
                    
````
