## ðŸ§  Key Insight

Simple XOR doesnâ€™t work here (because `a ^ a ^ a = a`).

Instead, we use **bit counting** or **bitmask state tracking**.

At each bit position:

* Bits from numbers appearing three times sum to a multiple of 3
* The unique number contributes the remainder

1. For each of the 32 bit positions:

## ðŸ§© Python Solution (Bit Counting â€“ Clear & Safe)

```python
class Solution(object):
    def singleNumber(self, nums):
        result = 0
        
        for i in range(32):
            bit_sum = 0
            for num in nums:
                bit_sum += (num >> i) & 1
            
            if bit_sum % 3:
                result |= (1 << i)
        
        # Handle negative numbers (32-bit signed)
        if result >= 2**31:
            result -= 2**32
        
        return result
```

## ðŸ§  Alternative (Advanced â€“ Bitmask FSM)

This is the **cleanest interview solution**, but harder to derive:

```python
class Solution(object):
    def singleNumber(self, nums):
        ones = twos = 0
        
        for num in nums:
            ones = (ones ^ num) & ~twos
            twos = (twos ^ num) & ~ones
        
        return ones
```

**Intuition**:

* `ones` tracks bits seen once
* `twos` tracks bits seen twice
* Bits seen three times are cleared automatically

# My Solution

````python
class Solution(object):
    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        res = 0
        for i in range(32):
            count = 0
            for n in nums:
                if (n >> i) & 1: count += 1
            if count % 3 != 0: res = res | 1 << i
        if res >= 2**31:
            res -= 2**32
        return res
````
