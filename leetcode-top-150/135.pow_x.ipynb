## ‚ö° LeetCode 50: Pow(x, n) ‚Äî Quick Review

### üß© Problem Summary

Implement `pow(x, n)` to compute ( x^n ):

* `x` is a floating-point number
* `n` is an integer (can be **negative**)
* You must do this **efficiently**

---

## üîç Key Observations

1. **Negative exponent**:

   * ( x^{-n} = \frac{1}{x^n} )
2. **Naive multiplication** (`n` times) is too slow for large `n`
3. Use **Exponentiation by Squaring** to reduce time complexity

---

## üß† Approach Sketch (Exponentiation by Squaring)

### Core Idea

Break the exponent in half repeatedly:

* If `n` is even:

  ```
  x^n = (x^2)^(n/2)
  ```
* If `n` is odd:

  ```
  x^n = x * x^(n-1)
  ```

### Steps

1. If `n < 0`, convert:

   * `x = 1/x`, `n = -n`
2. Initialize `result = 1`
3. While `n > 0`:

   * If `n` is odd ‚Üí multiply `result` by `x`
   * Square `x`
   * Divide `n` by 2
4. Return `result`

---

## ‚úÖ Python Solution (Optimal & Interview-Standard)

```python
class Solution(object):
    def myPow(self, x, n):
        if n == 0:
            return 1.0
        
        if n < 0:
            x = 1 / x
            n = -n
        
        result = 1.0
        
        while n > 0:
            if n % 2 == 1:
                result *= x
            x *= x
            n //= 2
        
        return result
```

---

## ‚è±Ô∏è Complexity Analysis

* **Time:** `O(log n)`
* **Space:** `O(1)`

# My Solution

````python
# Not space optimized version as using recursion stack

class Solution(object):
    def myPow(self, x, n):
        """
        :type x: float
        :type n: int
        :rtype: float
        """
        if x == 0: return 0
        if n == 0: return 1
        hm = {}
        def helper(x, n1):
            if abs(n1) == 1: return x if n1 > 0 else 1 / x
            mid = n1 // 2
            left = helper(x, mid) if (x, mid) not in hm else hm[(x, mid)]
            if (x,mid) not in hm: hm[(x, mid)] = left
            if n1 != mid:
                right = helper(x, n1-mid) if (x, n1-mid) not in hm else hm[(x, n1-mid)]
                if (x, n1-mid) not in hm: hm[(x, n1-mid)] = right
            else: right = 1
            return left * right
        return helper(x, n)
````
