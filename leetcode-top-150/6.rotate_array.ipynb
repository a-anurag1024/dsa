Given an integer array `nums`, rotate it **to the right by `k` steps**, where `k` is non-negative.  
The rotation must be done **in-place**, i.e., without using extra space for another array.

## âš™ï¸ Approach (Reverse Technique)
The key observation:  
Rotating the array right by `k` steps = moving the last `k` elements to the front.  

**Steps:**
1. Normalize `k` using `k %= len(nums)` to handle large `k`.  
2. **Reverse the entire array.**  
3. **Reverse the first `k` elements.**  
4. **Reverse the remaining `n - k` elements.**

Example:  
`nums = [1,2,3,4,5,6,7], k = 3`  
â†’ Reverse all â†’ `[7,6,5,4,3,2,1]`  
â†’ Reverse first 3 â†’ `[5,6,7,4,3,2,1]`  
â†’ Reverse rest â†’ `[5,6,7,1,2,3,4]`

## â±ï¸ Complexity
- **Time:** O(n) â€” each element reversed at most twice.  
- **Space:** O(1) â€” in-place operation.

## âœ… Code Snippet (Python)
```python
def rotate(nums, k):
    n = len(nums)
    k %= n

    def reverse(start, end):
        while start < end:
            nums[start], nums[end] = nums[end], nums[start]
            start += 1
            end -= 1

    reverse(0, n - 1)
    reverse(0, k - 1)
    reverse(k, n - 1)


## Approach - 2 (cyclic replacement)
Idea:
- Treat rotation as moving elements to (i + k) % n.
- Move each element to its new place in cycles until all are placed.
```python
def rotate(nums, k):
    n = len(nums)
    k %= n
    count = 0
    start = 0
    while count < n:
        current = start
        prev = nums[start]
        while True:
            next_idx = (current + k) % n
            nums[next_idx], prev = prev, nums[next_idx]
            current = next_idx
            count += 1
            if start == current:
                break
        start += 1
```
## â±ï¸ Complexity
- **Time:** O(n) â€” each element placed once in one of the cycles.  
- **Space:** O(1) â€” in-place operation.
