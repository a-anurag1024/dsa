## ðŸ“˜ Problem Summary
You are given a **sorted**, **unique** integer array `nums`.  
You must group consecutive numbers into "ranges" and return them in the smallest sorted list.

A range is formatted as:
- `"a->b"` if `a != b`
- `"a"` if the range contains only one element

---

## ðŸ§  Key Idea
Since the array is sorted and unique, consecutive numbers differ by exactly **1**.  
We can scan the array once and track the **start** and **end** of each consecutive segment.

---

## âœ… Approach (O(n) Time, O(1) Extra Space)
1. Initialize `start = nums[0]`.  
2. Iterate through the array:
   - If `nums[i] == nums[i-1] + 1`, continue the current range.
   - Otherwise:
     - Close the previous range.
     - Start a new range from `nums[i]`.
3. After the loop, push the final range.

**Formatting rule:**
- If `start == end`: store `"start"`
- Else: store `"start->end"`

---

## ðŸ“ Example
Input: `[0,1,2,4,5,7]`  
Output: `["0->2", "4->5", "7"]`

---

## ðŸ§© Sample Python Implementation
```python
class Solution(object):
    def summaryRanges(self, nums):
        if not nums:
            return []

        ranges = []
        start = nums[0]

        for i in range(1, len(nums)):
            if nums[i] != nums[i - 1] + 1:   # Break in sequence
                end = nums[i - 1]
                ranges.append(str(start) if start == end else f"{start}->{end}")
                start = nums[i]

        # Add the final range
        end = nums[-1]
        ranges.append(str(start) if start == end else f"{start}->{end}")

        return ranges

````python
class Solution(object):
    def summaryRanges(self, nums):
        """
        :type nums: List[int]
        :rtype: List[str]
        """
        if len(nums) == 0: return []
        curr_start = 0
        ret = []
        for end in range(len(nums)):
            if end + 1 < len(nums) and nums[end + 1] != nums[end] + 1:
                if curr_start != end:
                    ret.append(str(nums[curr_start]) + "->" + str(nums[end]))
                else: ret.append(str(nums[curr_start]))
                curr_start = end + 1
        end = len(nums)-1
        if curr_start != end:
            ret.append(str(nums[curr_start]) + "->" + str(nums[end]))
        else: ret.append(str(nums[curr_start]))
        return ret
````
