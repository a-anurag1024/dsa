## ðŸ§© Problem Summary

You are given:
- An array of words `words`
- A target line width `maxWidth`

You must **justify** the text so that:
1. Each line has **exactly `maxWidth`** characters.
2. Words are packed **greedily** (as many as fit per line).
3. Spaces between words are distributed **evenly**.
4. If spaces canâ€™t be distributed evenly, **left slots get extra spaces**.
5. The **last line** is **left-justified** (no extra spacing between words).

## âš™ï¸ Approach (Greedy Line Construction + Space Distribution)

### ðŸ”¹ Key Idea:

1. **Greedily group words** into lines until adding the next word would exceed `maxWidth`.

2. For each line:

   * If itâ€™s the **last line** or has **only one word**, **left-justify** it.
   * Otherwise, **distribute spaces evenly** between words:

     * Compute total spaces = `maxWidth - total_characters`.
     * Divide spaces into even chunks across gaps.
     * If not evenly divisible, assign extra spaces to the **leftmost gaps**.

3. **Join** the words and pad with spaces as needed to reach `maxWidth`.

## ðŸ§‘â€ðŸ’» Python Sketch

```python
def fullJustify(words, maxWidth):
    res, cur, num_letters = [], [], 0

    for w in words:
        # Check if adding the next word exceeds the maxWidth
        if num_letters + len(w) + len(cur) > maxWidth:
            spaces = maxWidth - num_letters
            if len(cur) == 1:
                res.append(cur[0] + ' ' * spaces)
            else:
                space_between = spaces // (len(cur) - 1)
                extra = spaces % (len(cur) - 1)
                line = ''
                for i in range(len(cur) - 1):
                    line += cur[i]
                    line += ' ' * (space_between + (1 if i < extra else 0))
                line += cur[-1]
                res.append(line)
            cur, num_letters = [], 0
        cur.append(w)
        num_letters += len(w)

    # Last line (left-justified)
    last_line = ' '.join(cur)
    res.append(last_line + ' ' * (maxWidth - len(last_line)))

    return res
```

# My solution

`````python
class Solution(object):
    def fullJustify(self, words, maxWidth):
        output_strings = []
        occupied_width = 0
        words_added = 0

        for i, word in enumerate(words):
            # check if adding this word exceeds the line width
            if occupied_width + len(word) + words_added > maxWidth:
                remaining_spaces = maxWidth - occupied_width
                if words_added > 1:
                    spaces_to_all = remaining_spaces // (words_added - 1)
                    extra_spaces = remaining_spaces % (words_added - 1)
                    string_to_add = ""
                    for k in range(words_added - 1):
                        string_to_add += words[i - words_added + k]
                        string_to_add += " " * (spaces_to_all + (1 if k < extra_spaces else 0))
                    string_to_add += words[i - 1]  # last word in line
                else:
                    string_to_add = words[i - 1] + " " * remaining_spaces
                output_strings.append(string_to_add)
                occupied_width = 0
                words_added = 0

            occupied_width += len(word)
            words_added += 1

        # Handle last line (left justified)
        if words_added:
            remaining = words[-words_added:]
            string_to_add = " ".join(remaining)
            string_to_add += " " * (maxWidth - len(string_to_add))
            output_strings.append(string_to_add)

        return output_strings
`````
