### **Problem Summary**

You are given an `m Ã— n` matrix with:

1. Each row sorted in non-decreasing order
2. The first element of each row **greater than** the last element of the previous row

Given a `target`, return `True` if it exists in the matrix, otherwise `False`.

**Constraint:** `O(log(m Â· n))` time complexity.

---

### **Key Observations**

* The matrix behaves like a **flattened sorted array**
* Row ordering guarantees **global sorted order**
* This allows **binary search** over all elements
* Index mapping is the key trick

---

## ðŸ”¹ Short Approach Sketch (Binary Search on Virtual Array)

1. Treat the matrix as a **1D sorted array** of size `m Ã— n`
2. Perform binary search on indices `[0 â€¦ m*n - 1]`
3. Convert 1D index â†’ 2D coordinates:

   ```python
   row = mid // n
   col = mid % n
   ```
4. Compare `matrix[row][col]` with `target`
5. Return `True` if found, else `False`

---

### **Time & Space Complexity**

* **Time:** `O(log(m Â· n))`
* **Space:** `O(1)`

## ðŸ”¹ Python Solution (Optimal)

```python
class Solution(object):
    def searchMatrix(self, matrix, target):
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """
        if not matrix or not matrix[0]:
            return False

        m, n = len(matrix), len(matrix[0])
        left, right = 0, m * n - 1

        while left <= right:
            mid = (left + right) // 2
            row = mid // n
            col = mid % n

            if matrix[row][col] == target:
                return True
            elif matrix[row][col] < target:
                left = mid + 1
            else:
                right = mid - 1

        return False
```

````python
class Solution(object):
    def searchMatrix(self, matrix, target):
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """
        m = len(matrix)
        n = len(matrix[0])
        # find the row
        st, en = 0, m - 1
        while st <= en:
            mid = (st+en) // 2
            if matrix[mid][0] == target: return True
            if matrix[mid][0] < target: st = mid + 1
            else: en = mid - 1
        row = st - 1
        if row < 0: return False
        # find the column
        st, en = 0, n - 1
        while st <= en:
            mid = (st+en) // 2
            if matrix[row][mid] == target: return True
            if matrix[row][mid] < target: st = mid + 1
            else: en = mid - 1
        return False
````
