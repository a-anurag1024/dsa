## ðŸ”¹ Quick Review: LeetCode 46 â€” Permutations

### **Problem Summary**

Given an array `nums` of **distinct integers**, return **all possible permutations**.

* Order **matters**
* Each number must be used **exactly once**
* Result size = `n!`

---

### **Key Observations**

* This is a **permutation generation** problem.
* Unlike combinations, **ordering is important**.
* At each position, you can choose **any unused element**.
* Classic **backtracking with used tracking**.

---

## ðŸ”¹ Short Approach Sketch (Backtracking)

1. Maintain:

   * `path`: current permutation being built
   * `used`: boolean array to mark chosen elements
2. If `len(path) == len(nums)`:

   * Add a copy of `path` to result.
3. Iterate through all indices:

   * If not used, choose the element
   * Mark as used â†’ recurse â†’ unmark (backtrack)

---

### **Time & Space Complexity**

* **Time:** `O(n!)`
* **Space:** `O(n)` recursion depth (excluding output)

---

## ðŸ”¹ Python Solution (Backtracking)

```python
class Solution(object):
    def permute(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        res = []
        used = [False] * len(nums)

        def backtrack(path):
            if len(path) == len(nums):
                res.append(path[:])
                return

            for i in range(len(nums)):
                if used[i]:
                    continue
                used[i] = True
                path.append(nums[i])
                backtrack(path)
                path.pop()
                used[i] = False

        backtrack([])
        return res
```

## ðŸ”¹ Alternative Approach (In-Place Swapping)

Uses less extra space (no `used` array):

```python
class Solution(object):
    def permute(self, nums):
        res = []

        def backtrack(start):
            if start == len(nums):
                res.append(nums[:])
                return

            for i in range(start, len(nums)):
                nums[start], nums[i] = nums[i], nums[start]
                backtrack(start + 1)
                nums[start], nums[i] = nums[i], nums[start]

        backtrack(0)
        return res
```


# My Solution

````python
class Solution(object):
    def permute(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        permutations = []
        visited = set()
        n = len(nums)
        def helper(curr):
            if len(curr) == n:
                permutations.append(curr[:])
            for num in nums:
                if num in visited: continue
                visited.add(num)
                curr.append(num)
                helper(curr)
                curr.pop()
                visited.remove(num)
        helper([])
        return permutations
````
