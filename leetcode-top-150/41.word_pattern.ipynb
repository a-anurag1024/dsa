## ðŸ“Œ Problem Summary
Given:
- A **pattern** string (e.g., `"abba"`)
- A sentence **s** split into words (e.g., `"dog cat cat dog"`)

Determine if **s follows the same pattern**, meaning there is a **bijective mapping** between:
- each character in `pattern`
- each unique word in `s`

### Requirements:
- Each pattern letter â†’ **exactly one unique word**
- Each word â†’ **exactly one unique pattern letter**  
  (one-to-one mapping, no duplicates either way)

---

## ðŸ§  Key Idea
This is the same concept as **Isomorphic Strings**, but:
- instead of characters â†’ map characters to **words**
- ensure both:
  - mapping from pattern â†’ words is consistent
  - words are not reused by other pattern characters

---

## âœ… Short Approach

### **Steps**
1. Split `s` into a list of words.
2. If `len(pattern) != len(words)` â†’ immediately false.
3. Maintain two structures:
   - `p_to_w`: pattern char â†’ word
   - `used_words`: to ensure unique mapping
4. For each index `i`:
   - If pattern char already mapped, ensure the mapped word matches.
   - Otherwise:
     - Ensure the current word is not used by another char.
     - Assign new mapping.

### Complexity
- **Time:** `O(n)`  
- **Space:** `O(n)` (for the words used)

---

## ðŸ”§ Python Code (Clean & Simple)

```python
class Solution:
    def wordPattern(self, pattern, s):
        words = s.split()
        
        if len(pattern) != len(words):
            return False
        
        p_to_w = {}
        used = set()
        
        for ch, word in zip(pattern, words):
            if ch in p_to_w:
                if p_to_w[ch] != word:
                    return False
            else:
                if word in used:
                    return False
                p_to_w[ch] = word
                used.add(word)
        
        return True

````python
class Solution(object):
    def wordPattern(self, pattern, s):
        """
        :type pattern: str
        :type s: str
        :rtype: bool
        """
        hm = {}
        done_words = set()
        nw = len(pattern)
        nl = len(s)
        j = 0
        for i, c in enumerate(pattern):
            st = j
            while j<nl and s[j] != ' ':
                j += 1
            word = s[st:j]
            if c not in hm:
                if word in done_words: return False
                hm[c] = word
                done_words.add(word)
            else:
                if hm[c] != word: return False
            j += 1
            if j >= nl and i != nw - 1: return False
        if j == nl+1: return True
        else: return False
````
