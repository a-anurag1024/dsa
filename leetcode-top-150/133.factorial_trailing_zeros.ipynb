Given an integer `n`, return the **number of trailing zeroes** in `n!` (factorial of `n`).

Trailing zeroes are produced by factors of **10**, and:

> `10 = 2 Ã— 5`

## ðŸ’¡ Key Insight (Very Important)

* In `n!`, there are **far more factors of 2 than 5**
* So the number of trailing zeroes depends **only on how many times 5 appears as a factor**

## ðŸ§  Approach Sketch (Optimal)

1. Count numbers divisible by `5` â†’ contribute **one** factor of 5
2. Count numbers divisible by `25 = 5Â²` â†’ contribute **one extra** factor
3. Count numbers divisible by `125 = 5Â³`, and so on
4. Stop when `5^k > n`

Mathematically:

```
trailing_zeroes = n//5 + n//25 + n//125 + ...
```

## âœ… Python Solution (Optimal & Interview-Standard)

```python
class Solution(object):
    def trailingZeroes(self, n):
        count = 0
        while n > 0:
            n //= 5
            count += n
        return count
```

# My Solution

````python
class Solution(object):
    def trailingZeroes(self, n):
        """
        :type n: int
        :rtype: int
        """
        fives = 0
        k = 5
        while n >= k:
            fives += n // k
            k *= 5
        return fives
````
