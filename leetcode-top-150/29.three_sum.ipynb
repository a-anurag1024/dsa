## üß© Problem Summary

Given an integer array `nums`, return **all unique triplets** `[nums[i], nums[j], nums[k]]`  
such that:
\[
nums[i] + nums[j] + nums[k] = 0
\]
and \( i \neq j \neq k \).

The solution must not contain **duplicate triplets**.

---

## üß† Examples

### Example 1:
**Input:**  
`nums = [-1,0,1,2,-1,-4]`  
**Output:**  
`[[-1,-1,2], [-1,0,1]]`

**Explanation:**  
Triplets that sum to zero:
- (-1) + (-1) + 2 = 0  
- (-1) + 0 + 1 = 0

---

### Example 2:
**Input:**  
`nums = [0,1,1]`  
**Output:**  
`[]`  
**Explanation:**  
No three numbers sum to zero.

---

### Example 3:
**Input:**  
`nums = [0,0,0]`  
**Output:**  
`[[0,0,0]]`  
**Explanation:**  
Only one valid triplet.

---

## ‚öôÔ∏è Approach (Sorting + Two Pointers)

### üîπ Key Idea:
1. **Sort** the array ‚Äî helps skip duplicates and apply two-pointer logic.
2. **Fix one element** `nums[i]` as the first number of the triplet.
3. Use **two pointers** (`left`, `right`) to find pairs that sum to `-nums[i]`.
4. Move pointers accordingly:
   - If the sum < 0 ‚Üí move `left` rightward (to increase sum).
   - If the sum > 0 ‚Üí move `right` leftward (to decrease sum).
   - If the sum = 0 ‚Üí store the triplet and skip duplicates.

---

### üîπ Step-by-Step Example

Input: `nums = [-1,0,1,2,-1,-4]`  
Sorted: `[-4, -1, -1, 0, 1, 2]`

1Ô∏è‚É£ Fix `-4` ‚Üí Need sum = `4` ‚Üí No match  
2Ô∏è‚É£ Fix `-1` (at i=1):
   - left=2 (`-1`), right=5 (`2`)
   - (-1) + (-1) + 2 = 0 ‚úÖ Add [-1, -1, 2]
   - Move pointers to skip duplicates  
3Ô∏è‚É£ Fix next `-1` (duplicate) ‚Üí skip  
4Ô∏è‚É£ Fix `0`: left=4, right=5 ‚Üí 0 + 1 + 2 > 0 ‚Üí move right--

‚úÖ Final result ‚Üí `[[-1,-1,2], [-1,0,1]]`

```python
def threeSum(nums):
    nums.sort()
    res = []

    for i in range(len(nums)):
        if i > 0 and nums[i] == nums[i - 1]:
            continue  # skip duplicate anchors

        left, right = i + 1, len(nums) - 1
        while left < right:
            total = nums[i] + nums[left] + nums[right]

            if total < 0:
                left += 1
            elif total > 0:
                right -= 1
            else:
                res.append([nums[i], nums[left], nums[right]])
                # Skip duplicates for left and right
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                left += 1
                right -= 1
    return res

````python
# Not efficient (hashmap not required)
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        nums.sort()
        ans_hm = {}
        for i, x in enumerate(nums):
            target = -x
            j = 0 if i != 0 else 1
            k = len(nums) - 1 if i!=len(nums)-1 else len(nums) - 2
            while j < k:
                curr_sum = nums[j] + nums[k]
                if curr_sum > target:
                    if k-1 != i: k -= 1
                    else: k -= 2
                elif curr_sum < target:
                    if j+1 != i: j += 1
                    else: j += 2
                else:
                    max_el = max(nums[i], nums[j], nums[k])
                    min_el = min(nums[i], nums[j], nums[k])
                    mid_el = nums[i] + nums[j] + nums[k] - max_el - min_el
                    triplet = [min_el, mid_el, max_el]
                    ans_hm[(min_el, max_el)] = triplet
                    if nums[j+1] == nums[j]: 
                        if j+1 != i: j+=1
                        else: j+=2
                    else:
                        if k-1 != i: k-=1
                        else: k-=2

        return list(ans_hm.values())

        
````
