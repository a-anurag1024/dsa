### **Problem Summary**

Given the head of a **singly linked list**, sort it in **ascending order** and return the sorted list.

---

### **Key Observations**

* You **cannot** efficiently random-access a linked list
* Built-in sort or array conversion breaks the spirit of the problem
* Required optimal solution: **O(n log n)** time
* **Merge Sort** is the natural fit for linked lists

  * Easy split
  * No extra array
  * Stable and efficient

---

## ðŸ”¹ Short Approach Sketch (Merge Sort on Linked List)

1. **Base case**:

   * Empty list or single node â†’ already sorted
2. **Split the list** into two halves using **slow & fast pointers**
3. **Recursively sort** both halves
4. **Merge** the two sorted linked lists
5. Return merged head

---

### **Time & Space Complexity**

* **Time:** `O(n log n)`
* **Space:** `O(log n)` (recursion stack)

---

## ðŸ”¹ Python Solution (Optimal â€“ Merge Sort)

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution(object):
    def sortList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        # Base case
        if not head or not head.next:
            return head

        # Step 1: Split the list
        slow, fast = head, head.next
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        mid = slow.next
        slow.next = None

        # Step 2: Sort both halves
        left = self.sortList(head)
        right = self.sortList(mid)

        # Step 3: Merge
        return self.merge(left, right)

    def merge(self, l1, l2):
        dummy = ListNode(0)
        curr = dummy

        while l1 and l2:
            if l1.val < l2.val:
                curr.next = l1
                l1 = l1.next
            else:
                curr.next = l2
                l2 = l2.next
            curr = curr.next

        curr.next = l1 if l1 else l2
        return dummy.next
```

---

## ðŸ”¹ Why Merge Sort Works Best Here

* Linked lists allow **O(1)** node rearrangement
* No index-based operations needed
* Splitting and merging are pointer-friendly

# My Solution

````python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution(object):
    def sortList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        def merge(l1, l2):
            dummy = ListNode()
            curr = dummy
            while l1 or l2:
                if not l1: curr.next = l2
                elif not l2: curr.next = l1
                elif l1.val >= l2.val: curr.next = l2
                else: curr.next = l1
                if curr.next == l1: l1 = l1.next
                else: l2 = l2.next
                curr = curr.next
            return dummy.next
        def sort(l):
            if not l or not l.next: return l

            slow = fast = l
            while fast.next and fast.next.next:
                slow = slow.next
                fast = fast.next.next

            mid = slow
            l2 = mid.next
            mid.next = None
            left = sort(l)
            right = sort(l2)
            return merge(left, right)

        return sort(head)
````
