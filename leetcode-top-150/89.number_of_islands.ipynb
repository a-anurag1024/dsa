## ğŸ“˜ Problem Summary
You are given an **m Ã— n grid** filled with:
- `'1'` â†’ **land**
- `'0'` â†’ **water**

An **island** is a group of connected `'1'`s (connected *horizontally or vertically*).  
Your task is to **count the number of islands**.

---

## ğŸ§  Key Idea
Whenever you find a `'1'`, that marks the **start of a new island**.  
From there, you:
- **DFS/BFS flood-fill** all connected `'1'`s,
- **mark them as visited** (convert to `'0'` or use a visited set),
- and continue scanning the grid.

Each time you start a new DFS/BFS, you have discovered **one island**.

---

## ğŸš€ Approach (DFS â€” Most Common)
1. Initialize a counter `islands = 0`.
2. Loop through every cell `(i, j)` in the grid.
3. When you find a `'1'`:
   - Increment `islands`.
   - Call DFS to mark the entire connected component as visited by turning `'1'` â†’ `'0'`.
4. Continue scanning the rest of the grid.
5. Return `islands`.

---

## ğŸ§© Time & Space Complexity
- **Time:** `O(m * n)` â€” every cell is visited once.
- **Space:** `O(m * n)` worst-case recursion (DFS stack)  
  or `O(min(m, n))` for BFS.

## ğŸ“ Sample DFS Code (Python)
```python
class Solution:
    def numIslands(self, grid):
        if not grid: return 0
        
        rows, cols = len(grid), len(grid[0])
        islands = 0
        
        def dfs(r, c):
            if r < 0 or c < 0 or r >= rows or c >= cols or grid[r][c] == "0":
                return
            grid[r][c] = "0"          # mark visited
            dfs(r+1, c)
            dfs(r-1, c)
            dfs(r, c+1)
            dfs(r, c-1)

        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == "1":
                    islands += 1
                    dfs(r, c)
        
        return islands

````python
class Solution(object):
    def numIslands(self, grid):
        """
        :type grid: List[List[str]]
        :rtype: int
        """
        m = len(grid)
        n = len(grid[0])
        num_of_islands = 0
        island_number_grid = grid[:]
        def dfs(i, j, island_number):
            if i > 0 and grid[i-1][j] == "1":
                island_number_grid[i-1][j] = island_number
                dfs(i-1, j, island_number)
            if j > 0 and grid[i][j-1] == "1":
                island_number_grid[i][j-1] = island_number
                dfs(i, j-1, island_number)
            if i < m - 1 and grid[i+1][j] == "1":
                island_number_grid[i+1][j] = island_number
                dfs(i+1, j, island_number)
            if j < n - 1 and grid[i][j+1] == "1":
                island_number_grid[i][j+1] = island_number
                dfs(i, j+1, island_number)
        for i in range(m):
            for j in range(n):
                if grid[i][j] == "1":
                    if island_number_grid[i][j] in ["0", "1"]:
                        num_of_islands += 1
                        if i < m - 1 and grid[i+1][j] == "1": dfs(i+1, j, num_of_islands)
                        if j < n - 1 and grid[i][j+1] == "1": dfs(i, j+1, num_of_islands)
        return num_of_islands
````
