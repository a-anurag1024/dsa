# üß¨ LeetCode 133: Clone Graph ‚Äî Quick Review

## üìò Problem Summary
You are given a reference to a **node in a connected, undirected graph**.  
Each node has:
- `val` ‚Üí an integer value  
- `neighbors` ‚Üí list of adjacent nodes  

Your task:  
‚û°Ô∏è **Return a deep copy of the entire graph**, ensuring all nodes and connections are cloned correctly.

---

## üß† Key Idea
This is a **graph traversal + hashmap** problem.

Why?
- The graph may contain **cycles**, so you must avoid infinite loops.
- You must maintain a **mapping from original ‚Üí cloned node** to reuse clones when revisiting nodes.

---

## üöÄ Approach (DFS or BFS)
### Steps:
1. Use a dictionary `old_to_new` to store clones:

original_node ‚Üí cloned_node

2. Traverse the graph (DFS/BFS).
3. For each node:
- If it's not cloned yet ‚Üí create a clone and store it in the map.
- Recursively/iteratively clone all its neighbors.
4. Return the clone of the starting node.

Both DFS and BFS work equally well.

---

## üß© Time & Space Complexity
- **Time:** `O(V + E)`  
Each node and edge is processed once.
- **Space:** `O(V)`  
Hashmap + recursion/queue.

---

## üìù Sample DFS Code (Python)
```python
class Solution:
 def cloneGraph(self, node):
     if not node:
         return None
     
     old_to_new = {}
     
     def dfs(curr):
         if curr in old_to_new:
             return old_to_new[curr]
         
         clone = Node(curr.val)
         old_to_new[curr] = clone
         
         for nei in curr.neighbors:
             clone.neighbors.append(dfs(nei))
         
         return clone
     
     return dfs(node)


# My Solution

````python

# Definition for a Node.
class Node(object):
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []


class Solution(object):
    def cloneGraph(self, node):
        """
        :type node: Node
        :rtype: Node
        """
        visited = {}
        def dfs(curr):
            if not curr: return None
            clone = Node(val=curr.val)
            visited[curr] = clone
            for neighbor in curr.neighbors:
                if neighbor in visited:
                    clone.neighbors.append(visited[neighbor])
                else:
                    clone_neighbor = dfs(neighbor)
                    clone.neighbors.append(clone_neighbor)
            return clone
        return dfs(node)
````
