Given the head of a singly linked list, determine if the list contains a **cycle**.

A cycle exists if, by following `next` pointers repeatedly, you eventually revisit a node.

You must return:
- `true` â†’ if a cycle exists  
- `false` â†’ otherwise  

This is a fundamental linked list problem frequently asked in interviews.

## ğŸ§  Key Idea
Use **Floydâ€™s Tortoise and Hare Algorithm** (Two-Pointer Technique)

### Why it works:
- A slow pointer moves **1 step at a time**.
- A fast pointer moves **2 steps at a time**.
- If there is a cycle, the fast pointer will eventually â€œlapâ€ the slow pointer â†’ they meet.
- If fast reaches `None`, no cycle exists.

This avoids using extra memory (like a set), giving **O(1) space**.

---

## ğŸ”‘ Approach (O(n) Time, O(1) Space)
1. Initialize two pointers:
   - `slow = head`
   - `fast = head`
2. Loop while `fast` and `fast.next` exist:
   - move `slow = slow.next`
   - move `fast = fast.next.next`
   - if `slow == fast`: cycle detected â†’ return `True`
3. If loop ends, return `False` (no cycle)

- fast: 3 â†’ 0 â†’ 2 â†’ 2 â†’ ...

class Solution(object):
    def hasCycle(self, head):

        while fast and fast.next:
            slow = slow.next

````python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """
        if head == None or head.next == None: return False
        fast = head
        slow = head
        while slow.next and fast.next and fast.next.next:
            fast = fast.next.next
            slow = slow.next
            if fast == slow: return True
        return False
        
````
