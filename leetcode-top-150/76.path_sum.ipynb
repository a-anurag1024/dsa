## üìå Problem Summary  
Given a binary tree and an integer `targetSum`, determine whether there exists a **root-to-leaf** path such that the sum of all node values along the path equals `targetSum`.

A **leaf** is a node with *no children*.

---

## üîç Examples  

### Example 1  
**Input:**  
`root = [5,4,8,11,null,13,4,7,2,null,null,null,1]`, `targetSum = 22`  
**Output:** `true`  
Because path `5 ‚Üí 4 ‚Üí 11 ‚Üí 2` sums to 22.

### Example 2  
**Input:** `root = [1,2,3]`, `targetSum = 5`  
**Output:** `false`

### Example 3  
**Input:** `root = []`, `targetSum = 0`  
**Output:** `false`

---

## üöÄ Key Idea  
Use DFS to check *whether any root-to-leaf path* equals the target.

At each node:
- Reduce `targetSum` by the node's value.
- If we reach a leaf and `targetSum == 0` ‚Üí return True.
- Recursively check left and right children.

---

## üìù Short Approach (Interview-Ready)

1. If `root` is `None`, return `False`.  
2. Compute `new_target = targetSum - root.val`.  
3. If leaf node and `new_target == 0` ‚Üí return `True`.  
4. Recursively check:
   - left subtree  
   - right subtree  
5. Return OR of both results.

Time Complexity: **O(n)**  
Space Complexity: **O(h)** recursion (h = height of tree)

---

## üßë‚Äçüíª Template Code (Recursive DFS)

```python
class Solution:
    def hasPathSum(self, root, targetSum):
        if not root:
            return False
        
        # check leaf
        if not root.left and not root.right:
            return targetSum == root.val
        
        new_target = targetSum - root.val
        
        return (self.hasPathSum(root.left, new_target) or
                self.hasPathSum(root.right, new_target))


# My solution

````python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
class Solution(object):
    def hasPathSum(self, root, targetSum):
        """
        :type root: Optional[TreeNode]
        :type targetSum: int
        :rtype: bool
        """
        def helper(curr, curr_sum):
            if not curr: return False
            if not curr.left and not curr.right:
                if curr_sum + curr.val == targetSum: return True
                else: return False
            if_path_sum = False
            curr_sum += curr.val
            if curr.left:
                if_path_sum = if_path_sum or helper(curr.left, curr_sum)
            if curr.right:
                if_path_sum = if_path_sum or helper(curr.right, curr_sum)
            return if_path_sum
        return helper(root, 0)
````
