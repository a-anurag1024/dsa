## ðŸ”¹ Quick Review: LeetCode 17 â€” Letter Combinations of a Phone Number

### **Problem Summary**

Given a string `digits` containing numbers from **2â€“9**, return **all possible letter combinations** that the number could represent using the standard phone keypad mapping.

If `digits` is empty, return an empty list.

---

### **Key Observations**

* Each digit maps to **multiple letters** (2â€“4).
* Total combinations = **cartesian product** of mapped letters.
* Order does **not** matter.
* This is a **classic backtracking / DFS** problem.

---

### **Why Backtracking?**

* At each digit, you must **try all possible letters**.
* Build combinations **character by character**.
* Backtrack once a branch is completed.

---

## ðŸ”¹ Short Approach Sketch

1. Create a mapping of digits â†’ letters.
2. Use **DFS / backtracking**:

   * At index `i`, try all letters mapped from `digits[i]`.
   * Append a letter, recurse to the next index.
   * When `i == len(digits)`, store the formed combination.
3. Return all collected combinations.
4. Edge case: if `digits` is empty â†’ return `[]`.

---

### **Time & Space Complexity**

* **Time:** `O(4^n)` (max 4 letters per digit)
* **Space:** `O(n)` recursion depth (excluding output)

---

## ðŸ”¹ Python Solution (Backtracking)

```python
class Solution(object):
    def letterCombinations(self, digits):
        """
        :type digits: str
        :rtype: List[str]
        """
        if not digits:
            return []

        phone = {
            "2": "abc",
            "3": "def",
            "4": "ghi",
            "5": "jkl",
            "6": "mno",
            "7": "pqrs",
            "8": "tuv",
            "9": "wxyz"
        }

        res = []

        def backtrack(index, path):
            # Base case: combination complete
            if index == len(digits):
                res.append(path)
                return

            # Explore all letters for current digit
            for ch in phone[digits[index]]:
                backtrack(index + 1, path + ch)

        backtrack(0, "")
        return res
```



**why not brute force**:

> *Weâ€™re effectively generating combinations but pruning early using DFS rather than materializing all states at once.*




# My Solution

````python
class Solution(object):
    def letterCombinations(self, digits):
        """
        :type digits: str
        :rtype: List[str]
        """
        d2l = {'2': ['a', 'b', 'c'], '3': ['d', 'e', 'f'],
        '4': ['g', 'h', 'i'], '5': ['j', 'k', 'l'], '6': ['m', 'n', 'o'],
        '7': ['p', 'q', 'r', 's'], '8': ['t', 'u', 'v'], '9': ['w', 'x', 'y', 'z']}
        response = []
        n = len(digits)
        def dfs(i, curr):
            if i >= n: 
                response.append(curr)
                return
            j = i + 1
            for l in d2l[digits[i]]:
                nxt = curr + l
                dfs(j, nxt)
            return
        dfs(0, "")
        return response
````
