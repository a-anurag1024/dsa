## üß© Problem Summary

Given a partially filled 9√ó9 Sudoku board, determine if it is **valid**.  
Only the filled cells need to be validated according to the following rules:

1. Each **row** must contain digits 1‚Äì9 **without repetition**.  
2. Each **column** must contain digits 1‚Äì9 **without repetition**.  
3. Each **3√ó3 sub-box** must contain digits 1‚Äì9 **without repetition**.

Note:
- Empty cells are denoted by `"."`.
- The board does **not** need to be solvable ‚Äî only **valid**.

---


## ‚öôÔ∏è Approach (Set-based Validation)

### üîπ Key Idea:
Use **hash sets** to track occurrences of numbers in:
- Each **row**
- Each **column**
- Each **3√ó3 sub-box**

If any digit appears twice in the same row, column, or sub-box ‚Üí the board is invalid.

---

### üîπ Step-by-Step Logic

1. Initialize three collections:
   - `rows = [set() for _ in range(9)]`
   - `cols = [set() for _ in range(9)]`
   - `boxes = [set() for _ in range(9)]`
2. For each cell `(r, c)`:
   - Skip if the value is `"."`
   - Compute the box index: `box_id = (r // 3) * 3 + (c // 3)`
   - Check if value already exists in:
     - `rows[r]`
     - `cols[c]`
     - `boxes[box_id]`
     ‚Üí If yes, return `False`
   - Otherwise, add the value to all three sets.
3. If no conflicts found, return `True`.

## üßë‚Äçüíª Python Sketch

```python
class Solution(object):
    def isValidSudoku(self, board):
        rows = [set() for _ in range(9)]
        cols = [set() for _ in range(9)]
        boxes = [set() for _ in range(9)]

        for r in range(9):
            for c in range(9):
                val = board[r][c]
                if val == ".":
                    continue

                box_id = (r // 3) * 3 + (c // 3)

                if (val in rows[r]) or (val in cols[c]) or (val in boxes[box_id]):
                    return False

                rows[r].add(val)
                cols[c].add(val)
                boxes[box_id].add(val)

        return True

````python
class Solution(object):
    def isValidSudoku(self, board):
        """
        :type board: List[List[str]]
        :rtype: bool
        """
        # Row checks
        for row in board:
            hm = {}
            for j in range(9):
                if row[j] == ".": continue
                elif row[j] in hm: 
                    return False
                else: hm[row[j]] = 1
        # Col checks
        for i in range(9):
            hm = {}
            for row in board:
                if row[i] == ".": continue
                elif row[i] in hm: 
                    return False
                else: hm[row[i]] = 1
        # box checks
        for b in range(9):
            hm = {}
            for b_i in range(9):
                i, j = ((b // 3)*3 + b_i // 3), (b % 3)*3 + (b_i % 3)
                if board[i][j] == ".": continue
                elif board[i][j] in hm: 
                    return False
                else: hm[board[i][j]] = 1
        return True
````
