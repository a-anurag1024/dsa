## üìù Problem Summary
There are `n` gas stations arranged in a circle.  
- `gas[i]` = amount of gas at station `i`  
- `cost[i]` = gas required to travel from station `i` to `(i + 1) % n`

Find the **starting station index** from which you can complete the entire circuit once in the clockwise direction.  
If it‚Äôs not possible, return **-1**.  
It‚Äôs guaranteed that if a solution exists, it‚Äôs **unique**.

---

## üí° Example

**Input:**  
`gas = [1,2,3,4,5]`, `cost = [3,4,5,1,2]`  
**Output:** `3`  

**Explanation:**  
- Start at station 3 (gas = 4).  
- You can complete the full circle as total gas ‚â• total cost.  
- If you start anywhere else, you‚Äôll run out of gas before completing the loop.

---

## ‚öôÔ∏è Approach (Greedy ‚Äî Cumulative Balance)
### Idea:
- A valid start exists **only if** total gas ‚â• total cost.
- We traverse once, keeping track of:
  - `total_tank` ‚Üí net gas gained/lost overall.  
  - `curr_tank` ‚Üí net gas gained/lost on the current route.  
- If `curr_tank` becomes negative at station `i`, it means we can‚Äôt start from or before `i`, so we:
  - Reset the start to `i + 1`,  
  - Reset `curr_tank = 0`.

The greedy insight:  
> ‚ÄúThe point right after a failure point is the only possible valid start.‚Äù

---

## ‚úÖ Code Snippet (Python)
```python
def canCompleteCircuit(gas, cost):
    total_tank = curr_tank = start = 0
    for i in range(len(gas)):
        total_tank += gas[i] - cost[i]
        curr_tank += gas[i] - cost[i]

        # If current route fails, move start forward
        if curr_tank < 0:
            start = i + 1
            curr_tank = 0

    return start if total_tank >= 0 else -1

````python
class Solution(object):
    def canCompleteCircuit(self, gas, cost):
        """
        :type gas: List[int]
        :type cost: List[int]
        :rtype: int
        """
        start = 0
        total_gas = 0
        curr_gas = 0
        for i in range(len(gas)):
            total_gas += gas[i] - cost[i]
            curr_gas += gas[i] - cost[i]
            if curr_gas < 0:
                start = i+1
                curr_gas = 0
        return start if total_gas >= 0 else -1
````
