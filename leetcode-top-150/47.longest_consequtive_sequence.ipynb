## ðŸ“Œ Problem Summary
Given an **unsorted** array `nums`, return the **length of the longest consecutive elements sequence**.

A consecutive sequence means numbers follow each other (e.g., `1,2,3,4`).  
**The solution must be O(n)** â€” sorting is *not allowed* (would be `O(n log n)`).

---

# ðŸ§  Key Idea  
Use a **hash set** for O(1) lookups and expand sequences only from **sequence starting points**.

A number is the **start of a sequence** if:
x - 1 is NOT in the set


Then we grow forward:
x, x+1, x+2, â€¦


This ensures each sequence is counted exactly once â†’ O(n) time.

---

# âœ… Short Approach (Hash Set + Expand Only Starts)

### Steps:
1. Insert all numbers into a hash set `nums_set`.
2. For each number `x`:
   - Only begin counting if `x - 1` is not in the set  
     (meaning `x` is the start of a sequence)
   - Expand forward: check `x+1, x+2, â€¦`
3. Track the longest length.

### Complexity:
- **Time:** `O(n)`  
  Each number is processed at most twice.  
- **Space:** `O(n)` for the hash set.

---

# ðŸ”§ Python Code (O(n) Solution)

```python
class Solution:
    def longestConsecutive(self, nums):
        nums_set = set(nums)
        longest = 0
        
        for x in nums_set:
            # Only start counting if x is the start of a sequence
            if x - 1 not in nums_set:
                length = 1
                curr = x
                
                while curr + 1 in nums_set:
                    curr += 1
                    length += 1
                
                longest = max(longest, length)
        
        return longest

````python
class Solution(object):
    def longestConsecutive(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        hs = set(nums)
        longest = 0
        for x in hs:
            if x - 1 not in hs:
                c = 1
                j = 1
                while True:
                    if x + j in hs:
                        c += 1
                        j += 1
                    else: break
                if c > longest: longest = c
        return longest
````
