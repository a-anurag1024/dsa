## ðŸ“ Problem Summary
You are given an array `prices` where `prices[i]` represents the stock price on day `i`.  
You can complete **multiple buy/sell transactions**, but can only hold **one stock at a time**.

Return the **maximum profit** you can achieve.

---

## ðŸ’¡ Example
**Input:**  
`prices = [7,1,5,3,6,4]`  
**Output:**  
`7`  
**Explanation:**  
- Buy at 1, sell at 5 â†’ profit = 4  
- Buy at 3, sell at 6 â†’ profit = 3  
- Total profit = 4 + 3 = **7**

---

## âš™ï¸ Approach (Greedy)
### Intuition:
Every time the price goes **up** from one day to the next, we can gain profit by â€œbuying beforeâ€ and â€œselling afterâ€.

### Steps:
1. Initialize `profit = 0`.
2. Iterate through prices:
   - Whenever `prices[i] > prices[i-1]`, add `(prices[i] - prices[i-1])` to profit.
3. Return the total accumulated `profit`.

This works because taking every increasing pair is equivalent to performing all profitable transactions.

---

## â±ï¸ Complexity
- **Time:** O(n) â€” single pass through the prices.  
- **Space:** O(1) â€” only one variable to track profit.

def maxProfit(prices):
    profit = 0
    for i in range(1, len(prices)):
        if prices[i] > prices[i - 1]:
            profit += prices[i] - prices[i - 1]
    return profit

````python
# Inefficient DP solution O(n^2) time and O(n) space
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        n = len(prices)
        max_profit_buy_that_day = [0]*n
        for j in range(n-1, -1, -1):
            if j == n-1:
                max_profit_buy_that_day[j] = 0
            else:
                curr_max = max_profit_buy_that_day[j+1]
                for i in range(j+1, n):
                    if prices[i] > prices[j]:
                        curr = prices[i] - prices[j]
                    else: curr = 0
                    if i != n-1: 
                        curr += max_profit_buy_that_day[i+1]
                    if curr > curr_max:
                        curr_max = curr
                max_profit_buy_that_day[j] = curr_max
        return max_profit_buy_that_day[0]
````

````python
# Efficient Greedy solution O(n) time and O(1) space
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        buy_price = prices[0]
        sell_price = prices[0]
        total_profit = 0
        for curr_price in prices[1:]:
            if curr_price > sell_price:
                sell_price = curr_price
            else:
                total_profit += max(0, sell_price-buy_price)
                buy_price = curr_price
                sell_price = curr_price
        total_profit += max(0, sell_price-buy_price)
        return total_profit
````
