## üìå Problem Summary  
Given the root of a binary tree, compute the **average value of nodes at each level** and return them as a list of floats.

Example:  
Level 0 ‚Üí average = 3  
Level 1 ‚Üí average = 14.5  
Level 2 ‚Üí average = 11  

Must return values within `1e-5` accuracy.

---

## üîç Example  
**Input:** `[3,9,20,null,null,15,7]`  
**Output:** `[3.00000, 14.50000, 11.00000]`

---

## üöÄ Key Idea  
Perform a **level-order BFS**:

For each level:
1. Compute the sum of all node values  
2. Count number of nodes at that level  
3. Append `sum / count` to result list  

Why BFS?  
Because each iteration corresponds naturally to one level of the tree.

---

## üìù Short Approach (Interview-Ready)

### BFS Steps:
1. Initialize a queue with root.  
2. While queue not empty:  
   - Determine the number of nodes in current level  
   - Pop that many nodes  
   - Accumulate their values  
   - Push their children back into the queue  
3. Compute average and append to list  

Time Complexity: **O(n)**  
Space Complexity: **O(n)** in worst case (last level of tree)

---

## üßë‚Äçüíª Template Code (BFS)

```python
from collections import deque

class Solution:
    def averageOfLevels(self, root):
        if not root:
            return []

        q = deque([root])
        result = []

        while q:
            level_size = len(q)
            level_sum = 0

            for _ in range(level_size):
                node = q.popleft()
                level_sum += node.val

                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)

            result.append(level_sum / level_size)

        return result
