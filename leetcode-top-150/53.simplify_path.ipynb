## ðŸ“˜ Problem Summary
You are given an **absolute Unix-style path** starting with `'/'`.  
Your task is to convert it into a **canonical simplified path** by following Unix rules:

### Meaning of symbols:
- `.` â†’ current directory (ignore it)
- `..` â†’ parent directory (pop last directory from stack)
- `//` â†’ multiple slashes treated as a single slash
- Everything else (e.g., `...`, `....`, `file`) is a **valid directory name**

### Canonical Path Rules:
- Must start with **one leading slash** (`/`)
- No trailing slash unless root (`/`)
- Directories must be separated by exactly one `/`
- No `.` or `..` in final path

---

## ðŸ§  Key Idea
Use a **stack** to simulate navigation through directories.

Split the path on `/` and process each component:
- If part is `""` or `"."` â†’ skip
- If part is `".."` â†’ pop from stack if possible
- Else â†’ push valid directory name onto stack

At the end, join the stack with `'/'` and prepend `'/'`.

---

## ðŸ”‘ Approach (O(n) Time, O(n) Space)
1. Split the path by `'/'`.
2. Initialize an empty stack.
3. For each component:
   - Skip `''` or `'.'`.
   - If `'..'`: pop if stack not empty.
   - Else: push valid directory name.
4. Construct the canonical path:  
   `"/" + "/".join(stack)`  
5. If stack is empty â†’ return `"/"`.

---

## ðŸ“ Examples
| Input | Output | Explanation |
|-------|--------|-------------|
| `"/home/"` | `"/home"` | remove trailing slash |
| `"/home//foo/"` | `"/home/foo"` | collapse `//` |
| `"/home/user/Documents/../Pictures"` | `"/home/user/Pictures"` | go up one directory |
| `"/../"` | `"/"` | cannot go above root |
| `"/.../a/../b/c/../d/./"` | `"/.../b/d"` | `...` is a valid directory |

## ðŸ§© Sample Python Implementation
```python
class Solution(object):
    def simplifyPath(self, path):
        stack = []

        for part in path.split('/'):
            if part == '' or part == '.':
                continue
            elif part == '..':
                if stack:
                    stack.pop()
            else:
                stack.append(part)

        return "/" + "/".join(stack)
