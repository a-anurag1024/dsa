### ğŸ’¡ Key Insight (Very Important)

This is a **classic dynamic programming** problem and maps **directly to the Fibonacci sequence**.

Why?

* To reach step `n`:

  * You could come from step `n-1` (1 step)
  * Or from step `n-2` (2 steps)

So:

```
ways(n) = ways(n-1) + ways(n-2)
```

3. Iteratively compute up to `n`
4. Return the final value

### ğŸ§® Python Solution (Optimized DP)

```python
class Solution(object):
    def climbStairs(self, n):
        if n <= 2:
            return n
        
        prev2, prev1 = 1, 2
        for _ in range(3, n + 1):
            curr = prev1 + prev2
            prev2, prev1 = prev1, curr
        
        return prev1
```

# My Solution

````python
class Solution(object):
    def climbStairs(self, n):
        """
        :type n: int
        :rtype: int
        """
        last_two = [2, 1]
        if n == 0: return 0
        if n <= 2: return last_two[-n]
        for j in range(n-2):
            new = last_two[0] + last_two[1]
            last_two = [new, last_two[0]]
        return last_two[0]
````
