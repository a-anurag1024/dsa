### **Problem Summary**

Given a **sorted array of distinct integers** and a `target`, return:

* the **index** if `target` exists, or
* the **index where it should be inserted** to maintain sorted order.

**Constraint:** Must run in `O(log n)` time.

---

### **Key Observations**

* Sorted array â†’ **Binary Search**
* Even if `target` is not found, binary search naturally converges to the **insertion position**
* Distinct elements remove ambiguity

---

## ðŸ”¹ Short Approach Sketch (Binary Search)

1. Initialize pointers:

   * `left = 0`, `right = len(nums) - 1`
2. While `left <= right`:

   * Compute `mid`
   * If `nums[mid] == target` â†’ return `mid`
   * If `nums[mid] < target` â†’ search right half
   * Else â†’ search left half
3. When loop ends, `left` points to the **correct insertion index**
4. Return `left`

---

### **Time & Space Complexity**

* **Time:** `O(log n)`
* **Space:** `O(1)`

---

## ðŸ”¹ Python Solution (Optimal Binary Search)

```python
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        left, right = 0, len(nums) - 1

        while left <= right:
            mid = (left + right) // 2

            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1

        # left is the insertion position
        return left
```

---

## ðŸ”¹ Why This Works

* Binary search narrows down where `target` *would* exist
* When the loop ends, `left` is the **smallest index** such that:

  ```
  nums[left] >= target
  ```
* This satisfies insertion order



# My Solution

````python
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        n = len(nums)
        left = 0
        right = n - 1
        while left <= right:
            mid = (left + right) // 2
            if target == nums[mid]: return mid
            elif target < nums[mid]: right = mid - 1
            else: left = mid + 1
        return left
````
