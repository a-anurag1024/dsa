# üéì LeetCode 207: Course Schedule ‚Äî Quick Review

## üìò Problem Summary
You are given:
- `numCourses` labeled from `0` to `numCourses - 1`
- A list of prerequisites where  
  `prerequisites[i] = [a, b]` means **course `b` must be taken before course `a`**

Your task:
‚û°Ô∏è **Return `True` if all courses can be finished**, otherwise return `False`.

---

## üß† Key Insight
This is a **cycle detection problem in a directed graph**.

- Each course ‚Üí **node**
- Each prerequisite `[a, b]` ‚Üí **directed edge** `b ‚Üí a`

If the graph contains a **cycle**, it is impossible to finish all courses.

---

## üöÄ Approach 1: DFS + Cycle Detection (Most Common)

### Idea:
- Use DFS to detect cycles.
- Maintain:
  - `visiting` set ‚Üí nodes in current DFS path
  - `visited` set ‚Üí nodes already processed (safe)

### Steps:
1. Build adjacency list.
2. For each course:
   - Run DFS if not visited.
3. If during DFS you revisit a node in `visiting`, a cycle exists ‚Üí return `False`.
4. If DFS finishes without cycles ‚Üí return `True`.

---

## üöÄ Approach 2: Kahn‚Äôs Algorithm (Topological Sort)

### Idea:
- Courses with **indegree 0** can be taken first.
- Remove them and reduce indegree of dependent courses.
- If you can process all courses ‚Üí no cycle.

---

## üß© Time & Space Complexity
- **Time:** `O(V + E)`  
- **Space:** `O(V + E)`

Where:
- `V = numCourses`
- `E = number of prerequisites`

---

## üìù Sample DFS Code (Python)
```python
from collections import defaultdict

class Solution:
    def canFinish(self, numCourses, prerequisites):
        graph = defaultdict(list)
        for a, b in prerequisites:
            graph[b].append(a)

        visiting = set()
        visited = set()

        def dfs(course):
            if course in visiting:
                return False
            if course in visited:
                return True

            visiting.add(course)
            for nxt in graph[course]:
                if not dfs(nxt):
                    return False
            visiting.remove(course)
            visited.add(course)
            return True

        for c in range(numCourses):
            if not dfs(c):
                return False

        return True


# My Solution

````python
class Solution(object):
    def canFinish(self, numCourses, prerequisites):
        """
        :type numCourses: int
        :type prerequisites: List[List[int]]
        :rtype: bool
        """
        graph = {}
        doable_check = {}
        for pre in prerequisites:
            if pre[0] not in graph: graph[pre[0]] = {pre[1]}
            else: graph[pre[0]].add(pre[1])
            if pre[1] not in graph: graph[pre[1]] = set()
        def check_doable(curr, visited):
            if curr in doable_check: return doable_check[curr]
            if curr in visited: 
                doable_check[curr] = False
                return False
            visited.add(curr)
            if curr not in graph: 
                doable_check[curr] = True
                return True
            doable = True
            for pre in graph[curr]:
                doable = doable and check_doable(pre, visited)
            doable_check[curr] = doable
            return doable
        for course in graph:
            doable = check_doable(course, set())
            if not doable: return False
        return True
````
