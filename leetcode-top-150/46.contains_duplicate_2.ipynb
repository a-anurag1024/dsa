## ðŸ“Œ Problem Summary  
Given an array `nums` and an integer `k`, determine if there exist **two equal elements** such that their **indices differ by at most k**:

nums[i] == nums[j] and |i - j| â‰¤ k


---

# ðŸ§  Key Insight  
We need to check **recent occurrences** only.  
This is a classic **sliding window + hash set** problem.

---

# âœ… Short Approach  

### **Approach: Sliding Window with Hash Set (Optimal)**  
Maintain a sliding window of size at most `k` containing the last `k` elements.

Steps:
1. Iterate through the array with index `i`.
2. Before inserting `nums[i]`, check:
   - If `nums[i]` already exists in the window â†’ return `True`
3. Add `nums[i]` to the window.
4. If window size exceeds `k`, remove `nums[i - k]`.
5. If the loop finishes with no duplicate found â†’ return `False`

### Complexity
- **Time:** `O(n)`  
- **Space:** `O(k)`

This is optimal because the window never holds more than `k` elements.

---

# ðŸ”§ Python Code (Sliding Window)

```python
class Solution:
    def containsNearbyDuplicate(self, nums, k):
        window = set()
        
        for i, num in enumerate(nums):
            if num in window:
                return True
            
            window.add(num)
            
            if len(window) > k:
                window.remove(nums[i - k])
        
        return False

````python
# The counting is not required in this case, as we only need to know if there is a duplicate within k distance.

from collections import Counter

class Solution(object):
    def containsNearbyDuplicate(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: bool
        """
        nb = Counter()
        n = len(nums)
        for i in range(min(n, k+1)):
            nb[nums[i]] += 1
        for i in range(n):
            if nb[nums[i]] >= 2:
                return True
            if i + k + 1 < n:
                nb[nums[i + k + 1]] += 1
            if i - k >= 0:
                nb[nums[i - k]] -= 1
        return False
````
