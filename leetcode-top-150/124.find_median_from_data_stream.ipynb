## ðŸ” Problem Recap

You need to design a data structure that:

* Continuously receives numbers from a data stream
* Can return the **median at any time**

**Median definition**

* Odd count â†’ middle element
* Even count â†’ average of the two middle elements

Operations to support:

* `addNum(num)`
* `findMedian()`

---

## ðŸ§  Key Insight

To get the median efficiently:

* You must **keep numbers ordered**
* Sorting every time is too slow

ðŸ‘‰ The optimal approach is to **split the numbers into two halves**:

* Left half â†’ smaller numbers
* Right half â†’ larger numbers

And always keep them **balanced**.

---

## ðŸš€ Optimal Strategy: Two Heaps

### Data Structures

1. **Max-heap (`left`)**
   Stores the smaller half (largest element at top)
2. **Min-heap (`right`)**
   Stores the larger half (smallest element at top)

Python only has min-heaps, so:

* Use **negative values** to simulate a max-heap

---

## âœï¸ Short Approach Sketch

1. Insert number into max-heap (`left`)
2. Move the largest from `left` to `right`
3. Balance sizes so `left` has **equal or one extra element**
4. Median:

   * Odd â†’ top of `left`
   * Even â†’ average of tops of `left` and `right`

---

## â±ï¸ Complexity

* **addNum** â†’ `O(log n)`
* **findMedian** â†’ `O(1)`
* **Space** â†’ `O(n)`

This is optimal and expected in interviews.

## ðŸ§© Python Solution (Optimal)

```python
import heapq

class MedianFinder(object):

    def __init__(self):
        self.left = []   # max-heap (store negatives)
        self.right = []  # min-heap

    def addNum(self, num):
        # Step 1: add to max-heap
        heapq.heappush(self.left, -num)
        
        # Step 2: move largest from left to right
        heapq.heappush(self.right, -heapq.heappop(self.left))
        
        # Step 3: balance heaps
        if len(self.right) > len(self.left):
            heapq.heappush(self.left, -heapq.heappop(self.right))

    def findMedian(self):
        if len(self.left) > len(self.right):
            return float(-self.left[0])
        return (-self.left[0] + self.right[0]) / 2.0
```

# My Solution

````python
import heapq

class MedianFinder(object):

    def __init__(self):
        self.n = 0
        self.left = []
        self.right = []


    def addNum(self, num):
        """
        :type num: int
        :rtype: None
        """
        if self.n % 2 == 1:
            if num <= - self.left[0]:
                mid_left = heapq.heappop(self.left)
                heapq.heappush(self.right, -mid_left)
                heapq.heappush(self.left, -num)
            else:
                heapq.heappush(self.right, num)
        else:
            if not self.left or num <= -self.left[0] or num <= self.right[0]:
                heapq.heappush(self.left, -num)
            else:
                mid_right = heapq.heappop(self.right)
                heapq.heappush(self.left, -mid_right)
                heapq.heappush(self.right, num)
        self.n += 1
        

    def findMedian(self):
        """
        :rtype: float
        """
        if self.n % 2 == 1:
            return - self.left[0]
        else:
            return (self.right[0] - self.left[0]) / 2.0
        


# Your MedianFinder object will be instantiated and called as such:
# obj = MedianFinder()
# obj.addNum(num)
# param_2 = obj.findMedian()
````
