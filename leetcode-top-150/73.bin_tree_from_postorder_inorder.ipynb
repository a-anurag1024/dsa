## ðŸš€ Key Idea  
### **Postorder gives the root last.**  
`postorder[-1]` is always the **root** of the current subtree.

### **Inorder tells how the subtree splits.**  
If root = `x` appears at index `mid` in inorder:

inorder: [ left-subtree | x | right-subtree ]

sizes: L nodes 1 R nodes


### Construction strategy (reverse of preorder solution):
1. Pop root from the **end** of postorder  
2. Find its index in inorder  
3. Build **right subtree first**, then left subtree  
   (because postorder is left-right-root, so when we go backwards it's root-right-left)

Use a hashmap for O(1) lookup of inorder positions.

---

## ðŸ“ Short Approach (Interview-Ready)

1. Build a map `{value â†’ inorder_index}`.  
2. Maintain a pointer `post_idx = len(postorder) - 1`.  
3. Define `build(l, r)` to construct subtree from inorder[l:r].  
4. In recursion:
   - If `l > r`: return None  
   - Root = postorder[post_idx], decrement post_idx  
   - Split inorder at root index `mid`  
   - **Build right subtree first**  
   - Then build left subtree  
5. Return root.

Time Complexity: **O(n)**  
Space Complexity: **O(n)**

## ðŸ§‘â€ðŸ’» Template Code (Optimal DFS)

```python
class Solution:
    def buildTree(self, inorder, postorder):
        idx_map = {val: i for i in inorder}
        post_idx = len(postorder) - 1
        
        def build(left, right):
            nonlocal post_idx
            if left > right:
                return None
            
            # Step 1: take root from postorder (end)
            root_val = postorder[post_idx]
            post_idx -= 1
            root = TreeNode(root_val)
            
            # Step 2: split inorder
            mid = idx_map[root_val]

            # Step 3: build right subtree FIRST, then left
            root.right = build(mid + 1, right)
            root.left = build(left, mid - 1)

            return root
        
        return build(0, len(inorder) - 1)


# My Solution

````python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
class Solution(object):
    def buildTree(self, inorder, postorder):
        """
        :type inorder: List[int]
        :type postorder: List[int]
        :rtype: Optional[TreeNode]
        """
        inorder_hm = {value:ind for ind, value in enumerate(inorder)}
        n = len(inorder)
        def helper(post_ind, in_l, in_r):
            if in_l > in_r: return None, post_ind
            curr_val = postorder[post_ind]
            curr_node = TreeNode(curr_val)
            post_ind -= 1
            mid = inorder_hm[curr_val]
            curr_node.right, post_ind = helper(post_ind, mid+1, in_r)
            curr_node.left, post_ind = helper(post_ind, in_l, mid-1)
            return curr_node, post_ind
        return helper(n-1, 0, n-1)[0]
        
````
