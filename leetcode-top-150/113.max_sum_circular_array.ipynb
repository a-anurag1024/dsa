### **Problem Summary**

Given a **circular array** `nums`, find the **maximum possible sum of a non-empty subarray**.

* Subarray must be **contiguous**
* Circular means it may **wrap around** from end â†’ beginning
* Each element can be used **at most once**

---

### **Key Observations**

* Two possible cases:

  1. **Normal (non-circular) subarray** â†’ standard Kadaneâ€™s Algorithm
  2. **Circular subarray** â†’ takes elements from end and beginning
* Circular max sum =
  **total_sum âˆ’ minimum subarray sum**
* Edge case:

  * If all numbers are negative, circular case becomes invalid

---

## ðŸ”¹ Short Approach Sketch

1. Use **Kadaneâ€™s Algorithm** to find:

   * `max_sub` â†’ maximum subarray sum
   * `min_sub` â†’ minimum subarray sum
2. Compute `total_sum` of array
3. If `max_sub < 0`:

   * Return `max_sub` (all elements negative)
4. Else:

   * Return `max(max_sub, total_sum âˆ’ min_sub)`

---

### **Time & Space Complexity**

* **Time:** `O(n)`
* **Space:** `O(1)`

## ðŸ”¹ Python Solution (Optimal)

```python
class Solution(object):
    def maxSubarraySumCircular(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        total_sum = 0
        curr_max = curr_min = 0
        max_sum = float('-inf')
        min_sum = float('inf')

        for num in nums:
            curr_max = max(num, curr_max + num)
            max_sum = max(max_sum, curr_max)

            curr_min = min(num, curr_min + num)
            min_sum = min(min_sum, curr_min)

            total_sum += num

        # If all numbers are negative
        if max_sum < 0:
            return max_sum

        return max(max_sum, total_sum - min_sum)
```

### ðŸ’¡ Tip

> *Why canâ€™t we always use `total_sum - min_sub`?*

Answer:

> *When all elements are negative, `min_sub == total_sum`, leading to an empty subarray, which is invalid.*

# My Solution

````python
class Solution(object):
    def maxSubarraySumCircular(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        max_sum = -float('inf')
        min_sum = float('inf')
        curr_sum = max_sum
        curr_min_sum = min_sum
        total_sum = 0
        for i in range(len(nums)):
            curr_sum = max(curr_sum+nums[i], nums[i])
            curr_min_sum = min(curr_min_sum+nums[i], nums[i])
            max_sum = max(max_sum, curr_sum)
            min_sum = min(min_sum, curr_min_sum)
            total_sum += nums[i]
        if min_sum == total_sum: return max_sum
        else: return max(max_sum, total_sum-min_sum)
````
