### **Problem Summary**

Given two integers `n` and `k`, return **all possible combinations** of `k` numbers chosen from the range `[1, n]`.

* Combinations are **unordered**
* Each number can be used **only once**
* Order inside a combination does **not** matter

---

### **Key Observations**

* This is **n choose k (C(n, k))**
* Permutations are **not allowed** (`[1,2] == [2,1]`)
* We must enforce **increasing order** to avoid duplicates
* Classic **backtracking / DFS with pruning** problem

---

## ðŸ”¹ Short Approach Sketch (Backtracking)

1. Use DFS with parameters:

   * `start`: next number to consider
   * `path`: current combination being built
2. If `len(path) == k`, add it to result.
3. Iterate from `start` to `n`:

   * Add current number to `path`
   * Recurse with `start = i + 1`
   * Backtrack
4. Optional optimization: prune when remaining numbers are insufficient.

---

### **Time & Space Complexity**

* **Time:** `O(C(n, k))`
* **Space:** `O(k)` recursion depth (excluding output)

## ðŸ”¹ Python Solution (Backtracking â€“ Optimal)

```python
class Solution(object):
    def combine(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: List[List[int]]
        """
        res = []

        def backtrack(start, path):
            # If combination is complete
            if len(path) == k:
                res.append(path[:])
                return

            # Try remaining numbers
            for i in range(start, n + 1):
                path.append(i)
                backtrack(i + 1, path)
                path.pop()

        backtrack(1, [])
        return res
```

# My Solution

````python
class Solution(object):
    def combine(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: List[List[int]]
        """
        combinations = []
        def dfs(i, curr):
            if len(curr) == k: 
                combinations.append(curr[:])
                return 
            for j in range(i+1, n+1):
                curr.append(j)
                dfs(j, curr)
                curr.pop()
        dfs(0, [])
        return combinations
````
