## ðŸš€ Key Idea  
Simulate in-order traversal **iteratively** using a stack.

### Important observation:  
In a BST, the next smallest element is always found by:

1. Going **all the way left**  
2. Processing the node  
3. Exploring its **right subtree**

So the iterator uses a stack to store the path to the next node.

---

## ðŸ“ Short Approach (Interview-Ready)

### Initialization:
- Push all left nodes from root onto the stack  
  â†’ the top of the stack is the smallest element.

### `next()`:
1. Pop the top node â†’ thatâ€™s the next smallest  
2. If that node has a right child:
   - Push its entire left chain onto the stack  
3. Return the popped value

### `hasNext()`:
- Return whether the stack is non-empty

### Complexity:
- Time per operation: **O(1) amortized** (each node pushed/popped once)  
- Space: **O(h)** stack (height of tree)

## ðŸ§‘â€ðŸ’» Template Code (Optimal)

```python
class BSTIterator:
    def __init__(self, root):
        self.stack = []
        self._push_left(root)

    def _push_left(self, node):
        while node:
            self.stack.append(node)
            node = node.left

    def next(self):
        top = self.stack.pop()
        if top.right:
            self._push_left(top.right)
        return top.val

    def hasNext(self):
        return len(self.stack) > 0
````


# My Solution

`````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class BSTIterator(object):

    def __init__(self, root):
        """
        :type root: Optional[TreeNode]
        """
        self.curr = root
        self.stack = []
        while self.curr.left:
            self.stack.append(self.curr)
            self.curr = self.curr.left

    def next(self):
        """
        :rtype: int
        """
        val = self.curr.val
        if self.curr.right:
            self.curr = self.curr.right
            while self.curr.left:
                self.stack.append(self.curr)
                self.curr = self.curr.left
        else:
            if self.stack:
                self.curr = self.stack.pop()
            else: self.curr = None
        return val


    def hasNext(self):
        """
        :rtype: bool
        """
        if self.curr: return True
        else: return False


# Your BSTIterator object will be instantiated and called as such:
# obj = BSTIterator(root)
# param_1 = obj.next()
# param_2 = obj.hasNext()
`````
