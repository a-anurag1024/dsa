### ü™ô LeetCode 322 ‚Äî Coin Change (Quick Review)

---

#### üîç Problem Recap

You are given:

* `coins`: different coin denominations (infinite supply of each)
* `amount`: total money to form

üëâ **Goal:** Find the **minimum number of coins** needed to make up `amount`.
If it‚Äôs **not possible**, return `-1`.

---

### üí° Core Insight

This is a **classic unbounded knapsack / DP optimization** problem.

At every amount `x`, you decide:

> *Which coin was used last?*

If you use coin `c`, then:

```
dp[x] = dp[x - c] + 1
```

You want the **minimum** over all valid coins.

---

### üß† DP State Definition

Let:

* `dp[i]` = minimum coins needed to make amount `i`

---

### üß± Base Case

```
dp[0] = 0     # 0 coins needed to make amount 0
```

Initialize all others with a **sentinel value** (`‚àû` or `amount + 1`).

---

### üîÅ DP Transition

For each amount `i` from `1 ‚Üí amount`:

```
dp[i] = min(dp[i], dp[i - coin] + 1)   for all coin ‚â§ i
```

---

### ‚úçÔ∏è Short Approach Sketch

1. Initialize DP array of size `amount + 1`
2. Set `dp[0] = 0`, rest as `amount + 1`
3. For each amount:

   * Try all coins
   * Update minimum coins needed
4. If `dp[amount]` is still sentinel ‚Üí return `-1`
5. Else return `dp[amount]`

---

### üßÆ Python Solution (Bottom-Up DP)

```python
class Solution(object):
    def coinChange(self, coins, amount):
        dp = [amount + 1] * (amount + 1)
        dp[0] = 0
        
        for i in range(1, amount + 1):
            for coin in coins:
                if coin <= i:
                    dp[i] = min(dp[i], dp[i - coin] + 1)
        
        return dp[amount] if dp[amount] != amount + 1 else -1
```

---

### ‚è±Ô∏è Complexity Analysis

* **Time Complexity:** `O(amount √ó len(coins))`
* **Space Complexity:** `O(amount)`


# My Solution

````python
class Solution(object):
    def coinChange(self, coins, amount):
        """
        :type coins: List[int]
        :type amount: int
        :rtype: int
        """
        dp = [0] + [-1]*amount
        for i in range(1, len(dp)):
            min_change = float('inf')
            for c in coins:
                if i - c >= 0 and dp[i-c] != -1:
                    min_change = min(min_change, dp[i-c]+1)
            if min_change <= amount: dp[i] = min_change
        return dp[-1]
````
