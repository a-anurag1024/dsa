## ðŸ§  Problem Summary

You are given:
- A string `s`
- An array of words `words` (all of **equal length**)

Return **all starting indices** in `s` where a substring is formed by **concatenating all words exactly once** in **any order**.

---

### Example 1:
**Input:**  
`s = "barfoothefoobarman"`, `words = ["foo","bar"]`  
**Output:** `[0, 9]`  
**Explanation:**  
- `"barfoo"` at index `0` = concatenation of `["bar","foo"]`
- `"foobar"` at index `9` = concatenation of `["foo","bar"]`

---

### Example 2:
**Input:**  
`s = "wordgoodgoodgoodbestword"`, `words = ["word","good","best","word"]`  
**Output:** `[]`  
**Explanation:**  
No substring matches all words.

---

### Example 3:
**Input:**  
`s = "barfoofoobarthefoobarman"`, `words = ["bar","foo","the"]`  
**Output:** `[6, 9, 12]`  
**Explanation:**  
Substrings `"foobarthe"`, `"barthefoo"`, and `"thefoobar"` are valid concatenations.

---

## âš™ï¸ Approach (Sliding Window + Hash Map Frequency Check)

### ðŸ”¹ Key Idea:
We need to find substrings of total length = `total_len = len(words) * word_len`  
such that every **word appears exactly as many times** as in `words`.

We can efficiently check each possible starting point using:
- **Fixed window sliding** of size `total_len`
- **Frequency comparison** of words within that window

---

### ðŸ”¹ Step-by-Step Logic

1. Let:
   - `word_len = len(words[0])`
   - `num_words = len(words)`
   - `total_len = word_len * num_words`
2. Build a frequency map:  
   `word_count = Counter(words)`
3. Iterate possible start offsets from `0` to `word_len - 1`:
   - Maintain a sliding window and a running count of words.
   - Move the window in steps of `word_len`.
   - If a word is invalid or exceeds expected count, slide the left boundary.
4. Record all valid start indices where counts match `word_count`.

---

## ðŸ§‘â€ðŸ’» Python Sketch

```python
from collections import Counter

def findSubstring(s, words):
    if not s or not words:
        return []

    word_len = len(words[0])
    total_len = len(words) * word_len
    word_count = Counter(words)
    res = []

    for i in range(word_len):  # handle different alignment offsets
        left = i
        curr_count = Counter()
        for j in range(i, len(s) - word_len + 1, word_len):
            word = s[j:j + word_len]
            if word in word_count:
                curr_count[word] += 1
                while curr_count[word] > word_count[word]:
                    curr_count[s[left:left + word_len]] -= 1
                    left += word_len
                if j + word_len - left == total_len:
                    res.append(left)
            else:
                curr_count.clear()
                left = j + word_len
    return res
