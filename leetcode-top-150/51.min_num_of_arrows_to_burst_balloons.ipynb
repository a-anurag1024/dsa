## ðŸ“˜ Problem Summary
You're given a list of balloon intervals `[xstart, xend]` representing the horizontal range of each balloon on the x-axis.

You can shoot an arrow **vertically upward** at any x position.  
It will burst **all balloons whose range covers that x-coordinate**.

Your goal: **shoot the minimum number of arrows** to burst all balloons.

---

## ðŸ§  Key Idea
This is a classic **interval scheduling / greedy overlap** problem.

### âž¤ Insight  
If balloons overlap, **one arrow can burst all overlapping balloons**.  
So we want to group overlapping intervals and use one arrow per group.

### ðŸ”‘ Greedy Strategy  
**Sort balloons by their end coordinate** (`xend`).  
Then:
- Shoot an arrow at the end of the first balloon.
- For each next balloon:
  - If it overlaps (i.e., `start <= current_arrow_position`), it gets burst automatically.
  - Otherwise, shoot a new arrow and set the new arrow position to this balloonâ€™s end.

This ensures **minimum arrows** (same as "minimum number of intervals to cover points").

---

## âœï¸ Approach (O(n log n) Sorting + O(n) Scan)
1. Sort `points` by `xend` (ascending).
2. Initialize:
   - `arrows = 1`
   - `last_arrow_pos = points[0][1]`
3. Iterate remaining balloons:
   - If `xstart <= last_arrow_pos`: it overlaps â†’ continue.
   - Else: shoot a new arrow â†’ `arrows += 1`, update `last_arrow_pos`.
4. Return `arrows`.

---

## ðŸ“ Example
Input:  
`[[10,16],[2,8],[1,6],[7,12]]`

Sorted by end â†’  
`[[1,6],[2,8],[7,12],[10,16]]`

Arrows:
- Arrow at `6` bursts first two.
- Next balloon starts at `7` â†’ need new arrow at `12`.

Total = **2**

## ðŸ§© Sample Python Implementation
```python
class Solution(object):
    def findMinArrowShots(self, points):
        if not points:
            return 0

        points.sort(key=lambda x: x[1])  # sort by end
        arrows = 1
        last_arrow = points[0][1]

        for start, end in points[1:]:
            if start > last_arrow:        # no overlap â†’ new arrow needed
                arrows += 1
                last_arrow = end

        return arrows
