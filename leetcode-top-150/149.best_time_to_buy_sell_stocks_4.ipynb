## âœï¸ Short Approach Sketch

1. If `k >= n//2`, solve as **unlimited transactions**
2. Otherwise:

   * Initialize `buy` and `sell` arrays of size `k+1`
3. For each price:

   * Update `buy[t]` and `sell[t]` for `t = 1..k`
4. Return `sell[k]`

## ðŸ§® Python Solution (Optimal)

```python
class Solution(object):
    def maxProfit(self, k, prices):
        n = len(prices)
        if n == 0 or k == 0:
            return 0
        
        # Optimization: unlimited transactions
        if k >= n // 2:
            profit = 0
            for i in range(1, n):
                if prices[i] > prices[i - 1]:
                    profit += prices[i] - prices[i - 1]
            return profit
        
        buy = [float('-inf')] * (k + 1)
        sell = [0] * (k + 1)
        
        for price in prices:
            for t in range(1, k + 1):
                buy[t] = max(buy[t], sell[t - 1] - price)
                sell[t] = max(sell[t], buy[t] + price)
        
        return sell[k]
```

# My Solution

````python
class Solution:
    def maxProfit(self, k: int, prices: List[int]) -> int:
        dp = [[float('-inf'), 0] for _ in range(k)]
        for price in prices:
            for j in range(k):
                dp[j][0] = max(dp[j][0], (dp[j-1][1] if j > 0 else 0) - price)
                dp[j][1] = max(dp[j][1], dp[j][0] + price)
        return dp[-1][-1]
````
