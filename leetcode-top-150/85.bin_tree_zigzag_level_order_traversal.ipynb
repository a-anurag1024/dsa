## üìå Problem Summary  
Given the root of a binary tree, return its **zigzag (or spiral) level order traversal**:

- Level 0 ‚Üí left to right  
- Level 1 ‚Üí right to left  
- Level 2 ‚Üí left to right  
- ‚Ä¶alternating directions each level  

---

## üîç Example  

**Input:**  
`[3,9,20,null,null,15,7]`

**Output:**  
[
[3],
[20, 9],
[15, 7]
]


---

## üöÄ Key Idea  
Use **BFS level-order traversal**, but **reverse every alternate level**.

Two clean approaches:

### ‚≠ê Approach 1 ‚Äî BFS + Reverse  
- Perform standard BFS  
- If the current level index is odd, reverse the list  

### ‚≠ê Approach 2 ‚Äî Double-ended approach (appendleft)  
- Use a deque to build each level from the appropriate side  
- Avoids list reversal cost, but BFS + reverse is already optimal for interviews

Time Complexity: **O(n)**  
Space Complexity: **O(n)**  

---

## üìù Short Approach (Interview-Ready)

1. BFS with a queue  
2. Track `level_index` (0-based)  
3. For each level:
   - Collect values normally  
   - If `level_index` is odd ‚Üí reverse the list  
4. Append to result  
5. Increment `level_index`  

---

## üßë‚Äçüíª Template Code (BFS + Reverse)

```python
from collections import deque

class Solution:
    def zigzagLevelOrder(self, root):
        if not root:
            return []

        q = deque([root])
        result = []
        level_index = 0

        while q:
            level_size = len(q)
            level = []

            for _ in range(level_size):
                node = q.popleft()
                level.append(node.val)

                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)

            if level_index % 2 == 1:
                level.reverse()

            result.append(level)
            level_index += 1

        return result

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
from collections import deque

class Solution(object):
    def zigzagLevelOrder(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: List[List[int]]
        """
        traversal = []
        next_nodes = deque()
        if not root: return traversal
        next_nodes.append(root)
        left_to_right = True
        while next_nodes:
            level_traversal = []
            n = len(next_nodes)
            for _ in range(n):
                curr = next_nodes.popleft()
                level_traversal.append(curr.val)
                if curr.left: next_nodes.append(curr.left)
                if curr.right: next_nodes.append(curr.right)
            if left_to_right: traversal.append(level_traversal)
            else: traversal.append(level_traversal[-1::-1])
            left_to_right = not left_to_right
        return traversal
````
