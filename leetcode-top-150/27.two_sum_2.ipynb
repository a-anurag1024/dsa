You are given a **1-indexed**, **sorted** array `numbers` in non-decreasing order.  
Find **two numbers** that add up to a given `target`.  
Return their indices `[index1, index2]`, where:
- `1 <= index1 < index2 <= len(numbers)`
- Exactly **one solution** exists.
- You **cannot reuse** the same element.
- Must use **O(1) extra space**.

## âš™ï¸ Approach (Two-Pointer Technique)

### ğŸ”¹ Key Idea:
Because the array is **sorted**, we can use two pointers â€” one starting from the **beginning** and one from the **end** â€” to find the pair efficiently.

---

### ğŸ”¹ Step-by-Step Logic

1. Initialize two pointers:  
   `left = 0`, `right = len(numbers) - 1`
2. While `left < right`:
   - Compute the sum: `current_sum = numbers[left] + numbers[right]`
   - If `current_sum == target`:  
     â†’ Return `[left + 1, right + 1]` (1-indexed result)
   - If `current_sum < target`:  
     â†’ Move `left` one step right (to increase sum)
   - If `current_sum > target`:  
     â†’ Move `right` one step left (to decrease sum)
3. End when the pair is found (guaranteed to exist).

```python
def twoSum(numbers, target):
    left, right = 0, len(numbers) - 1
    while left < right:
        total = numbers[left] + numbers[right]
        if total == target:
            return [left + 1, right + 1]
        elif total < target:
            left += 1
        else:
            right -= 1

````python
class Solution(object):
    def twoSum(self, numbers, target):
        """
        :type numbers: List[int]
        :type target: int
        :rtype: List[int]
        """
        n = len(numbers)
        ind1 = 0
        ind2 = n - 1
        while ind1 < n and ind2 > -1:
            curr_sum = numbers[ind1] + numbers[ind2]
            if curr_sum == target:
                return [ind1+1, ind2+1]
            elif curr_sum < target:
                ind1 += 1
            else:
                ind2 -= 1
        return
````
