## üîÅ LeetCode 9: Palindrome Number ‚Äî Quick Review

### üß© Problem Summary

Given an integer `x`, determine whether it is a **palindrome** ‚Äî i.e., it reads the same **forward and backward**.

* `121` ‚Üí ‚úÖ palindrome
* `-121` ‚Üí ‚ùå not a palindrome
* `10` ‚Üí ‚ùå not a palindrome

---

## üí° Key Observations

1. **Negative numbers can never be palindromes**

   * `-121` reversed becomes `121-` ‚Üí invalid.
2. Numbers ending with `0` (except `0` itself) **cannot be palindromes**

   * `10 ‚Üí 01` ‚ùå
3. You **don‚Äôt need to reverse the full number**

   * Reversing **half** is enough (better space & time behavior).

---

## üß† Approach Sketch (Optimal ‚Äì No String Conversion)

### Idea: Reverse only the **second half** of the number

* Keep extracting digits from `x`
* Build `reversed_half`
* Stop when `reversed_half >= x`

At the end:

* For **even digits** ‚Üí `x == reversed_half`
* For **odd digits** ‚Üí `x == reversed_half // 10`

---

## ‚úÖ Python Solution (Math-Based, Interview Preferred)

```python
class Solution(object):
    def isPalindrome(self, x):
        # Edge cases
        if x < 0 or (x % 10 == 0 and x != 0):
            return False
        
        reversed_half = 0
        
        while x > reversed_half:
            reversed_half = reversed_half * 10 + x % 10
            x //= 10
        
        # Handles both even and odd digit lengths
        return x == reversed_half or x == reversed_half // 10
```

---

## ‚è±Ô∏è Complexity Analysis

* **Time:** `O(log‚ÇÅ‚ÇÄ n)` ‚Äî number of digits
* **Space:** `O(1)` ‚Äî constant space

# My Solution

````python
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        if x < 0: return False
        n = 0
        while x >= 10**n:
            n += 1
        st, en = 0, n - 1
        while st <= en:
            right = (x // (10**st)) % 10
            left = (x // (10**en)) % 10
            if left != right: return False
            st += 1
            en -= 1
        return True
````
