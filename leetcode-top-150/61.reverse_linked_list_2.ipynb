## ðŸ“˜ Problem Summary
You're given a singly linked list and two integers `left` and `right` (1-indexed, `left <= right`).

Your task:
Reverse the **portion of the list from position `left` to position `right`**  
and return the head of the modified list.

Only the sublist between `left` and `right` should be reversedâ€”everything else must remain unchanged.

---

## ðŸ§  Key Idea
Use **pointer manipulation** to:
1. Reach the node just before the reversal region.
2. Reverse the sublist between `left` and `right`.
3. Reconnect the reversed sublist back to the main list.

This is a classic in-place partial reversal problem.

---

## ðŸ”‘ Approach (O(n) Time, O(1) Space)
### Steps:
1. Use a **dummy node** to simplify edge cases (e.g., reversing from position 1).
2. Move a pointer `prev` until it reaches the node **just before** `left`.
3. Start reversing the sublist using the **standard in-place reverse technique**:
   - Keep moving nodes one by one from `curr.next` to the front.
4. Continue until the sublist is reversed.
5. Return `dummy.next`.

This avoids creating new nodes and reverses only the required part.

## ðŸ§© Sample Python Implementation
```python
class Solution(object):
    def reverseBetween(self, head, left, right):
        if not head or left == right:
            return head

        dummy = ListNode(0)
        dummy.next = head
        prev = dummy

        # Step 1: Move prev to (left - 1)-th node
        for _ in range(left - 1):
            prev = prev.next

        # Step 2: Reverse the sublist from left to right
        curr = prev.next
        for _ in range(right - left):
            nxt = curr.next
            curr.next = nxt.next
            nxt.next = prev.next
            prev.next = nxt

        return dummy.next

````python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution(object):
    def reverseBetween(self, head, left, right):
        """
        :type head: Optional[ListNode]
        :type left: int
        :type right: int
        :rtype: Optional[ListNode]
        """
        curr_ind = 1
        start = None
        curr = head
        if left == right: return head
        while curr_ind < left:
            start = curr
            curr = curr.next
            curr_ind += 1
        rev_head = curr
        rev_curr = rev_head
        rev_next = rev_curr.next
        while curr_ind < right:
            sub = rev_next.next
            rev_next.next = rev_curr
            rev_curr = rev_next
            rev_next = sub
            curr_ind += 1
        if start:
            start.next = rev_curr
        else:
            head = rev_curr
        rev_head.next = rev_next
        return head
````
