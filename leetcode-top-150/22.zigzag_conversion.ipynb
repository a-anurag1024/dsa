## âš™ï¸ Approach (Simulate the Zigzag Writing)

### ðŸ”¹ Key Idea:
Think of writing characters in a **downward and upward cycle** across rows.

- Maintain a list of strings (one for each row).
- Traverse the input string, adding each character to the appropriate row.
- Change direction when reaching the **top** or **bottom** row.

---

### ðŸ”¹ Step-by-Step

1. Initialize `rows = [""] * numRows`
2. Track the **current row** and **direction** (`down = True/False`)
3. For each character:
   - Append it to the current row string.
   - If at top (`row == 0`) â†’ change direction to down.
   - If at bottom (`row == numRows - 1`) â†’ change direction to up.
   - Move to next row accordingly.
4. Join all rows to form the final result.

## ðŸ§‘â€ðŸ’» Python Sketch

```python
def convert(s: str, numRows: int) -> str:
    if numRows == 1 or numRows >= len(s):
        return s

    rows = [""] * numRows
    cur_row, step = 0, 1

    for char in s:
        rows[cur_row] += char
        if cur_row == 0:
            step = 1       # Move downward
        elif cur_row == numRows - 1:
            step = -1      # Move upward
        cur_row += step

    return "".join(rows)


# My solution

````python
class Solution(object):
    def convert(self, s, numRows):
        """
        :type s: str
        :type numRows: int
        :rtype: str
        """
        output_string = [[] for _ in range(numRows)]
        stride = 2*numRows - 2 if numRows > 1 else 1
        for i in range(len(s)):
            seq_num = i % stride
            if seq_num < numRows: row_num = seq_num
            else: row_num = numRows - 1 - (seq_num - numRows + 1)
            output_string[row_num].append(s[i])
        return "".join(["".join(a) for a in output_string])
````
