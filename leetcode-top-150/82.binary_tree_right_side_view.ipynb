**Topic:** Binary Trees / BFS / DFS

## üìå Problem Summary  
You are standing on the **right side** of a binary tree.  
Return the list of node values you would see from **top to bottom**.

Essentially:  
‚û°Ô∏è **Return the rightmost node at each depth level.**

---

## üîç Examples

### Example 1  
**Input:** `[1,2,3,null,5,null,4]`  
Right side view:  
**Output:** `[1,3,4]`

### Example 2  
**Input:** `[1,2,3,4,null,null,null,5]`  
Right side view:  
**Output:** `[1,3,4,5]`

---

## üöÄ Key Idea  
Two common approaches:

### ‚≠ê Approach 1 ‚Äî BFS Level Order (Simplest)
For each level:
- Traverse children left ‚Üí right  
- The **last node of each level** is visible from the right side  
- Append it to the result  

### ‚≠ê Approach 2 ‚Äî DFS Preorder (Right-first)
Perform a DFS with priority:  
`root ‚Üí right ‚Üí left`  
At each depth:
- If it‚Äôs the **first time** visiting this depth ‚Üí record the node  
  *(because the rightmost node is visited first)*

Time Complexity: **O(n)**  
Space Complexity: **O(h)** (h = height of tree)

---

## üìù Short Approach (Interview-Ready Description)

‚úî Use BFS or DFS  
‚úî For each tree level, capture the **rightmost** node  
‚úî Works for skewed, balanced, or sparse trees  
‚úî Avoids storing entire levels unnecessarily  
‚úî O(n) time ‚Äî must visit each node once  

---

## üßë‚Äçüíª Template Code ‚Äî BFS Version (Simple & Clean)

```python
from collections import deque

class Solution:
    def rightSideView(self, root):
        if not root:
            return []

        q = deque([root])
        result = []

        while q:
            level_size = len(q)
            for i in range(level_size):
                node = q.popleft()
                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)
                # last element in level
                if i == level_size - 1:
                    result.append(node.val)

            return result

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
from collections import deque
class Solution(object):
    def rightSideView(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: List[int]
        """
        if not root:
            return []

        q = deque()
        q.append(root)
        result = []

        while q:
            level_size = len(q)
            for i in range(level_size):
                node = q.popleft()
                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)
            result.append(node.val)

        return result
````
