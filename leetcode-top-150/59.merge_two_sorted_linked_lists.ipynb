## ðŸ“˜ Problem Summary
You are given two **sorted linked lists**, `list1` and `list2`.  
Your task is to merge them into a **single sorted linked list** by directly splicing nodes (not creating new ones unnecessarily).

Return the head of the merged list.

---

## ðŸ§  Key Idea
Use the **two-pointer merge** technique (similar to merge step in merge sort):

- Maintain a pointer for each list.
- Compare node values.
- Append the smaller node to the merged list.
- Move the pointer of the list from which the node was taken.
- Continue until one list is exhausted.
- Append any remaining nodes from the other list.

Use a **dummy node** to simplify handling of the list head.

---

## ðŸ”‘ Approach (O(n + m) Time, O(1) Extra Space)
1. Create a dummy node â†’ `tail = dummy`.
2. Use two pointers: `p = list1`, `q = list2`.
3. While both `p` and `q` exist:
   - If `p.val < q.val`, append `p`, move `p`.
   - Else append `q`, move `q`.
4. Append whichever list still has remaining nodes.
5. Return `dummy.next`.

---

## ðŸ“ Example
Input:  
`list1 = [1,2,4]`  
`list2 = [1,3,4]`

Merged:

1 â†’ 1 â†’ 2 â†’ 3 â†’ 4 â†’ 4

## ðŸ§© Sample Python Implementation
```python
class Solution(object):
    def mergeTwoLists(self, list1, list2):
        dummy = ListNode(0)
        tail = dummy

        p, q = list1, list2

        while p and q:
            if p.val < q.val:
                tail.next = p
                p = p.next
            else:
                tail.next = q
                q = q.next
            tail = tail.next

        tail.next = p or q  # append leftover nodes

        return dummy.next

````python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        curr_merge = None
        head = None
        while list1 or list2:
            if list1 is None:
                added = "2"
            elif list2 is None:
                added = "1"
            else:
                if list1.val <= list2.val: added = "1"
                else: added = "2"
            if added == "1":
                c = list1
                list1 = list1.next
            else:
                c = list2
                list2 = list2.next
            if curr_merge:
                curr_merge.next = c
                curr_merge = curr_merge.next
            else:
                head = curr_merge = c
        return head
````
