* You must handle:

  * Different lengths

---

## âœï¸ Short Approach Sketch (Manual Addition)

1. Start from the **rightmost bits** of both strings
2. Keep a `carry`
3. At each step:

   * `sum = bit_a + bit_b + carry`
   * Result bit = `sum % 2`
   * New carry = `sum // 2`
4. Continue until both strings and carry are exhausted
5. Reverse the result

---

## â±ï¸ Complexity

* **Time**: `O(max(len(a), len(b)))`
* **Space**: `O(max(len(a), len(b)))`

## ðŸ§© Python Solution (Manual Binary Addition)

```python
class Solution(object):
    def addBinary(self, a, b):
        i, j = len(a) - 1, len(b) - 1
        carry = 0
        res = []

        while i >= 0 or j >= 0 or carry:
            total = carry
            if i >= 0:
                total += int(a[i])
                i -= 1
            if j >= 0:
                total += int(b[j])
                j -= 1

            res.append(str(total % 2))
            carry = total // 2

        return ''.join(reversed(res))
```

# My Solution

````python
class Solution(object):
    def addBinary(self, a, b):
        """
        :type a: str
        :type b: str
        :rtype: str
        """
        a_idx = len(a)
        b_idx = len(b)
        carry = 0
        res = ""
        while a_idx > 0 or b_idx > 0:
            a_idx -= 1
            b_idx -= 1
            s = carry
            if a_idx >= 0: s += int(a[a_idx])
            if b_idx >= 0: s += int(b[b_idx])
            if s in [0, 1]:
                res = str(s) + res
                carry = 0
            else :
                res = "0" + res if s == 2 else "1" + res
                carry = 1
        if carry != 0:
            res = "1" + res
        return res
````
