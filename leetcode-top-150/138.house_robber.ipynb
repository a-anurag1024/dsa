### üè† LeetCode 198 ‚Äî House Robber (Quick Review)

---

#### üîç Problem Recap

You are robbing houses along a street.
Each house has some money, but **you cannot rob two adjacent houses**.

üëâ **Goal:** Maximize the total money robbed without triggering the alarm.

---

### üí° Core Insight

At **each house**, you have **two choices**:

1. **Skip** the current house ‚Üí take the max till previous house
2. **Rob** the current house ‚Üí add its value to the max till house `i-2`

This leads to a **classic Dynamic Programming (DP)** recurrence.

---

### üß† DP State Definition

Let:

* `dp[i]` = maximum money you can rob from houses `0..i`

#### Recurrence

```
dp[i] = max(
    dp[i-1],          # skip current house
    dp[i-2] + nums[i] # rob current house
)
```

#### Base Cases

* `dp[0] = nums[0]`
* `dp[1] = max(nums[0], nums[1])`

---

### ‚úçÔ∏è Short Approach Sketch

1. Handle edge cases (`n = 0` or `n = 1`)
2. Track only the last **two DP states** (space optimization)
3. For each house:

   * Decide between robbing or skipping
4. Return the final maximum

---

### üßÆ Python Solution (Space Optimized)

```python
class Solution(object):
    def rob(self, nums):
        if not nums:
            return 0
        if len(nums) == 1:
            return nums[0]
        
        prev2 = nums[0]
        prev1 = max(nums[0], nums[1])
        
        for i in range(2, len(nums)):
            curr = max(prev1, prev2 + nums[i])
            prev2, prev1 = prev1, curr
        
        return prev1
```

---

### ‚è±Ô∏è Complexity Analysis

* **Time Complexity:** `O(n)`
* **Space Complexity:** `O(1)`


# My Solution

````python
class Solution(object):
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if len(nums) <= 2: return max(nums)
        second_last = max(nums[0], nums[1])
        last = max(second_last, nums[0]+nums[2])
        for i in range(3, len(nums)):
            new_last = max(last, nums[i]+second_last)
            second_last = last
            last = new_last
        return last
````
