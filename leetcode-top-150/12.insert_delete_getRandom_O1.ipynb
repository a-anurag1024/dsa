## üìù Problem Summary
Design a data structure `RandomizedSet` that supports the following operations in **O(1)** average time:

| Operation | Description |
|------------|--------------|
| `insert(val)` | Inserts a value if not present. Returns `True` if inserted successfully. |
| `remove(val)` | Removes a value if present. Returns `True` if removed successfully. |
| `getRandom()` | Returns a random element with **equal probability**. |

---
---

## ‚öôÔ∏è Approach (Hash Map + Dynamic Array)
### Idea:
To achieve **O(1)** operations:
- Use a **list (`arr`)** to store values for random access.
- Use a **hash map (`pos`)** to store value ‚Üí index mappings for O(1) lookup and deletion.

### Steps:
#### üü¢ `insert(val)`
- If `val` not in `pos`:
  - Append `val` to `arr`.
  - Store `pos[val] = len(arr) - 1`.
- Return `True` if inserted, else `False`.

#### üî¥ `remove(val)`
- If `val` in `pos`:
  - Swap it with the **last element** in `arr` (to delete in O(1)).
  - Update index of the swapped element in `pos`.
  - Pop the last element and remove `val` from `pos`.
- Return `True` if removed, else `False`.

#### üé≤ `getRandom()`
- Return a random element using Python‚Äôs `random.choice(arr)`.

## ‚úÖ Code Snippet (Python)
```python
import random

class RandomizedSet(object):

    def __init__(self):
        self.arr = []
        self.pos = {}

    def insert(self, val):
        if val in self.pos:
            return False
        self.arr.append(val)
        self.pos[val] = len(self.arr) - 1
        return True

    def remove(self, val):
        if val not in self.pos:
            return False
        idx = self.pos[val]
        last = self.arr[-1]
        self.arr[idx] = last
        self.pos[last] = idx
        self.arr.pop()
        del self.pos[val]
        return True

    def getRandom(self):
        return random.choice(self.arr)

````python
#Inefficient getRandom() function as O(N) copying from dictionary keys to list

import numpy as np

class RandomizedSet(object):

    def __init__(self):
        self.hashmap = {}
        self.n = 0

    def insert(self, val):
        """
        :type val: int
        :rtype: bool
        """
        if val in self.hashmap:
            return False
        else:
            self.hashmap[val] = self.n
            self.n += 1
            return True

    def remove(self, val):
        """
        :type val: int
        :rtype: bool
        """
        if val in self.hashmap:
            del self.hashmap[val]
            self.n -= 1
            return True
        else:
            return False
        

    def getRandom(self):
        """
        :rtype: int
        """
        return np.random.choice(self.hashmap.keys())
        

# Your RandomizedSet object will be instantiated and called as such:
# obj = RandomizedSet()
# param_1 = obj.insert(val)
# param_2 = obj.remove(val)
# param_3 = obj.getRandom()
````
