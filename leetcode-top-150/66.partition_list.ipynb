## ğŸ“Œ Problem Summary  
Given a linked list and a value `x`, partition the list so that:

- All nodes **less than x** come before  
- All nodes **greater than or equal to x**  

The **relative order** of nodes within each group must remain unchanged.

---

### Example 1  
**Input:** `[1,4,3,2,5,2], x = 3`  
**Output:** `[1,2,2,4,3,5]`

### Example 2  
**Input:** `[2,1], x = 2`  
**Output:** `[1,2]`

## ğŸš€ Key Idea  
Use **two separate linked lists**:
- One for nodes **< x**  
- One for nodes **>= x**  

Then connect them at the end.

This preserves ordering and avoids complex pointer juggling.

---

## ğŸ“ Short Approach (Interview-Ready)

1. Create two dummy lists:  
   - `left` â†’ nodes < x  
   - `right` â†’ nodes >= x  
2. Traverse original list:  
   - If `node.val < x` â†’ append to `left` list  
   - Else â†’ append to `right` list  
3. Connect: `left_tail.next = right_head`.  
4. Set `right_tail.next = None` to avoid cycles.  
5. Return `left_head`.

Time Complexity: **O(n)**  
Space Complexity: **O(1)** (reusing original nodes)

class Solution:
    def partition(self, head, x):
        left_dummy = ListNode(0)
        right_dummy = ListNode(0)

        left = left_dummy
        right = right_dummy
        curr = head

        while curr:
            if curr.val < x:
                left.next = curr
                left = left.next
            else:
                right.next = curr
                right = right.next
            curr = curr.next

        # Connect partitions
        left.next = right_dummy.next
        right.next = None  # end the list

        return left_dummy.next

````python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution(object):
    def partition(self, head, x):
        """
        :type head: Optional[ListNode]
        :type x: int
        :rtype: Optional[ListNode]
        """
        prev = dummy = ListNode()
        dummy.next = curr = head
        anchor = None
        while curr:
            if curr.val >= x and anchor is None:
                anchor = prev
            if curr.val < x and anchor:
                next_curr = curr.next
                prev.next = next_curr
                curr.next = anchor.next
                anchor.next = curr
                anchor = anchor.next
                if next_curr is None: break
                else: curr = next_curr
                continue
            curr = curr.next
            prev = prev.next
        return dummy.next
````
