## âš™ï¸ Approach (KMP â€” Knuth-Morris-Pratt Algorithm)

### ðŸ”¹ Key Idea:
Instead of restarting comparisons every time a mismatch happens,  
**KMP** uses information from previous matches to skip redundant checks.

It preprocesses the `needle` to build an array called **LPS (Longest Prefix which is also Suffix)**,  
which helps determine how much to shift the pattern when a mismatch occurs.

---

### ðŸ§© Step-by-Step Outline

1. **Build LPS array** for the pattern (`needle`):
   - `lps[i]` = length of longest prefix which is also a suffix for `needle[:i+1]`.
   - Helps know where to continue matching from when mismatch occurs.

2. **Search Phase:**
   - Iterate through `haystack` and `needle` using two pointers (`i` and `j`).
   - If characters match â†’ increment both pointers.
   - If mismatch â†’ use `lps[j-1]` to move `j` efficiently (avoid rechecking).
   - When `j == len(needle)`, a full match is found â€” return `i - j`.

---

## ðŸ§‘â€ðŸ’» Python Sketch

```python
def strStr(haystack: str, needle: str) -> int:
    if not needle:
        return 0

    # Step 1: Build LPS (Longest Prefix Suffix) array
    def build_lps(pattern):
        lps = [0] * len(pattern)
        prev_len = 0
        i = 1
        while i < len(pattern):
            if pattern[i] == pattern[prev_len]:
                prev_len += 1
                lps[i] = prev_len
                i += 1
            else:
                if prev_len != 0:
                    prev_len = lps[prev_len - 1]
                else:
                    lps[i] = 0
                    i += 1
        return lps

    lps = build_lps(needle)

    # Step 2: Pattern matching using KMP
    i = j = 0
    while i < len(haystack):
        if haystack[i] == needle[j]:
            i += 1
            j += 1
        if j == len(needle):
            return i - j  # match found
        elif i < len(haystack) and haystack[i] != needle[j]:
            if j != 0:
                j = lps[j - 1]
            else:
                i += 1
    return -1


# My solution

````python
class Solution(object):
    def strStr(self, haystack, needle):
        """
        :type haystack: str
        :type needle: str
        :rtype: int
        """
        k = -1
        nh = len(haystack)
        nl = len(needle)
        if nl == 0: return k
        for i in range(len(haystack)):
            if i+nl <= nh:
                if haystack[i:i+nl] == needle:
                    return i
        return -1
````
