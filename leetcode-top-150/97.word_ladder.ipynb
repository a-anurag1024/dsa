# üî§ LeetCode 127: Word Ladder ‚Äî Quick Review

## üìò Problem Summary
You are given:
- `beginWord` ‚Üí starting word
- `endWord` ‚Üí target word
- `wordList` ‚Üí dictionary of valid intermediate words

Rules:
- Each transformation changes **exactly one letter**
- Every intermediate word must be in `wordList`
- `beginWord` does **not** need to be in `wordList`

Goal:
‚û°Ô∏è Return the **length of the shortest transformation sequence**  
‚û°Ô∏è If impossible ‚Üí return `0`

---

## üß† Key Insight
This is a **shortest-path problem in an unweighted graph**.

- Each word ‚Üí **node**
- Edge exists if two words differ by **one character**
- Each transformation costs **1 step**

üëâ Use **Breadth-First Search (BFS)** to guarantee the shortest path.

---

## üöÄ Optimized Approach (BFS + Intermediate Patterns)

### Why optimization is needed
Naive comparison of all word pairs is `O(N¬≤)` ‚Üí too slow.

### Trick:
Replace one letter with `*` to build **generic patterns**:
hot ‚Üí ot, ht, ho*


Words sharing a pattern are neighbors.

---

## üöÄ Steps
1. If `endWord` is not in `wordList` ‚Üí return `0`
2. Preprocess `wordList` to build:

pattern ‚Üí list of words

3. BFS from `beginWord`
4. For each word:
- Generate its patterns
- Visit all neighbors via those patterns
5. Count BFS levels until `endWord` is found

---

## üß© Time & Space Complexity
- **Preprocessing:** `O(N √ó L)`
- **BFS:** `O(N √ó L)`
- **Space:** `O(N √ó L)`

Where:
- `N` = number of words
- `L` = word length

---

## üìù Sample BFS Code (Python)
```python
from collections import defaultdict, deque

class Solution:
 def ladderLength(self, beginWord, endWord, wordList):
     if endWord not in wordList:
         return 0
     
     L = len(beginWord)
     patterns = defaultdict(list)
     
     for word in wordList:
         for i in range(L):
             patterns[word[:i] + "*" + word[i+1:]].append(word)
     
     q = deque([(beginWord, 1)])
     visited = set([beginWord])
     
     while q:
         curr, level = q.popleft()
         if curr == endWord:
             return level
         
         for i in range(L):
             pat = curr[:i] + "*" + curr[i+1:]
             for nxt in patterns[pat]:
                 if nxt not in visited:
                     visited.add(nxt)
                     q.append((nxt, level + 1))
             patterns[pat] = []  # avoid reprocessing
     
     return 0
