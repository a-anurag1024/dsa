# üîí LeetCode 130: Surrounded Regions ‚Äî Quick Review

## üìò Problem Summary
You are given an **m √ó n board** containing:
- `'X'` ‚Üí walls  
- `'O'` ‚Üí open cells

A region of `'O'` should be **captured** (turned into `'X'`) *only if* the region:
- Is fully surrounded by `'X'`, and  
- **Does NOT touch the border**.

Your task:  
‚û°Ô∏è **Modify the board in-place**, turning all surrounded `'O'` ‚Üí `'X'`.

---

## üß† Key Insight
Any `'O'` that is **connected to the border** can *never* be captured.

So the idea is:
1. Find all **border `'O'` cells**.
2. DFS/BFS from them to mark all **safe regions**.
3. Flip:
   - Unmarked `'O'` ‚Üí `'X'` (captured)
   - Marked `'O'` ‚Üí `'O'` (remain unchanged)

---

## üöÄ Approach (DFS/BFS from border O's)
1. Traverse the **first row, last row, first column, last column**.
2. When you find `'O'` on the border:
   - Run DFS/BFS to mark all reachable `'O'`s as safe (commonly using `'#'`).
3. After marking:
   - Convert all remaining `'O'` ‚Üí `'X'`.
   - Convert all `'#'` ‚Üí `'O'` (restore safe ones).

This guarantees only truly surrounded regions get flipped.

---

## üß© Time & Space Complexity
- **Time:** `O(m * n)` ‚Äî every cell processed at most once  
- **Space:** `O(m * n)` worst-case DFS stack or BFS queue

---

## üìù Sample DFS Code (Python)
```python
class Solution:
    def solve(self, board):
        if not board: return
        m, n = len(board), len(board[0])

        def dfs(i, j):
            if i < 0 or j < 0 or i >= m or j >= n or board[i][j] != "O":
                return
            board[i][j] = "#"
            dfs(i+1, j)
            dfs(i-1, j)
            dfs(i, j+1)
            dfs(i, j-1)

        # Step 1: Mark border-connected regions
        for i in range(m):
            dfs(i, 0)
            dfs(i, n-1)
        for j in range(n):
            dfs(0, j)
            dfs(m-1, j)

        # Step 2: Flip surrounded regions
        for i in range(m):
            for j in range(n):
                if board[i][j] == "O":
                    board[i][j] = "X"
                elif board[i][j] == "#":
                    board[i][j] = "O"
