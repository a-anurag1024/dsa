## üß© Problem Summary

You are given an integer array `ratings` representing the ratings of `n` children standing in a line.  
You must distribute candies to these children such that:

1. Each child gets **at least one candy** üç≠  
2. Children with a **higher rating** than their **neighbors** receive **more candies**

üéØ **Goal:**  
Return the **minimum number of candies** needed to satisfy the above conditions.

---

## üß† Example

### Example 1:
**Input:** `ratings = [1, 0, 2]`  
**Output:** `5`  
**Explanation:**  
Candies distribution: `[2, 1, 2]`

### Example 2:
**Input:** `ratings = [1, 2, 2]`  
**Output:** `4`  
**Explanation:**  
Candies distribution: `[1, 2, 1]`

---

## ‚öôÔ∏è Approach

We need to ensure **both left and right neighbor rules** are satisfied.

### üîπ Step-by-Step Strategy

1. **Initialize** a candies array with all `1`s (minimum requirement).
2. **Left-to-Right pass:**  
   - If `ratings[i] > ratings[i-1]`, give one more candy than `i-1`.  
   - Ensures left neighbor condition.
3. **Right-to-Left pass:**  
   - If `ratings[i] > ratings[i+1]`, ensure `candies[i]` is at least one more than `i+1`.  
   - Ensures right neighbor condition.
4. **Sum up all candies** to get the minimum total.

---

## üí° Intuition

The two-pass approach ensures fairness from **both directions** ‚Äî  
each child gets enough candies relative to both left and right neighbors.

This problem elegantly demonstrates **greedy optimization from local constraints**.

---

## üßÆ Complexity

| Type | Complexity |
|------|-------------|
| ‚è±Ô∏è Time | `O(n)` |
| üß† Space | `O(n)` |

---

## üßë‚Äçüíª Python Sketch

```python
def candy(ratings):
    n = len(ratings)
    candies = [1] * n

    # Left to right
    for i in range(1, n):
        if ratings[i] > ratings[i - 1]:
            candies[i] = candies[i - 1] + 1

    # Right to left
    for i in range(n - 2, -1, -1):
        if ratings[i] > ratings[i + 1]:
            candies[i] = max(candies[i], candies[i + 1] + 1)

    return sum(candies)


## Another more elegant solution

### üß© Idea

Observe the `ratings` array as a **sequence of slopes**:

* Increasing ‚Üí up-slope
* Decreasing ‚Üí down-slope

You can:

1. Identify all **peaks** and **valleys**
2. Compute the total candies by summing arithmetic progressions for increasing and decreasing slopes, ensuring the peak is not double-counted.

### ‚öôÔ∏è Steps (Conceptual)

For each segment:

* For a continuous increasing slope of length `inc`, add `1 + 2 + ... + inc`
* For a continuous decreasing slope of length `dec`, add `1 + 2 + ... + dec`
* Adjust for overlaps at peaks.

### ‚è±Ô∏è Complexity

* **Time:** `O(n)`
* **Space:** `O(1)`

‚úÖ *Mathematically elegant.*
‚ùå *Harder to implement correctly ‚Äî more error-prone.*

--
