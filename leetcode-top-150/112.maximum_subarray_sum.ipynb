### **Problem Summary**

Given an integer array `nums`, find the **contiguous subarray** with the **maximum sum** and return that sum.

* Subarray must be **non-empty**
* Works with **negative numbers**
* Order matters (contiguous segment)

---

### **Key Observations**

* Brute force (`O(nÂ²)`) is too slow
* At each index, decide:

  * **Extend** the previous subarray
  * **Start fresh** from current element
* This leads to **Kadaneâ€™s Algorithm**
* Dynamic Programming with **constant space**

---

## ðŸ”¹ Short Approach Sketch (Kadaneâ€™s Algorithm)

1. Initialize:

   * `curr_sum` = `nums[0]`
   * `max_sum` = `nums[0]`
2. Iterate from index `1` to `n-1`:

   * `curr_sum = max(nums[i], curr_sum + nums[i])`
   * `max_sum = max(max_sum, curr_sum)`
3. Return `max_sum`

---

### **Time & Space Complexity**

* **Time:** `O(n)`
* **Space:** `O(1)`

## ðŸ”¹ Python Solution (Optimal)

```python
class Solution(object):
    def maxSubArray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        curr_sum = nums[0]
        max_sum = nums[0]

        for i in range(1, len(nums)):
            curr_sum = max(nums[i], curr_sum + nums[i])
            max_sum = max(max_sum, curr_sum)

        return max_sum
```


# My Solution

````python
class Solution(object):
    def maxSubArray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        max_sum = -float('inf')
        curr_sum = max_sum
        for i in range(len(nums)):
            curr_sum = max(curr_sum+nums[i], nums[i])
            max_sum = max(max_sum, curr_sum)
        return max_sum
````
