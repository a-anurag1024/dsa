## ðŸ”¹ Quick Review: LeetCode 22 â€” Generate Parentheses

### **Problem Summary**

Given `n` pairs of parentheses, generate **all valid (well-formed)** combinations.

* Each combination must be **balanced**
* Order matters
* Output size follows the **Catalan number** sequence

---

### **Key Observations**

* You cannot place `')'` unless there is a matching `'('`
* At any point:

  * `open_count â‰¤ n`
  * `close_count â‰¤ open_count`
* This is **not brute force** â†’ invalid states must be pruned early
* Classic **backtracking with constraints**

---

## ðŸ”¹ Short Approach Sketch (Backtracking)

1. Use DFS with:

   * `open_used`: number of `'('` used
   * `close_used`: number of `')'` used
   * `path`: current string
2. If `open_used == close_used == n` â†’ add to result
3. Add `'('` if `open_used < n`
4. Add `')'` if `close_used < open_used`
5. Backtrack after each choice

---

### **Time & Space Complexity**

* **Time:** `O(Câ‚™)` (Catalan number, output-bound)
* **Space:** `O(n)` recursion depth (excluding output)

---

## ðŸ”¹ Python Solution (Backtracking â€“ Optimal)

```python
class Solution(object):
    def generateParenthesis(self, n):
        """
        :type n: int
        :rtype: List[str]
        """
        res = []

        def backtrack(open_used, close_used, path):
            if open_used == close_used == n:
                res.append(path)
                return

            # Add '(' if we still can
            if open_used < n:
                backtrack(open_used + 1, close_used, path + "(")

            # Add ')' only if valid
            if close_used < open_used:
                backtrack(open_used, close_used + 1, path + ")")

        backtrack(0, 0, "")
        return res
```

---

## ðŸ”¹ Why This Works

* Invalid sequences are **never generated**
* Ensures correctness via **strict constraints**
* Much more efficient than generating all `2^(2n)` strings

# My Solution

````python

class Solution(object):
    def generateParenthesis(self, n):
        """
        :type n: int
        :rtype: List[str]
        """
        combinations = []
        st = []
        curr = []
        def helper(i):
            if len(curr) == 2*n:
                combinations.append("".join(curr[:]))
                return
            # opening a parantheses
            if i < n:
                curr.append('(')
                st.append('(')
                helper(i+1)
                curr.pop()
                st.pop()
            if st:
                st.pop()
                curr.append(')')
                helper(i)
                st.append('(')
                curr.pop()
        helper(0)
        return combinations
````
