## ðŸ“Œ Problem Summary  
Two strings **s** and **t** are *isomorphic* if characters in **s** can be replaced (one-to-one mapping) to transform it into **t** while preserving order.

### Rules:
- Each character in **s** must map to exactly one character in **t**.
- No two characters in **s** can map to the same character in **t**.
- A character may map to itself.

---

## ðŸ§  Key Idea
We need a **bijective (one-to-one)** mapping between the characters of `s` and `t`.

To validate this:
- Track mapping `s â†’ t`
- Track used characters in `t` to prevent duplicate assignments

---

## âœ… Short Approach

### **Method: Hash Maps / Dictionaries**
1. Maintain:
   - A map `s_to_t` for characters in `s` â†’ `t`
   - A set `mapped` to ensure no two `s` chars map to the same `t` char
2. For each index:
   - If `s[i]` is already mapped, ensure it maps to `t[i]`
   - If not mapped:
     - Ensure `t[i]` is not already assigned to another char
     - Assign mapping
3. If all checks pass â†’ return `True`

### Complexity:
- **Time:** `O(n)`
- **Space:** `O(1)` (max 256 possible chars)

class Solution:
    def isIsomorphic(self, s, t):
        s_to_t = {}
        used = set()
        
        for a, b in zip(s, t):
            if a in s_to_t:
                if s_to_t[a] != b:
                    return False
            else:
                if b in used:
                    return False
                s_to_t[a] = b
                used.add(b)
        
        return True
