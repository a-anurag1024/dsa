# üî¢ LeetCode #26 ‚Äî Remove Duplicates from Sorted Array

**Difficulty:** Easy  
**Topic:** Array, Two Pointers  
**Pattern:** In-place modification on sorted array

## üìù Problem Summary
Given a **sorted** integer array `nums`, remove the **duplicates in-place** so that each unique element appears **only once**.  
The **relative order** of elements must be maintained.  

Return the number of **unique elements** `k`, where the first `k` elements of `nums` contain the distinct values in sorted order.

---

## üí° Example
**Input:**  
`nums = [1,1,2]`  
**Output:**  
`k = 2, nums = [1,2,_]`

**Input:**  
`nums = [0,0,1,1,1,2,2,3,3,4]`  
**Output:**  
`k = 5, nums = [0,1,2,3,4,_,_,_,_,_]`

---

## ‚öôÔ∏è Approach (Two-Pointer Technique)
1. Since the array is **sorted**, duplicates will be **adjacent**.  
2. Use two pointers:
   - `i` ‚Üí tracks the **position of the last unique element**.
   - `j` ‚Üí scans through the array.
3. For each `j`:
   - If `nums[j] != nums[i]`, increment `i` and set `nums[i] = nums[j]`.
4. After traversal, the count of unique elements is `i + 1`.

---

## ‚è±Ô∏è Complexity
- **Time:** O(n) ‚Äî single pass through the array.  
- **Space:** O(1) ‚Äî done in-place without extra memory.

## ‚úÖ Code Snippet (Python)
```python
def removeDuplicates(nums):
    if not nums:
        return 0
    i = 0
    for j in range(1, len(nums)):
        if nums[j] != nums[i]:
            i += 1
            nums[i] = nums[j]
    return i + 1

````python
class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if len(nums) < 2:
            return len(nums)

        i = 0 
        for j in range(1, len(nums)):
            if nums[j] != nums[i]:
                nums[i+1] = nums[j]
                i = i + 1
        return i + 1

        
````
