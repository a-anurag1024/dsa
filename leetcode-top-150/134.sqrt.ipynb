## âœ… Python Solution (Interview-Standard)

```python
class Solution(object):
    def mySqrt(self, x):
        if x < 2:
            return x
        
        left, right = 1, x // 2
        
        while left <= right:
            mid = (left + right) // 2
            sq = mid * mid
            
            if sq == x:
                return mid
            elif sq < x:
                left = mid + 1
            else:
                right = mid - 1
        
        return right
```

````python
# Not optimal due to linear search but with reduced search space
class Solution(object):
    def mySqrt(self, x):
        """
        :type x: int
        :rtype: int
        """
        if x <= 1: return x
        if x == 2: return 1
        num_of_digits = 0
        curr = x
        while curr > 0:
            num_of_digits += 1
            curr //= 10
        digits_of_sqrt = num_of_digits // 2 
        st_num = 10**(digits_of_sqrt-1) if digits_of_sqrt > 0 else 0
        en_num = 10**(digits_of_sqrt+1)
        for n in range(st_num, en_num):
            if n*n == x: return n
            if n*n > x: return n - 1
````
