## ðŸ§  Key Observations

* Binary representation consists of only `0` and `1`
* You only need to **count how many `1`s** are present
* Bit manipulation gives the most elegant solution

---

## âœï¸ Short Approach Sketch (Bit Trick)

1. Initialize `count = 0`
2. While `n` is not zero:

   * Remove the **lowest set bit** using `n & (n - 1)`
   * Increment count
3. Return count

Each loop removes **one `1` bit**, so the loop runs only as many times as there are set bits.

---

## â±ï¸ Complexity

* **Time**: `O(number of set bits)` â†’ very fast
* **Space**: `O(1)`

## ðŸ§© Python Solution (Optimal â€“ Brian Kernighanâ€™s Algorithm)

```python
class Solution(object):
    def hammingWeight(self, n):
        count = 0
        while n:
            n = n & (n - 1)
            count += 1
        return count
```

## ðŸ§© Alternative Solution (Bit-by-Bit)

```python
class Solution(object):
    def hammingWeight(self, n):
        count = 0
        while n:
            count += n & 1
            n >>= 1
        return count
```

# My Solution

````python
class Solution(object):
    def hammingWeight(self, n):
        """
        :type n: int
        :rtype: int
        """
        n_one = 0
        while n > 0:
            n_one += 1 if n & 1 else 0
            n >>= 1
        return n_one
````
