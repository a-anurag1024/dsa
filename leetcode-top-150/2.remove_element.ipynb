# üßÆ LeetCode #27 ‚Äî Remove Element

**Difficulty:** Easy  
**Topic:** Array, Two Pointers  
**Pattern:** In-place modification

## üìù Problem Summary
Given an integer array `nums` and an integer `val`, remove all occurrences of `val` **in-place** and return the number of elements `k` that are **not equal** to `val`.

After the operation:
- The first `k` elements of `nums` should contain values **not equal to `val`**.
- The remaining elements beyond `k` can be ignored.

---

## üí° Example
**Input:**  
`nums = [3,2,2,3], val = 3`  
**Output:**  
`k = 2, nums = [2,2,_,_]`

**Input:**  
`nums = [0,1,2,2,3,0,4,2], val = 2`  
**Output:**  
`k = 5, nums = [0,1,4,0,3,_,_,_]`

---

## ‚öôÔ∏è Approach (Two-Pointer Technique)
1. Initialize a pointer `i = 0` ‚Äî this marks the **position to place** the next valid element.
2. Iterate through the array with another pointer `j`:
   - If `nums[j] != val`,  
     ‚Üí assign `nums[i] = nums[j]`  
     ‚Üí increment `i`.
3. After the loop ends, `i` will represent the count `k` (number of valid elements).
4. Return `i`.

---

## ‚è±Ô∏è Complexity
- **Time:** O(n) ‚Äî each element checked once.  
- **Space:** O(1) ‚Äî in-place modification.

---

## ‚úÖ Code Snippet (Python)
```python
def removeElement(nums, val):
    i = 0
    for j in range(len(nums)):
        if nums[j] != val:
            nums[i] = nums[j]
            i += 1
    return i


# My Original Solution

````python
class Solution(object):
    def removeElement(self, nums, val):
        """
        :type nums: List[int]
        :type val: int
        :rtype: int
        """
        i = 0
        j = len(nums) - 1
        num_of_vals = len(nums)
        while i <= j:
            if nums[j] == val:
                num_of_vals -= 1
                j -= 1
            elif nums[i] == val:
                nums[i], nums[j] = nums[j], nums[i]
                num_of_vals -= 1
                i += 1
                j -= 1
            else:
                i += 1
        return num_of_vals
````
