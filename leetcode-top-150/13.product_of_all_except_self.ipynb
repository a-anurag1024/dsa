# ‚ú≥Ô∏è LeetCode #238 ‚Äî Product of Array Except Self

**Difficulty:** Medium  
**Topic:** Array, Prefix Product, Suffix Product  
**Pattern:** Multiplicative prefix-suffix traversal

## üìù Problem Summary
Given an integer array `nums`, return an array `answer` where:
\[
\text{answer[i]} = \prod_{\substack{j=0 \\ j \ne i}}^{n-1} \text{nums[j]}
\]
You must:
- Solve it in **O(n)** time,  
- **Not use division**,  
- Ensure results fit in 32-bit integers.

---

## üí° Example
**Input:**  
`nums = [1,2,3,4]`  
**Output:**  
`[24,12,8,6]`  

**Explanation:**  
- Product except index 0 ‚Üí 2 √ó 3 √ó 4 = 24  
- Product except index 1 ‚Üí 1 √ó 3 √ó 4 = 12  
- Product except index 2 ‚Üí 1 √ó 2 √ó 4 = 8  
- Product except index 3 ‚Üí 1 √ó 2 √ó 3 = 6

---

## ‚öôÔ∏è Approach (Prefix‚ÄìSuffix Product)
### Idea:
To compute product except `nums[i]`,  
we can multiply:
- the **product of all elements to the left** of `i` (prefix product), and  
- the **product of all elements to the right** of `i` (suffix product).

We can do this efficiently in **two passes** using one output array.

---

### üîç Steps:
1. Initialize `answer[i]` to store **prefix products**:
   - `answer[i] = answer[i-1] * nums[i-1]`
2. Traverse backward to multiply **suffix products**:
   - Maintain a running variable `suffix = 1`.
   - Update `answer[i] *= suffix`.
   - Then set `suffix *= nums[i]`.

---

## ‚úÖ Code Snippet (Python)
```python
def productExceptSelf(nums):
    n = len(nums)
    answer = [1] * n

    # Step 1: Prefix products
    for i in range(1, n):
        answer[i] = answer[i - 1] * nums[i - 1]

    # Step 2: Multiply with suffix products
    suffix = 1
    for i in range(n - 1, -1, -1):
        answer[i] *= suffix
        suffix *= nums[i]

    return answer


# My solution

````python
class Solution(object):
    def productExceptSelf(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        output = [1]*len(nums)
        tot_prod = 1
        for i in range(len(nums)):
            output[i] = tot_prod
            tot_prod *= nums[i]
        rem_prod = nums[-1]
        for i in range(len(nums)-2, -1, -1):
            output[i] = output[i] * rem_prod
            rem_prod *= nums[i]
        return output
````
