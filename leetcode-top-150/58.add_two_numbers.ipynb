# ‚ûï LeetCode 2: Add Two Numbers ‚Äî Quick Review

## üìò Problem Summary
You are given two **non-empty linked lists** representing two non-negative integers.  
- Digits are stored in **reverse order** (1's place first).
- Each node contains a **single digit**.
- Add the two numbers and return the result as a **linked list**, also in reverse order.

Carry must be handled properly (just like elementary addition).

---

## üß† Key Idea
Simulate digit-by-digit addition exactly like you would on paper:

1. Traverse both lists simultaneously.
2. Add the corresponding digits plus any carry.
3. Create a new node with `(sum % 10)` and update `carry = sum // 10`.
4. Continue until both lists are exhausted **and** no carry remains.

This naturally builds the result in reverse order.

---

## üîë Approach (O(n) Time, O(n) Space)
1. Create a dummy head node to simplify list construction.
2. Use pointers `p` and `q` to traverse `l1` and `l2`.
3. Maintain `carry = 0`.
4. Loop while `p` or `q` or `carry`:
   - Extract digits:  
     `x = p.val if p else 0`  
     `y = q.val if q else 0`
   - Compute:  
     `s = x + y + carry`
   - New digit ‚Üí `s % 10`, update carry ‚Üí `s // 10`
   - Append new node to result.
5. Return `dummy.next`.

---

## üìù Example
Input:  
`l1 = [2,4,3]`  
`l2 = [5,6,4]`  

Digit-wise addition (reverse order):
- 2 + 5 = 7
- 4 + 6 = 0 (carry 1)
- 3 + 4 + 1 = 8


Output:  
`[7,0,8]` ‚Üí represents 807

---

## üß© Sample Python Implementation
```python
class Solution(object):
    def addTwoNumbers(self, l1, l2):
        dummy = ListNode(0)
        curr = dummy
        carry = 0

        p, q = l1, l2

        while p or q or carry:
            x = p.val if p else 0
            y = q.val if q else 0

            s = x + y + carry
            carry = s // 10

            curr.next = ListNode(s % 10)
            curr = curr.next

            if p: p = p.next
            if q: q = q.next

        return dummy.next


# My Solution

````python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution(object):
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: Optional[ListNode]
        :type l2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        carry = 0
        curr = None
        while l1 or l2:
            val = carry
            if l1: val += l1.val
            if l2: val += l2.val
            carry = val // 10
            val = val % 10
            if not curr:
                curr = ListNode(val=val)
                head = curr
            else:
                curr.next = ListNode(val=val)
                curr = curr.next
            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next
        if carry:
            curr.next = ListNode(val=carry)
        return head if curr != None else None
````
