## ðŸ“˜ Problem Summary
You're given an array of tokens representing an arithmetic expression in **Reverse Polish Notation (RPN)**:

- Valid operators: `+`, `-`, `*`, `/`
- Division truncates **toward zero** (same as Python `int(a / b)`)
- Input is guaranteed to be valid

Your task: **Evaluate the expression and return the resulting integer.**

---

## ðŸ§  Key Idea
Use a **stack** to simulate RPN evaluation.

### RPN Evaluation Rules:
- When you see a **number**, push it onto the stack.
- When you see an **operator**:
  - Pop the top two values (`b` then `a`)
  - Compute `a op b`
  - Push the result back onto the stack

At the end, the stack contains exactly **one number**, which is the result.

---

## ðŸ”‘ Approach (O(n) Time, O(n) Space)
1. Create a stack.
2. For each token:
   - If it's a number â†’ push `int(token)` to stack.
   - If it's an operator:
     - pop `b`, pop `a`
     - compute `a op b`
     - push the result
3. Return the only element left in the stack.

### ðŸ“Œ Note on Division:
- Division must **truncate toward zero**:
  ```python
  int(a / b)
  ```

## Sample Python Implementation 

```python
class Solution(object):
    def evalRPN(self, tokens):
        stack = []
        ops = {"+", "-", "*", "/"}

        for token in tokens:
            if token not in ops:
                stack.append(int(token))
            else:
                b = stack.pop()
                a = stack.pop()

                if token == "+":
                    stack.append(a + b)
                elif token == "-":
                    stack.append(a - b)
                elif token == "*":
                    stack.append(a * b)
                else:  # division truncates toward zero
                    stack.append(int(float(a) / b))

        return stack[-1]
```

# My Solution

````python
from collections import deque

class Solution(object):
    def evalRPN(self, tokens):
        """
        :type tokens: List[str]
        :rtype: int
        """
        st = deque()
        for t in tokens:
            if t not in ['+', '-', '*', '/']:
                st.append(t)
            else:
                s = int(st.pop())
                f = int(st.pop())
                if t == '+': r = f + s
                elif t == '-': r = f - s
                elif t == '*': r = f * s
                elif t == '/': r = int(f / float(s)) 
                st.append(str(r))
        return int(st.pop())
````
