#### ðŸ” Problem Recap

Youâ€™re given two strings:

* `word1`
* `word2`

ðŸ‘‰ **Goal:** Find the **minimum number of operations** to convert `word1` into `word2`.

Allowed operations:

1. **Insert** a character
2. **Delete** a character
3. **Replace** a character

This is also known as **Levenshtein Distance**.

---

## ðŸ’¡ Core Insight

This is a **classic 2D Dynamic Programming** problem where each state represents:

> *How many edits are needed to match prefixes of the two strings?*

Each character comparison gives you **three choices**.

---

## ðŸ§  DP State Definition

Let:

```
dp[i][j] = minimum operations to convert
           word1[0:i] â†’ word2[0:j]
```

---

## ðŸ§± Base Cases

* `dp[0][j] = j` â†’ insert all characters of `word2`
* `dp[i][0] = i` â†’ delete all characters of `word1`
* `dp[0][0] = 0`

---

## ðŸ” DP Transition

If characters match:

```
word1[i-1] == word2[j-1]
dp[i][j] = dp[i-1][j-1]
```

Else:

```
dp[i][j] = 1 + min(
    dp[i-1][j],     # delete
    dp[i][j-1],     # insert
    dp[i-1][j-1]    # replace
)
```

---

## âœï¸ Short Approach Sketch

1. Create a DP table of size `(m+1) Ã— (n+1)`
2. Initialize base cases
3. Fill table row by row
4. Return `dp[m][n]`

---

## ðŸ§® Python Solution (Space Optimized O(n))

```python
class Solution(object):
    def minDistance(self, word1, word2):
        m, n = len(word1), len(word2)
        
        dp = list(range(n + 1))
        
        for i in range(1, m + 1):
            prev_diag = dp[0]
            dp[0] = i
            for j in range(1, n + 1):
                temp = dp[j]
                if word1[i - 1] == word2[j - 1]:
                    dp[j] = prev_diag
                else:
                    dp[j] = 1 + min(
                        dp[j],       # delete
                        dp[j - 1],   # insert
                        prev_diag    # replace
                    )
                prev_diag = temp
        
        return dp[n]
```

# My Solution

````python
class Solution:
    def minDistance(self, word1: str, word2: str) -> int:
        m, n = len(word1), len(word2)
        last_dp = [i for i in range(n+1)]
        for j in range(m):
            curr_dp = [j+1]
            for i in range(n):
                if word1[j] == word2[i]:
                    curr = last_dp[i]
                else:
                    curr = 1 + last_dp[i]
                if i > 0: curr = min(curr, 1 + curr_dp[i])
                if i < n: curr = min(curr, 1 + last_dp[i+1])
                curr_dp.append(curr)
            last_dp = curr_dp[:]
        return last_dp[-1]
                
````
