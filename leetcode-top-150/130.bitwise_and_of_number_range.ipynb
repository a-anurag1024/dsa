## üîç Problem Recap

You‚Äôre given two integers `left` and `right` representing a **range [left, right]**.
Return the **bitwise AND** of **all numbers** in this range (inclusive).

---

## üß† Key Insight

**Bitwise AND keeps a bit = 1 only if it is 1 in *every* number.**

If `left != right`, lower bits will **eventually flip from 0 to 1 or 1 to 0** across the range ‚Üí they become `0` in the final AND.

üëâ Only the **common left (most significant) prefix** of `left` and `right` survives.

---

## ‚úèÔ∏è Short Approach Sketch (Shift Until Equal)

1. While `left < right`:

   * Right-shift both numbers by 1
   * Count shifts
2. When `left == right`, that value is the common prefix
3. Left-shift back by the number of shifts

---

## ‚è±Ô∏è Complexity

* **Time**: `O(log(right - left))` (at most 32 shifts)
* **Space**: `O(1)`

---

## üß© Python Solution (Most Common & Clean)

```python
class Solution(object):
    def rangeBitwiseAnd(self, left, right):
        shifts = 0
        
        while left < right:
            left >>= 1
            right >>= 1
            shifts += 1
        
        return left << shifts
```

---

## üß© Alternative Solution (Brian Kernighan Trick)

```python
class Solution(object):
    def rangeBitwiseAnd(self, left, right):
        while right > left:
            right = right & (right - 1)
        return right
```

**Why it works**:

* Clears the lowest set bit of `right`
* Repeatedly removes unstable bits until `right <= left`

### üîë One-Line Memory Hook

> **Only the common binary prefix survives a range AND**


# My Solution

````python
class Solution(object):
    def rangeBitwiseAnd(self, left, right):
        """
        :type left: int
        :type right: int
        :rtype: int
        """
        shift = 0
        while left < right:
            left >>= 1
            right >>= 1
            shift += 1
        return left << shift
````
