## üìå Problem Summary  
Given the root of a binary tree, flatten it **in-place** into a linked list following:

- The list uses **right pointers** only  
- All **left pointers must be null**  
- The node order must match **preorder traversal**  
  (`root ‚Üí left ‚Üí right`)

---

### Example 1  
**Input:**

**Output:**  
`[1,null,2,null,3,null,4,null,5,null,6]`

### Example 2  
**Input:** `[]`  
**Output:** `[]`

### Example 3

**Output:** `[0]`

/

/ \

After: 1 ‚Üí 2 ‚Üí 3 ‚Üí 4 ‚Üí 5 ‚Üí 6


### ‚öôÔ∏è Follow-up constraint:  
**Flatten in-place ‚Üí O(1) extra space**, no arrays or helper lists.

## üìù Short Approaches (Interview-Ready)

### **Approach 1 ‚Äî Reverse Postorder DFS (Optimal & Clean)**  
Use recursion:  
1. Maintain a pointer `prev = None`.  
2. Visit nodes in order **right ‚Üí left ‚Üí root** (reverse preorder).  
3. Set:

root.right = prev

root.left = None

prev = root

4. Eventually, the entire tree is flattened.

Time: **O(n)**  
Space: **O(h)** recursion stack  

---

### **Approach 2 ‚Äî Iterative Morris Traversal (O(1) space)**  
For each node:
1. If it has a left subtree:
- Find the **rightmost node** in the left subtree  
- Connect its `.right` to current node‚Äôs `.right`  
- Move the entire left subtree to the right  
2. Move to `curr.right`.

Time: **O(n)**  
Space: **O(1)**  

---

## üßë‚Äçüíª Template Code (Recursive Reverse-Preorder)

```python
class Solution:
 def flatten(self, root):
     self.prev = None
     
     def dfs(node):
         if not node:
             return
         dfs(node.right)
         dfs(node.left)
         
         node.right = self.prev
         node.left = None
         self.prev = node

     dfs(root)
```

## Optimal Solution (iterative)
```python
class Solution:
    def flatten(self, root):
        curr = root
        while curr:
            if curr.left:
                # find rightmost of left subtree
                rightmost = curr.left
                while rightmost.right:
                    rightmost = rightmost.right
                rightmost.right = curr.right
                curr.right = curr.left
                curr.left = None
            curr = curr.right
```

# My Solution

````python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
class Solution(object):
    def flatten(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: None Do not return anything, modify root in-place instead.
        """
        def helper(curr):
            tail = curr
            left = curr.left if curr.left else None
            right = curr.right if curr.right else None
            if left:
                last_left = helper(left)
                tail.right = left
                tail = last_left
            if right:
                last_right = helper(right)
                tail.right = right
                tail = last_right
            curr.left = None
            return tail
        if root:
            helper(root)
        return root
````
