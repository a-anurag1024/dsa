## üìå Problem Summary  
Given the head of a **sorted** linked list, remove all nodes that have duplicate values ‚Äî meaning **any value that appears more than once should be fully removed**, not kept once.

The output must remain sorted.

---

## üîç Examples  

### Example 1  
**Input:** `[1,2,3,3,4,4,5]`  
**Output:** `[1,2,5]`

### Example 2  
**Input:** `[1,1,1,2,3]`  
**Output:** `[2,3]`

---

## üöÄ Key Idea  
Use pointer scanning to detect duplicate segments and skip them entirely.

Since the list is **sorted**, duplicates appear in contiguous blocks.

---

## üìù Short Approach (Interview-Ready)

1. Create a `dummy` node pointing to head (simplifies deleting the head).  
2. Maintain two pointers:  
   - `prev` ‚Üí last node confirmed to be distinct  
   - `curr` ‚Üí current node being examined  
3. As long as `curr.val == curr.next.val`, move `curr` forward to skip *all* duplicates.  
4. If duplicates were found, set `prev.next = curr.next` to **remove the entire block**.  
5. Else (no duplicates), simply move `prev = prev.next`.  
6. Continue until the end.

Time Complexity: **O(n)**  
Space Complexity: **O(1)**

---

## üßë‚Äçüíª Template Code (Clean & Minimal)
```python
class Solution:
    def deleteDuplicates(self, head):
        dummy = ListNode(0, head)
        prev = dummy
        curr = head

        while curr:
            # Check if current node starts a duplicate block
            if curr.next and curr.val == curr.next.val:
                # Move curr to the end of the duplicate block
                while curr.next and curr.val == curr.next.val:
                    curr = curr.next
                # Skip entire duplicate block
                prev.next = curr.next
            else:
                prev = prev.next  # value is unique, keep it

            curr = curr.next

        return dummy.next

````python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution(object):
    def deleteDuplicates(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        dummy = ListNode()
        dummy.next = head
        prev = dummy
        curr = head
        while curr:
            curr_val = curr.val
            curr = curr.next
            if curr and curr.val == curr_val:
                while curr and curr.val == curr_val:
                    curr = curr.next
                prev.next = curr
            else:
                prev = prev.next
        return dummy.next
````
