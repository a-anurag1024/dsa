## ðŸ§  Intuition

A **Trie (Prefix Tree)** stores characters level by level.  
- Each node represents a character.
- A path from root â†’ node represents a prefix.
- Words sharing prefixes share the same path, saving space and time.

This allows:
- **Search** in `O(L)`
- **Prefix check** in `O(L)`
where `L` is the length of the word/prefix.

## Python Implementation
```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word: str) -> None:
        curr = self.root
        for ch in word:
            if ch not in curr.children:
                curr.children[ch] = TrieNode()
            curr = curr.children[ch]
        curr.is_end = True

    def search(self, word: str) -> bool:
        curr = self.root
        for ch in word:
            if ch not in curr.children:
                return False
            curr = curr.children[ch]
        return curr.is_end

    def startsWith(self, prefix: str) -> bool:
        curr = self.root
        for ch in prefix:
            if ch not in curr.children:
                return False
            curr = curr.children[ch]
        return True
```

# My Solution

````python
class Node:
    def __init__(self, val=None, is_end=False, nexts={}):
        self.val = val
        self.is_end = is_end
        self.nexts = nexts

class Trie(object):

    def __init__(self):
        self.root = Node(val="<s>", nexts={})        


    def insert(self, word):
        """
        :type word: str
        :rtype: None
        """
        curr = self.root
        for c in word:
            if c not in curr.nexts:
                curr.nexts[c] = Node(val=c, nexts={})
            curr = curr.nexts[c]
        curr.is_end = True
        

    def search(self, word):
        """
        :type word: str
        :rtype: bool
        """
        curr = self.root
        for c in word:
            if c not in curr.nexts: return False
            curr = curr.nexts[c]
        if curr.is_end: return True
        return False
        

    def startsWith(self, prefix):
        """
        :type prefix: str
        :rtype: bool
        """
        curr = self.root
        for c in prefix:
            if c not in curr.nexts: return False
            curr = curr.nexts[c]
        return True


# Your Trie object will be instantiated and called as such:
# obj = Trie()
# obj.insert(word)
# param_2 = obj.search(word)
# param_3 = obj.startsWith(prefix)
````
