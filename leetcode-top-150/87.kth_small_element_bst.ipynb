## ðŸ“Œ Problem Summary  
Given the root of a **Binary Search Tree (BST)** and an integer `k`, return the **k-th smallest** value in the tree.

Because BST's inorder traversal yields nodes in **sorted order**, the k-th smallest element is simply the k-th value in an inorder traversal.

**Output:** `1`

### Example 2

## ðŸš€ Key Idea  
### BST Property:  
Inorder traversal produces **sorted** output.

Thus, to find the k-th smallest:
- Perform inorder until you count k nodes  
- Return the k-th nodeâ€™s value  

Two main implementations:

### â­ Approach 1 â€” Recursive Inorder Traversal  
- Slow but simple  
- Stop early once k nodes are visited  

### â­ Approach 2 â€” Iterative Inorder (Stack)  
- Uses a controlled traversal  
- Runs in **O(h + k)** time (h = tree height)  
- Uses **O(h)** space  

(Optimal for large trees)

---

## ðŸ“ Short Approach (Interview-Ready)

1. Initialize an empty stack  
2. Push all left nodes from root  
3. While k > 0:  
   - Pop the top node â†’ this is the next smallest  
   - Decrement k  
   - If k == 0 â†’ return node.val  
   - Push all left nodes from node.right  
4. Answer is returned when k drops to 0

Time Complexity: **O(h + k)**  
Space Complexity: **O(h)**

class Solution:
    def kthSmallest(self, root, k):
        stack = []

        # push left chain
        while root:
            stack.append(root)
            root = root.left

        # process k nodes
        while k > 0:
            node = stack.pop()
            k -= 1
            if k == 0:
                return node.val

            # push left chain of right subtree
            right = node.right
            while right:
                stack.append(right)
                right = right.left

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def kthSmallest(self, root, k):
        """
        :type root: Optional[TreeNode]
        :type k: int
        :rtype: int
        """
        self.k_smallest = None
        def inorder(curr, ind_number):
            if not curr: return ind_number

            ind_number = inorder(curr.left, ind_number)
            if ind_number == k:
                self.k_smallest = curr.val
            ind_number = inorder(curr.right, ind_number + 1)
            return ind_number
        inorder(root, 1)
        return self.k_smallest
````
