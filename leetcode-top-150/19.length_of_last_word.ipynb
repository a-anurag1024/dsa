# üìù LeetCode 58 ‚Äî Length of Last Word (Easy)

## üß© Problem Summary

Given a string `s` consisting of **words and spaces**, return the **length of the last word** in the string.

A **word** is defined as a **maximal substring** consisting of non-space characters only.

---

## üß† Examples

### Example 1:
**Input:** `s = "Hello World"`  
**Output:** `5`  
**Explanation:** Last word ‚Üí `"World"`, length = 5

---

### Example 2:
**Input:** `s = "   fly me   to   the moon  "`  
**Output:** `4`  
**Explanation:** Last word ‚Üí `"moon"`, length = 4

---

### Example 3:
**Input:** `s = "luffy is still joyboy"`  
**Output:** `6`  
**Explanation:** Last word ‚Üí `"joyboy"`, length = 6

---

## ‚öôÔ∏è Approach (Simple String Traversal)

### üîπ Idea 1: Using `split()` (Pythonic way)
1. Use `split()` to separate words (automatically ignores extra spaces).  
2. Return the length of the last element in the list.

```python
def lengthOfLastWord(s):
    return len(s.split()[-1])
```

### üîπ Idea 2: Manual Traversal (Optimized O(1) space)

1. Start from the end of the string.
2. Skip trailing spaces.
3. Count characters until another space (or start of string) is found.

```python
def lengthOfLastWord(s):
    i = len(s) - 1
    length = 0

    # Skip trailing spaces
    while i >= 0 and s[i] == ' ':
        i -= 1

    # Count last word
    while i >= 0 and s[i] != ' ':
        length += 1
        i -= 1

    return length
```

‚úÖ Efficient ‚Äî single pass
‚úÖ No extra space
‚úÖ Works even for strings with trailing spaces

---

## üßÆ Complexity

| Type     | Complexity                                          |
| -------- | --------------------------------------------------- |
| ‚è±Ô∏è Time  | `O(n)`                                              |
| üß† Space | `O(1)` (for traversal) / `O(n)` (for split version) |



# My solution

````python
class Solution(object):
    def lengthOfLastWord(self, s):
        """
        :type s: str
        :rtype: int
        """
        last_word_start = -1
        last_word_length = 0
        for i in range(len(s)):
            if s[i] == " ":
                last_word_start = -1
            else:
                if last_word_start == -1:
                    last_word_start = i
                    last_word_length = 1
                else: last_word_length += 1
        return last_word_length
````
