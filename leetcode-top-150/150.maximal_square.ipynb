Youâ€™re given an `m Ã— n` **binary matrix** (`'0'` and `'1'`).

## ðŸ’¡ Core Insight

This is a **2D Dynamic Programming** problem.

For a cell to be the **bottom-right corner** of a square of side `k`, its:

* top cell
* left cell
* top-left (diagonal) cell

must all support squares of size `k-1`.

---

## ðŸ§  DP State Definition

Let:

```
dp[i][j] = side length of the largest square
           ending at cell (i, j)
```

---

## ðŸ” DP Transition

If `matrix[i][j] == '1'`:

```
dp[i][j] = 1 + min(
    dp[i-1][j],     # top
    dp[i][j-1],     # left
    dp[i-1][j-1]    # diagonal
)
```

Else:

```
dp[i][j] = 0
```

---

## ðŸ§± Base Case

* First row and first column:

  * `dp[i][j] = 1` if `matrix[i][j] == '1'`

---

## âœï¸ Short Approach Sketch

1. Create a DP table (or reuse input)
2. For each `'1'`, compute largest square ending there
3. Track maximum side length found
4. Return `max_sideÂ²`

## ðŸ§® Python Solution (Space Optimized â€” 1D DP)

```python
class Solution(object):
    def maximalSquare(self, matrix):
        if not matrix or not matrix[0]:
            return 0
        
        m, n = len(matrix), len(matrix[0])
        dp = [0] * (n + 1)
        max_side = 0
        prev = 0
        
        for i in range(1, m + 1):
            prev = 0
            for j in range(1, n + 1):
                temp = dp[j]
                if matrix[i - 1][j - 1] == '1':
                    dp[j] = 1 + min(dp[j], dp[j - 1], prev)
                    max_side = max(max_side, dp[j])
                else:
                    dp[j] = 0
                prev = temp
        
        return max_side * max_side
```

# My Solution

````python
class Solution:
    def maximalSquare(self, matrix: List[List[str]]) -> int:
        m, n = len(matrix), len(matrix[0])
        dp = [[0 for _ in range(n)] for _ in range(m)]
        max_sq = 0
        for i in range(m):
            if matrix[i][0] == "1":
                dp[i][0] = 1
                max_sq = 1
        for j in range(n):
            if matrix[0][j] == "1":
                dp[0][j] = 1
                max_sq = 1

        for i in range(1, m):
            for j in range(1, n):
                if matrix[i][j] == "1":
                    dp[i][j] = 1 + min(dp[i-1][j], dp[i-1][j-1], dp[i][j-1])
                    max_sq = max(max_sq, dp[i][j])

        return max_sq*max_sq
        
````
