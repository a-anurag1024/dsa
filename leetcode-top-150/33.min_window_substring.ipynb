## ðŸ§© Problem Summary

Given two strings `s` and `t`, find the **smallest window substring** in `s` that contains **all characters of `t`**, including duplicates.  
If no such window exists, return `""`.

---

## ðŸ§  Examples

### Example 1:
**Input:**  
`s = "ADOBECODEBANC"`, `t = "ABC"`  
**Output:** `"BANC"`  
**Explanation:**  
The substring `"BANC"` contains `'A'`, `'B'`, and `'C'`.

---

### Example 2:
**Input:**  
`s = "a"`, `t = "a"`  
**Output:** `"a"`  
**Explanation:**  
The entire string is a valid window.

---

### Example 3:
**Input:**  
`s = "a"`, `t = "aa"`  
**Output:** `""`  
**Explanation:**  
`t` needs two `'a'`s, but `s` has only one.

---

## âš™ï¸ Approach (Sliding Window + Frequency Matching)

### ðŸ”¹ Key Idea:
Use two pointers to create a **sliding window** over `s` that expands and contracts dynamically while maintaining a **count of required characters**.

The goal:
- Expand the right pointer to include all characters in `t`.
- Once all characters are covered, shrink from the left to minimize the window.

---

### ðŸ”¹ Step-by-Step Logic

1. Build a **frequency map** `need` for characters in `t`.  
   (Tracks how many of each character we still need.)
2. Use two pointers:  
   - `left` (window start)  
   - `right` (window end)
3. Expand `right` â†’ include characters in the window.  
   Decrement counts in `need` when a required char is found.
4. When all characters are matched:
   - Try shrinking from the left.
   - Update the minimum window size whenever a smaller valid window is found.
5. Continue until `right` reaches the end of `s`.

## ðŸ§‘â€ðŸ’» Python Sketch

```python
from collections import Counter

def minWindow(s, t):
    if not s or not t:
        return ""

    need = Counter(t)
    missing = len(t)
    left = start = end = 0

    for right, ch in enumerate(s, 1):  # right is 1-based for easier slicing
        if need[ch] > 0:
            missing -= 1
        need[ch] -= 1

        # When all characters are matched, shrink from the left
        if missing == 0:
            while left < right and need[s[left]] < 0:
                need[s[left]] += 1
                left += 1
            if end == 0 or right - left < end - start:
                start, end = left, right

            # Move left pointer ahead to find new potential window
            need[s[left]] += 1
            missing += 1
            left += 1

    return s[start:end]


# My solution

````python
from collections import Counter

class Solution(object):
    def minWindow(self, s, t):
        if not s or not t:
            return ""

        t_count = Counter(t)
        curr_count = Counter()
        required = len(t_count)
        formed = 0
        st = 0
        min_len = float('inf')
        min_substr = ""

        for en, ch in enumerate(s):
            if ch in t_count:
                curr_count[ch] += 1
                if curr_count[ch] == t_count[ch]:
                    formed += 1

            while formed == required:
                window_len = en - st + 1
                if window_len < min_len:
                    min_len = window_len
                    min_substr = s[st:en + 1]

                left_char = s[st]
                if left_char in t_count:
                    curr_count[left_char] -= 1
                    if curr_count[left_char] < t_count[left_char]:
                        formed -= 1
                st += 1

        return min_substr
````
