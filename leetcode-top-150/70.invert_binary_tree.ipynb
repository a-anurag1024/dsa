## üìå Problem Summary  
Given the root of a binary tree, **invert** it ‚Äî meaning swap the left and right children of every node ‚Äî and return the root.

This is conceptually the "mirror" of the tree.

---

## üîç Examples  

### Example 1  
**Input:** `[4,2,7,1,3,6,9]`  
**Output:** `[4,7,2,9,6,3,1]`

### Example 2  
**Input:** `[2,1,3]`  
**Output:** `[2,3,1]`

### Example 3  
**Input:** `[]`  
**Output:** `[]`

---

## üöÄ Key Idea  
Use a simple recursive DFS or iterative BFS/DFS approach to swap children of each node.

The problem is extremely similar to tree traversal but with one extra action:  
`swap(node.left, node.right)`.

---

## üìù Short Approach (Interview-Ready)

### **Recursive Approach**
1. If root is `None`, return `None`.  
2. Invert left subtree  
3. Invert right subtree  
4. Swap left and right  
5. Return root  

### **Iterative Approach (Queue or Stack)**
1. Use BFS or DFS  
2. For each node, swap children  
3. Push children into queue/stack  

Time Complexity: **O(n)**  
Space Complexity:  
- DFS: O(h) recursion  
- BFS: O(n) in worst case

## üßë‚Äçüíª Template Code (Recursive)
```python
class Solution:
    def invertTree(self, root):
        if not root:
            return root
        
        # invert children
        root.left, root.right = root.right, root.left
        
        # recurse
        self.invertTree(root.left)
        self.invertTree(root.right)
        
        return root

````python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
class Solution(object):
    def invertTree(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: Optional[TreeNode]
        """
        def helper(node):
            if not node: return None
            temp = node.left
            node.left = helper(node.right)
            node.right = helper(temp)
            return node
        return helper(root)
````
