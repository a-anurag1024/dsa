**Difficulty:** Medium  
**Topic:** Binary Trees / BFS  

---

## üìå Problem Summary  
Given the root of a binary tree, return its **level order traversal**.  
This means:  
- Visit nodes **level by level**  
- At each level, go **left to right**  
- Output each level as a separate list

---

## üîç Example  
**Input:** `[3,9,20,null,null,15,7]`  
**Output:**  
[
[3],
[9,20],
[15,7]
]


---

## üöÄ Key Idea  
Use **Breadth-First Search (BFS)** with a queue.

BFS naturally processes nodes level by level.  
For each level:
- Count how many nodes are in the queue  
- Pop exactly that many  
- Record their values  
- Push their children  

---

## üìù Short Approach (Interview-Ready)

1. If root is `None`, return empty list  
2. Push root to a queue  
3. While queue is not empty:  
   - Determine `level_size = len(queue)`  
   - Loop `level_size` times:  
     - Pop a node  
     - Add its value to current level list  
     - Push its left and right children  
   - Append current level list to the result  

Time Complexity: **O(n)**  
Space Complexity: **O(n)** (queue at widest level)

---

## üßë‚Äçüíª Template Code (BFS)

```python
from collections import deque

class Solution:
    def levelOrder(self, root):
        if not root:
            return []

        q = deque([root])
        result = []

        while q:
            level_size = len(q)
            level = []

            for _ in range(level_size):
                node = q.popleft()
                level.append(node.val)

                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)

            result.append(level)

        return result

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
from collections import deque 

class Solution(object):
    def levelOrder(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: List[List[int]]
        """
        traversal = []
        next_nodes = deque()
        if not root:
            return traversal
        next_nodes.append(root)
        while next_nodes:
            n = len(next_nodes)
            level_traversal = []
            for _ in range(n):
                curr = next_nodes.popleft()
                level_traversal.append(curr.val)
                if curr.left: next_nodes.append(curr.left)
                if curr.right: next_nodes.append(curr.right)
            traversal.append(level_traversal)
        return traversal
````
