## ðŸ“ Problem Summary
Given an integer array `citations`, where `citations[i]` is the number of citations for the i-th paper,  
return the **h-index** of the researcher.

**Definition:**  
The *h-index* is the **maximum value `h`** such that the researcher has published **at least `h` papers** with **â‰¥ `h` citations each**.

---

## ðŸ’¡ Example
**Input:**  
`citations = [3,0,6,1,5]`  
**Output:** `3`  

**Explanation:**  
- Papers sorted by citation count: `[0,1,3,5,6]`  
- 3 papers have at least 3 citations â†’ **h = 3**

---

## âš™ï¸ Approach 1 â€” Sorting-Based (Simple & Intuitive)
1. Sort `citations` in **descending order**.  
2. Traverse through the sorted list:
   - For each paper at index `i` (0-based), check if `citations[i] >= i + 1`.  
   - The largest valid `i + 1` is your **h-index**.
3. Return that value.

## âœ… Code Snippet (Python)
```python
def hIndex(citations):
    citations.sort(reverse=True)
    h = 0
    for i, c in enumerate(citations):
        if c >= i + 1:
            h = i + 1
        else:
            break
    return h


# My Solution

````python
class Solution(object):
    def hIndex(self, citations):
        """
        :type citations: List[int]
        :rtype: int
        """
        n = len(citations)
        greater_than_map = {i:0 for i in range(n+1)}
        for c in citations:
            if c >= n: greater_than_map[n] += 1
            else:
                greater_than_map[c] += 1
        papers_cumm = 0
        for k in range(n,-1,-1):
            papers_cumm += greater_than_map[k]
            if papers_cumm >= k: return k
        return 0
````
