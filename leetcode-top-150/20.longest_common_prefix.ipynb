# üå∏ LeetCode 14 ‚Äî Longest Common Prefix (Easy)

---

## üß© Problem Summary

Given an array of strings `strs`, return the **longest common prefix** shared by all strings.  
If there is **no common prefix**, return an empty string `""`.

---

## üß† Examples

### Example 1:
**Input:** `["flower", "flow", "flight"]`  
**Output:** `"fl"`  
**Explanation:**  
Common prefix among all strings = `"fl"`

---

### Example 2:
**Input:** `["dog", "racecar", "car"]`  
**Output:** `""`  
**Explanation:**  
No common prefix ‚Üí return empty string

---

## ‚öôÔ∏è Approach (Horizontal Scanning)

### üîπ Idea:
Start with the **first word** as the initial prefix.  
Iteratively compare it with the next string and **shrink** the prefix until it matches.

---

### üîπ Step-by-Step

1. Initialize `prefix = strs[0]`
2. For each subsequent string:
   - While the string **does not start** with `prefix`, shorten `prefix` by removing the last character.
3. Return the final `prefix`.

If `prefix` becomes empty at any point ‚Üí return `""`.

---

## üßë‚Äçüíª Python Sketch

```python
def longestCommonPrefix(strs):
    if not strs:
        return ""
    
    prefix = strs[0]
    for s in strs[1:]:
        while not s.startswith(prefix):
            prefix = prefix[:-1]
            if not prefix:
                return ""
    return prefix


# My solution

````python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        common_prefix = ""
        i = 0
        min_len = min([len(s) for s in strs])
        while strs and i < min_len:
            c = strs[0][i]
            for j in range(1, len(strs)):
                if strs[j][i] != c:
                    return common_prefix
            common_prefix += c
            i += 1
        return common_prefix
````
