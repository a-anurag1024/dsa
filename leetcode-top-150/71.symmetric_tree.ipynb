## üìå Problem Summary  
Given the root of a binary tree, determine whether the tree is **symmetric**, meaning the left subtree is a mirror reflection of the right subtree.

A tree is symmetric if:
- Values match in mirror positions  
- Left subtree‚Äôs left matches right subtree‚Äôs right  
- Left subtree‚Äôs right matches right subtree‚Äôs left  

---

## üîç Examples  

### Example 1  
**Input:** `[1,2,2,3,4,4,3]`  
**Output:** `true`

### Example 2  
**Input:** `[1,2,2,null,3,null,3]`  
**Output:** `false`

---

## üöÄ Key Idea  
Check if the **left** and **right** subtrees are *mirror images*.

### Mirror condition:
For two nodes `a` and `b`:
1. Both null ‚Üí ‚úîÔ∏è symmetric  
2. One null ‚Üí ‚ùå asymmetric  
3. Values differ ‚Üí ‚ùå  
4. Recursively check:
   - `a.left`  vs `b.right`  
   - `a.right` vs `b.left`  

---

## üìù Short Approach (Interview-Ready)

### **Recursive Mirror Function**
1. Base cases:
   - Both nodes null ‚Üí return True  
   - One null ‚Üí return False  
   - Values differ ‚Üí return False  
2. Recursively compare:  

isMirror(a.left, b.right) AND isMirror(a.right, b.left)


### **Iterative BFS/Queue Approach**
Push pairs of nodes into a queue.  
For each pair, check:
- both null ‚Üí continue  
- one null ‚Üí fail  
- values differ ‚Üí fail  
- enqueue mirror children pairs  

Time Complexity: **O(n)**  
Space Complexity: **O(h)** or **O(n)** depending on recursion/queue

## üßë‚Äçüíª Template Code (Recursive)
```python
class Solution:
 def isSymmetric(self, root):
     def isMirror(a, b):
         if not a and not b:
             return True
         if not a or not b or a.val != b.val:
             return False
         return isMirror(a.left, b.right) and isMirror(a.right, b.left)
     
     return isMirror(root.left, root.right)

````python
# Although this solution is of the same time and space complexity, it is little more wasteful in terms of space

# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
class Solution(object):
    def isSymmetric(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: bool
        """
        curr = [root]
        while True:
            vals = []
            next_l = []
            for node in curr:
                vals.append(node.val if node else None)
                if not node: next_l.extend([None, None])
                else: next_l.extend([node.left, node.right])
            i = 0
            j = len(vals)-1
            while i < j:
                if vals[i] != vals[j]: return False
                i += 1
                j -= 1
            if any(next_l):
                curr = next_l
            else: break
        return True
````
