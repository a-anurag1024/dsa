## ðŸ§  State Definition (Key)


## ðŸ§® Python Solution (Optimal O(n), O(1))

```python
class Solution(object):
    def maxProfit(self, prices):
        buy1 = float('-inf')
        buy2 = float('-inf')
        sell1 = 0
        sell2 = 0
        
        for p in prices:
            buy1 = max(buy1, -p)
            sell1 = max(sell1, buy1 + p)
            buy2 = max(buy2, sell1 - p)
            sell2 = max(sell2, buy2 + p)
        
        return sell2
```
