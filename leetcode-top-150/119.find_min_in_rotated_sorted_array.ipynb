### **Problem Summary**

You are given a **sorted array of unique integers** that has been **rotated** an unknown number of times.
Return the **minimum element** in the array.

**Constraint:** `O(log n)` time.

---

### **Key Observations**

* The array consists of **two sorted segments**
* The minimum element is the **pivot point**
* At least one half of the array is always sorted
* Binary search can be adapted using comparisons with boundaries

---

## ðŸ”¹ Short Approach Sketch (Binary Search on Pivot)

1. Initialize `left = 0`, `right = n - 1`
2. While `left < right`:

   * Compute `mid`
   * If `nums[mid] > nums[right]`:

     * Minimum lies in the **right half**
     * Move `left = mid + 1`
   * Else:

     * Minimum lies in the **left half or at mid**
     * Move `right = mid`
3. When loop ends, `left == right`
4. Return `nums[left]`

---

### **Time & Space Complexity**

* **Time:** `O(log n)`
* **Space:** `O(1)`

---

## ðŸ”¹ Python Solution (Optimal)

```python
class Solution(object):
    def findMin(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        left, right = 0, len(nums) - 1

        while left < right:
            mid = (left + right) // 2

            if nums[mid] > nums[right]:
                left = mid + 1
            else:
                right = mid

        return nums[left]
```

````python
class Solution(object):
    def findMin(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        st, en = 0, len(nums) - 1
        while st <= en:
            mid = (st + en) // 2
            if mid !=0 and nums[mid-1] > nums[mid]: return nums[mid]
            if nums[mid] < nums[en]: en = mid - 1
            else: st = mid + 1
        return nums[0]
````
