## ðŸ“Œ Problem Summary  
Given an integer array `nums` and an integer `target`, return the **indices** of the two numbers such that:

nums[i] + nums[j] = target


Constraints:
- Exactly **one** solution always exists.
- You cannot use the **same element twice**.
- Order of output does not matter.

---

## ðŸ§  Key Insight  
As you scan through the list, you only need to know if the **complement**  
(`target - nums[i]`) has been seen before.

A **hash map (dictionary)** gives O(1) lookup time.

---

## âœ… Short Approach (Optimal â€” One Pass Hash Map)

### **Steps**
1. Create hashmap: `value â†’ index`.
2. Iterate through array:
   - For each number `x`, compute its complement `target - x`.
   - If complement exists in map â†’ return its index + current index.
   - Else store `x` in map for future lookups.

### Complexity
- **Time:** `O(n)`
- **Space:** `O(n)`

## ðŸ”§ Python Code (Best Solution)

```python
class Solution:
    def twoSum(self, nums, target):
        hm = {}  # value -> index
        
        for i, num in enumerate(nums):
            comp = target - num
            if comp in hm:
                return [hm[comp], i]
            hm[num] = i

````python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        hm = {}
        for i in range(len(nums)):
            if target - nums[i] not in hm:
                hm[nums[i]] = i
            else:
                return [hm[target-nums[i]], i]
````
