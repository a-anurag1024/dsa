## üìù Problem Summary
Given a **sorted** integer array `nums`, remove duplicates **in-place** such that **each unique element appears at most twice**.  
The **relative order** must be preserved, and no extra array can be used.

Return the number of elements `k` such that the first `k` positions of `nums` contain the allowed elements.

---

## üí° Example
**Input:**  
`nums = [1,1,1,2,2,3]`  
**Output:**  
`k = 5, nums = [1,1,2,2,3,_]`

**Input:**  
`nums = [0,0,1,1,1,1,2,3,3]`  
**Output:**  
`k = 7, nums = [0,0,1,1,2,3,3,_,_]`

---

## ‚öôÔ∏è Approach (Two-Pointer Technique)
1. Use a pointer `i` to track the **next position** to place a valid element.  
2. Iterate through the array with `n`:
   - For each `nums[n]`, check if it can appear again:
     - If `i < 2` (first two positions) ‚Üí always place it.
     - Else, if `nums[n] != nums[i-2]` ‚Üí safe to place (ensures ‚â§ 2 occurrences).
3. Assign `nums[i] = nums[n]` and increment `i`.
4. After iteration, return `i` as the count of valid elements.

---

## ‚è±Ô∏è Complexity
- **Time:** O(n) ‚Äî single pass.  
- **Space:** O(1) ‚Äî in-place, no extra storage.

## ‚úÖ Code Snippet (Python)
```python
def removeDuplicates(nums):
    i = 0
    for n in nums:
        if i < 2 or n != nums[i - 2]:
            nums[i] = n
            i += 1
    return i


# My original Solution

````python
class Solution(object):
    def removeDuplicates(self, nums):
        if not nums: 
            return 0
        i = 0
        is_double_counted = False
        for j in range(1, len(nums)):
            if nums[j] == nums[i]:
                if not is_double_counted:
                    i += 1
                    nums[i] = nums[j]
                    is_double_counted = True
            else:
                i += 1
                nums[i] = nums[j]
                is_double_counted = False
        return i + 1
````
