# üß¨ LeetCode 433: Minimum Genetic Mutation ‚Äî Quick Review

## üìò Problem Summary
You are given:
- `startGene` ‚Üí starting gene string (length = 8)
- `endGene` ‚Üí target gene string
- `bank` ‚Üí list of valid gene strings

Rules:
- A **mutation** changes exactly **one character**
- Each intermediate gene **must exist in `bank`**
- Characters allowed: `'A'`, `'C'`, `'G'`, `'T'`

Goal:
‚û°Ô∏è Return the **minimum number of mutations** to reach `endGene`  
‚û°Ô∏è If impossible ‚Üí return `-1`

---

## üß† Key Insight
This is a **shortest path problem in an unweighted graph**.

- Each gene ‚Üí **node**
- Edge exists if two genes differ by **exactly one character**
- Each mutation costs **1 step**

üëâ **Breadth-First Search (BFS)** guarantees the minimum number of mutations.

---

## üöÄ Approach (BFS)
### Steps:
1. Put all genes in `bank` into a set for `O(1)` lookup.
2. Start BFS from `startGene`.
3. For each gene:
   - Try all possible **single-character mutations**
   - If the new gene exists in `bank` and is unvisited:
     - Add to queue
     - Mark as visited
4. Stop when `endGene` is reached.
5. If BFS ends without finding it ‚Üí return `-1`.

---

## üß© Time & Space Complexity
- **Time:** `O(N √ó 8 √ó 4)`  
  where `N` = size of bank  
- **Space:** `O(N)` for queue + visited set

---

## üìù Sample BFS Code (Python)
```python
from collections import deque

class Solution:
    def minMutation(self, startGene, endGene, bank):
        bank = set(bank)
        if endGene not in bank:
            return -1
        
        q = deque([(startGene, 0)])
        visited = set([startGene])
        genes = ['A', 'C', 'G', 'T']
        
        while q:
            curr, steps = q.popleft()
            if curr == endGene:
                return steps
            
            for i in range(8):
                for g in genes:
                    if g != curr[i]:
                        nxt = curr[:i] + g + curr[i+1:]
                        if nxt in bank and nxt not in visited:
                            visited.add(nxt)
                            q.append((nxt, steps + 1))
        
        return -1

````python
from collections import deque

class Solution(object):
    def minMutation(self, startGene, endGene, bank):
        """
        :type startGene: str
        :type endGene: str
        :type bank: List[str]
        :rtype: int
        """
        characters = {'A', 'T', 'G', 'C'}
        q = deque([(startGene, 0)])
        visited = set()
        while q:
            state, num = q.popleft()
            if state == endGene: return num
            for i in range(8):
                current_char = state[i]
                for next_char in list(characters-set(current_char)):
                    new_gene = state[:i] + next_char + state[i+1:]
                    if new_gene in visited: continue
                    elif new_gene not in bank: continue
                    else:
                        q.append((new_gene, num+1))
                        visited.add(new_gene)
        return -1
````
