## üìå Problem Summary  
Given the roots of two binary trees `p` and `q`, determine if the two trees are **identical**.

Two trees are the same if:
1. They have the **same structure**  
2. All corresponding nodes have the **same value**

---

## üîç Examples  

### Example 1  
**Input:** `p = [1,2,3], q = [1,2,3]`  
**Output:** `true`

### Example 2  
**Input:** `p = [1,2], q = [1,null,2]`  
**Output:** `false`  
(The structure differs)

### Example 3  
**Input:** `p = [1,2,1], q = [1,1,2]`  
**Output:** `false`  
(Same structure, but values differ)

---

## üöÄ Key Idea  
Perform a **recursive comparison**:

Two nodes are equal if:
- Both are `None` ‚Üí trees match at this position  
- Both are not `None`, values match, and  
  - left subtrees match  
  - right subtrees match  

This is a typical DFS tree traversal problem.

---

## üìù Short Approach (Interview-Ready)

Recursive check:
1. If both nodes are `None` ‚Üí return True  
2. If only one is `None` ‚Üí return False  
3. If values differ ‚Üí return False  
4. Recurse on:
   - left children  
   - right children  
5. Return the AND of both results  

Time Complexity: **O(n)**  
Space Complexity: **O(h)** recursion stack, where `h` = tree height  

---

## üßë‚Äçüíª Template Code (Simple DFS)
```python
class Solution:
    def isSameTree(self, p, q):
        if not p and not q:
            return True
        if not p or not q:
            return False
        if p.val != q.val:
            return False

        return (self.isSameTree(p.left, q.left) and
                self.isSameTree(p.right, q.right))


# My Solution

````python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
    
class Solution(object):
    def isSameTree(self, p, q):
        """
        :type p: Optional[TreeNode]
        :type q: Optional[TreeNode]
        :rtype: bool
        """
        def traversal(p_node, q_node):
            if p_node == None or q_node == None:
                if p_node == q_node: return True
                else: return False
            if p_node.val != q_node.val:
                return False
            return traversal(p_node.left, q_node.left) and traversal(p_node.right, q_node.right)
        return traversal(p, q)
        
````
