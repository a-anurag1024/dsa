## üìå Problem Summary  
Given the **root** of a binary tree, return its **maximum depth**, defined as the length of the longest path from the root down to any leaf node.

---

## üîç Examples  

### Example 1  
**Input:** `[3,9,20,null,null,15,7]`  
**Output:** `3`  
(Depth path: `3 ‚Üí 20 ‚Üí 15` or `3 ‚Üí 20 ‚Üí 7`)

### Example 2  
**Input:** `[1,null,2]`  
**Output:** `2`

---

## üöÄ Key Idea  
Maximum depth can be computed using:

### ‚úÖ **DFS (Recursion)**  
Depth =  
1 + max(depth(left subtree), depth(right subtree))


### OR

### ‚úÖ **BFS (Level Order Traversal)**  
Count number of levels in the tree.

Both methods take **O(n)** time.

---

## üìù Short Approach (Interview-Ready)

### **Recursive DFS Approach**
1. If `root` is `None`, depth is `0`.  
2. Recursively compute:  

left_depth = depth(root.left)

right_depth = depth(root.right)

return 1 + max(left_depth, right_depth)


### **Iterative BFS Approach (Optional)**
1. Use a queue starting at root.  
2. For each level, process all nodes and increment depth counter.  

Time Complexity: **O(n)**  
Space Complexity:  
- DFS: O(h), where h = height  
- BFS: O(n) worst case

## üßë‚Äçüíª Template Code (DFS - Simple & Clean)
```python
class Solution:
 def maxDepth(self, root):
     if not root:
         return 0
     return 1 + max(self.maxDepth(root.left),
                    self.maxDepth(root.right))


# My Solution

````python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
class Solution(object):
    def maxDepth(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """
        def helper(node, curr_level=0):
            if node is None:
                return curr_level
            left_max = helper(node.left, curr_level+1)
            right_max = helper(node.right, curr_level+1)
            return max(left_max, right_max)
        return helper(root)
````
