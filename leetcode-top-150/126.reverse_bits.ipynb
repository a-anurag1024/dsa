## âœï¸ Short Approach Sketch (Bit Manipulation)

1. Initialize `result = 0`
2. Repeat 32 times:

   * Shift `result` left by 1
   * Add the least significant bit of `n` (`n & 1`)
   * Shift `n` right by 1
3. Return `result`

This simulates reversing bits one by one.

## ğŸ§© Python Solution (Bitwise â€“ Interview Preferred)

```python
class Solution(object):
    def reverseBits(self, n):
        res = 0
        for _ in range(32):
            res = (res << 1) | (n & 1)
            n >>= 1
        return res
```

````python
class Solution(object):
    def reverseBits(self, n):
        """
        :type n: int
        :rtype: int
        """
        def dec_to_bin(num):
            b = []
            curr = num
            while curr > 0:
                b.append(str(curr % 2))
                curr = curr // 2
            return "".join(b[-1::-1])
        def bin_to_dec(num):
            n = len(num)
            d = 0
            for i in range(n):
                d += int(num[i]) * (2**(n-i-1))
            return d
        bin_num = dec_to_bin(n)
        padded_bin_num = "0"*(32-len(bin_num)) + bin_num
        return bin_to_dec(padded_bin_num[-1::-1])
````
