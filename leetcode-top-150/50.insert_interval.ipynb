## ðŸ“˜ Problem Summary
You are given:
- A list of **non-overlapping**, **sorted** intervals.
- A `newInterval` that must be inserted into this list.

Your job:
Insert `newInterval` into the intervals list while:
- Preserving sorted order  
- Merging any overlapping intervals  
- Ensuring the result has **no overlaps**

---

## ðŸ§  Key Idea
We can treat the insertion as a **3-phase scan**:

1. **Add all intervals ending before the new interval starts**  
   (`interval[1] < new_start`)

2. **Merge all intervals overlapping with the new interval**  
   (`interval[0] <= new_end`)  
   Update:  
   - `new_start = min(new_start, interval[0])`  
   - `new_end   = max(new_end,   interval[1])`

3. **Add the remaining intervals**  
   (intervals starting after new interval ends)

This avoids re-sorting and works in **O(n)**.

---

## ðŸ§© Approach Sketch
1. Initialize result list.
2. Push all intervals ending before `newInterval` into result.
3. Merge all intervals overlapping with `newInterval`.
4. Push the merged `newInterval`.
5. Push all remaining intervals.
6. Return result.

---

## ðŸ“ Example
Input: `[[1,3],[6,9]]`, newInterval = `[2,5]`  
Output: `[[1,5],[6,9]]`

---

## ðŸ§© Sample Python Implementation
```python
class Solution(object):
    def insert(self, intervals, newInterval):
        res = []
        new_start, new_end = newInterval
        i = 0
        n = len(intervals)

        # 1. Add intervals before newInterval
        while i < n and intervals[i][1] < new_start:
            res.append(intervals[i])
            i += 1

        # 2. Merge overlapping intervals
        while i < n and intervals[i][0] <= new_end:
            new_start = min(new_start, intervals[i][0])
            new_end = max(new_end, intervals[i][1])
            i += 1

        # Push merged newInterval
        res.append([new_start, new_end])

        # 3. Add the remaining intervals
        while i < n:
            res.append(intervals[i])
            i += 1

        return res

````python
class Solution(object):
    def insert(self, intervals, newInterval):
        if not intervals:
            return [newInterval]

        new_start, new_end = newInterval
        n = len(intervals)

        left, right = 0, n
        while left < right:
            mid = (left + right) // 2
            if intervals[mid][0] < new_start:
                left = mid + 1
            else:
                right = mid
        insert_pos = left   # this is where newInterval would fit by start

        res = intervals[:insert_pos]

        if res and res[-1][1] >= new_start:   # overlaps previous
            res[-1][1] = max(res[-1][1], new_end)
        else:
            res.append([new_start, new_end])

        for i in range(insert_pos, n):
            curr = intervals[i]
            if res[-1][1] >= curr[0]:  # overlap
                res[-1][1] = max(res[-1][1], curr[1])
            else:
                res.append(curr)

        return res
````
