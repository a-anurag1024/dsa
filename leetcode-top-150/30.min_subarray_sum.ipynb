## âš™ï¸ Approach (Sliding Window / Two-Pointer)

### ðŸ”¹ Key Idea:
Since all numbers are **positive**, we can use a **sliding window** to maintain a growing sum and shrink it as soon as it exceeds the target.

---

### ðŸ”¹ Step-by-Step Logic

1. Initialize:
   - `left = 0` (window start)
   - `curr_sum = 0`
   - `min_len = inf`
2. Iterate `right` from `0` to `len(nums) - 1`:
   - Add `nums[right]` to `curr_sum`.
   - While `curr_sum >= target`:
     - Update `min_len = min(min_len, right - left + 1)`
     - Shrink the window by subtracting `nums[left]` and moving `left += 1`.
3. After the loop:
   - If `min_len` is still infinity â†’ return `0` (no subarray found)
   - Otherwise, return `min_len`.

## ðŸ§‘â€ðŸ’» Python Sketch

```python
def minSubArrayLen(target, nums):
    left = 0
    curr_sum = 0
    min_len = float('inf')

    for right in range(len(nums)):
        curr_sum += nums[right]

        while curr_sum >= target:
            min_len = min(min_len, right - left + 1)
            curr_sum -= nums[left]
            left += 1

    return 0 if min_len == float('inf') else min_len


# My solution

````python
class Solution(object):
    def minSubArrayLen(self, target, nums):
        """
        :type target: int
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)
        lead = trail = 0
        min_len = n + 1
        curr_sum = nums[0]
        while lead<n and trail<n:
            if curr_sum < target:
                lead += 1
                curr_sum += nums[lead] if lead < n else 0
            else:
                curr_len = lead - trail + 1
                if curr_len < min_len : min_len = curr_len
                trail += 1
                curr_sum -= nums[trail-1]
        return min_len if min_len <= n else 0
````
