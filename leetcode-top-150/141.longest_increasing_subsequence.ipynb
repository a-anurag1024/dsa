## ğŸ§  Approach 1: DP (Conceptually Simple)

### DP State

* `dp[i]` = length of LIS **ending at index `i`**

### Transition

For every `j < i`:

```
if nums[j] < nums[i]:
    dp[i] = max(dp[i], dp[j] + 1)
```

### Base Case

```

```

### Result

```

```

### âœï¸ Short Approach Sketch (DP)

1. Initialize `dp` with all `1`s
2. For each `i`, check all `j < i`
3. Update `dp[i]` if increasing condition holds
4. Return the maximum value in `dp`

### ğŸ§® Python Solution (O(nÂ²) DP)

```python
class Solution(object):
    def lengthOfLIS(self, nums):
        if not nums:
            return 0
        
        n = len(nums)
        dp = [1] * n
        
        for i in range(n):
            for j in range(i):
                if nums[j] < nums[i]:
                    dp[i] = max(dp[i], dp[j] + 1)
        
        return max(dp)
```

### âœï¸ Short Approach Sketch (Optimized)

1. Initialize empty `tails`
2. For each number:

   * Binary search its position in `tails`
   * Replace or append
3. Return `len(tails)`

### ğŸ§® Python Solution (O(n log n))

```python
import bisect

class Solution(object):
    def lengthOfLIS(self, nums):
        tails = []
        
        for num in nums:
            idx = bisect.bisect_left(tails, num)
            if idx == len(tails):
                tails.append(num)
            else:
                tails[idx] = num
        
        return len(tails)
```
