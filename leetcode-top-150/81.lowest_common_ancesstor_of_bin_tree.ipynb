## üìå Problem Summary  
Given the root of a binary tree and two nodes `p` and `q`, return their **Lowest Common Ancestor (LCA)**.

The LCA is defined as the **lowest** (i.e., deepest) node that has **both p and q as descendants**, allowing a node to be a descendant of itself.

---

## üîç Examples  

### Example 1  
**Input:**  
`root = [3,5,1,6,2,0,8,null,null,7,4], p = 5, q = 1`  
**Output:** `3`  

### Example 2  
**Input:**  
`p = 5, q = 4`  
**Output:** `5`  
Because a node can be its own descendant.

### Example 3  
**Input:**  
`root = [1,2], p = 1, q = 2`  
**Output:** `1`

---

## üöÄ Key Idea  
Perform a **postorder DFS** (left ‚Üí right ‚Üí node).  
At each node:

1. If node is **None**, return None  
2. If node is **p** or **q**, return node  
3. Recursively search left and right subtrees  
4. Combine results:
   - If **both** left and right return non-null ‚Üí this node is the LCA  
   - Else return whichever side is non-null  

This works because:
- A node returns itself upward if it is p or q  
- Once both p and q appear in different subtrees, their LCA is the first node where this happens  

### Time Complexity:  
**O(n)** ‚Äî every node visited at most once  

### Space Complexity:  
**O(h)** recursion stack (h = tree height)

## üßë‚Äçüíª Template Code (Recursive DFS)

```python
class Solution:
    def lowestCommonAncestor(self, root, p, q):
        if not root or root == p or root == q:
            return root
        
        left = self.lowestCommonAncestor(root.left, p, q)
        right = self.lowestCommonAncestor(root.right, p, q)
        
        if left and right:
            return root  # p and q found in different branches
        
        return left if left else right

````python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        
class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        ancestor = None
        def dfs(curr):
            global ancestor
            if not curr:
                return False, False, None
            cp, cq = curr == p, curr == q
            lp, lq, found_l = dfs(curr.left)
            rp, rq, found_r = dfs(curr.right)
            found = (found_l or found_r)
            if not found:
                if (lp and rq) or (lq and rp): 
                    found = curr
                if (cp and (rq or lq)) or (cq and (rp or lp)): 
                    found = curr
            contains_p = cp or lp or rp
            contains_q = cq or lq or rq
            return contains_p, contains_q, found
        _, _, ancestor = dfs(root)
        return ancestor
        
````
