## ðŸ“Œ Problem Summary  
Given an array of strings, group together words that are **anagrams** of each other.  
You may return the groups in **any order**.

Anagrams share:
- the same characters  
- with the same frequencies  
- but possibly in different order  

---

# ðŸ§  Key Idea  
Use a **hash map** where the key uniquely represents the character frequency of a word.

### Two common ways to create the key:

### **1. Sorted string (simple & effective)**
Sort each word:
- `"eat"` â†’ `"aet"`
- `"tea"` â†’ `"aet"`

Use the sorted word as key.

### **2. Frequency count tuple (faster for long strings)**
For each word, build a 26-length frequency array:
- `"eat"` â†’ `(1,0,0,0,1,0,0,...,1)` â†’ tuple  
Use tuple as key.

Both methods produce correct anagram groups.

---

# âœ… Short Approach

### **Approach: Hashmap grouping**
1. Initialize dictionary: `anagrams = {}`  
2. For each word:
   - Sort the word (or compute freq tuple)
   - Use this as key in hashmap
   - Append the word to its group  
3. Return all hashmap values.

### Complexity  
#### Using sorted key:
- **Time:** `O(n * k log k)`  
- **Space:** `O(nk)`  
(k = max length of any string)

#### Using frequency tuple:
- **Time:** `O(n * k)`  
- **Space:** `O(nk)`

# ðŸ”§ Python Code (Method 1 â€” Sorted Key)

```python
class Solution:
    def groupAnagrams(self, strs):
        groups = {}
        
        for word in strs:
            key = ''.join(sorted(word))
            if key not in groups:
                groups[key] = []
            groups[key].append(word)
        
        return list(groups.values())
```

class Solution:
    def groupAnagrams(self, strs):
        groups = {}
        
        for word in strs:
            freq = [0] * 26
            for ch in word:
                freq[ord(ch) - ord('a')] += 1
            key = tuple(freq)
            
            if key not in groups:
                groups[key] = []
            groups[key].append(word)
        
        return list(groups.values())
