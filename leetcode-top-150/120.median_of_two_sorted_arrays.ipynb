## ðŸ”¹ Quick Review: LeetCode 4 â€” Median of Two Sorted Arrays

### **Problem Summary**

Given two **sorted arrays** `nums1` and `nums2`, return the **median** of the combined dataset.

* Arrays can have **different sizes**
* Must run in **`O(log(m + n))`**
* You **cannot** merge the arrays explicitly

---

### **Key Observations**

* Brute-force merge â†’ `O(m + n)` âŒ (not allowed)
* Median depends on **balanced partitioning**
* We can binary-search **only the smaller array**
* Goal: split both arrays so that:

  * Left half has `(m + n + 1) // 2` elements
  * All left elements â‰¤ all right elements

---

## ðŸ”¹ Short Approach Sketch (Binary Search on Partition)

1. Always binary search on the **smaller array**
2. Choose partition index `i` in `nums1`
3. Compute corresponding partition `j` in `nums2`
4. Check validity:

   * `nums1[i-1] â‰¤ nums2[j]`
   * `nums2[j-1] â‰¤ nums1[i]`
5. Once valid:

   * If total length is odd â†’ median = `max(left)`
   * If even â†’ median = average of `max(left)` and `min(right)`
6. Adjust binary search range if partition is invalid

---

### **Time & Space Complexity**

* **Time:** `O(log(min(m, n)))`
* **Space:** `O(1)`

---

## ðŸ”¹ Python Solution (Optimal & Interview-Standard)

```python
class Solution(object):
    def findMedianSortedArrays(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: float
        """
        # Ensure nums1 is the smaller array
        if len(nums1) > len(nums2):
            nums1, nums2 = nums2, nums1

        m, n = len(nums1), len(nums2)
        left, right = 0, m

        while left <= right:
            i = (left + right) // 2
            j = (m + n + 1) // 2 - i

            maxLeft1 = float('-inf') if i == 0 else nums1[i - 1]
            minRight1 = float('inf') if i == m else nums1[i]

            maxLeft2 = float('-inf') if j == 0 else nums2[j - 1]
            minRight2 = float('inf') if j == n else nums2[j]

            if maxLeft1 <= minRight2 and maxLeft2 <= minRight1:
                if (m + n) % 2 == 1:
                    return float(max(maxLeft1, maxLeft2))
                return (
                    max(maxLeft1, maxLeft2) +
                    min(minRight1, minRight2)
                ) / 2.0
            elif maxLeft1 > minRight2:
                right = i - 1
            else:
                left = i + 1
```

## ðŸ”¹ Why This Works

* Partition ensures **equal-sized halves**
* Binary search efficiently adjusts partition point
* Only local comparisons are required
* No full merge needed
